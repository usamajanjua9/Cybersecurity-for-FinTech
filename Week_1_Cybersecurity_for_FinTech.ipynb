##  1. Introduction to the Field

---

### **1.1 What is Cybersecurity?** 

**Definition:**
Cybersecurity is the practice of protecting systems, networks, and data from digital attacks, damage, or unauthorized access.

**Scope:**

* Protects **hardware** (servers, ATMs, mobile devices)
* Protects **software** (apps, banking systems, wallets)
* Protects **data** (transactions, customer details, account info)

**Importance:**

* Prevents financial loss 
* Maintains customer trust 
* Ensures compliance with regulations

### **1.2 Role of Cybersecurity in Digital Finance & FinTech** 

* **FinTech systems** (mobile wallets, online banking, digital KYC, crypto exchanges) are prime targets for attackers.
* Billions are lost every year due to fraud, scams, and breaches.
* Without strong cybersecurity, customers wont trust online financial systems.

**Examples in FinTech:**

*  Mobile Wallets  Need PIN + encryption for protection
*  Online Payments  Require secure protocols (HTTPS, TLS)
*  Crypto Exchanges  At risk of hacks (Mt. Gox, Binance)

### **1.3 Cybersecurity vs Information Security vs Privacy** 

| Term                        | Focus                                           | Example in FinTech                                                 |
| --------------------------- | ----------------------------------------------- | ------------------------------------------------------------------ |
| **Cybersecurity**         | Protects digital systems, networks, apps        | Stopping a hacker from breaching a payment gateway                 |
| **Information Security**  | Protects *all information* (digital + physical) | Locking up printed financial reports                               |

# **2. CIA Triad  The Pillars of Security**

The **CIA Triad** is the foundation of all cybersecurity.
It stands for **Confidentiality , Integrity , and Availability **.
Think of it as the **three legs of a stool**   if one breaks, the system falls.

---

##  **2.1 Overview of CIA Triad**

* Introduced as a **core security model** to guide policies and controls.
* Ensures that systems and data remain **secure, trustworthy, and usable**.
* Widely applied in **banking, payments, and FinTech ecosystems**.

 *Analogy:*

* **Confidentiality**  keeping your ATM PIN secret 
* **Integrity**  ensuring your account balance is correct after a transaction 
* **Availability**  being able to withdraw cash anytime 

<img src="https://www.asisonline.org/globalassets/security-management/security-technology/2022/october/1022-sectec-peden-availability-achieves-positive-outcome-cia-triad.jpg" width="350" alt="Blockchain Animation">

---

##  **2.2 Confidentiality **

### **Definition and Importance**

* Ensures **only authorized people** can access sensitive data.
* Prevents **unauthorized disclosure** of information.
* In FinTech, confidentiality is **non-negotiable** since **money = trust**.

**Why it matters:**

* Protects customers financial details.
* Safeguards institutions from fraud.
* Builds user confidence in digital banking.

---

### **Methods to Ensure Confidentiality**

1. **Encryption **

   * Converts plain text  unreadable ciphertext.
   * Example: AES encryption in mobile banking apps.
   * Ensures stolen data is useless without the decryption key.

2. **Access Control **

   * Restricting who can access what (role-based access).
   * Example: Only the **finance department** can approve transactions.

3. **Least Privilege Principle **

   * Users should have **only the minimum access needed** to do their job.
   * Example: A call center rep can view account details but **cannot transfer funds**.

---

### **FinTech Examples of Confidentiality**

*  **Mobile Wallet PIN**: Protecting the customers 4-digit PIN from exposure.
*  **Online Banking**: Using TLS (HTTPS) to encrypt communication between browser and bank server.
*  **Credit Card Transactions**: PCI-DSS compliance ensures card numbers are encrypted.

# **2.3 Integrity **

---

### **Definition**

Integrity means ensuring that **data remains accurate, consistent, and trustworthy** throughout its lifecycle.

* Data should not be **altered, deleted, or manipulated** without authorization.
* Any changes must be **traceable and verifiable**.

 In simple terms: *What goes in should be the same as what comes out.*

---

### **Why Integrity Matters in Cybersecurity**

* Incorrect or tampered data can lead to **financial losses, fraud, and wrong decisions**.
* In FinTech, even **1-digit change in a transaction log** can cause huge discrepancies.
* Integrity violations can damage **trust between banks, businesses, and customers**.

<img src="https://i.pinimg.com/originals/89/db/a8/89dba89b2b6a4efdf56e766a6e4f091a.gif" width="350" alt="Blockchain Animation">

---

### **Mechanisms to Ensure Integrity**

1. **Hashing **

* A mathematical function that converts data  fixed-size string (hash).
* If even **1 character changes**, the hash changes completely.
* Example: SHA-256 used in blockchain & banking.

---

2. **Checksums **

* A simple method to verify file or data integrity during transfer.
* Example: When downloading banking software updates, a checksum ensures it hasnt been tampered with.

---

3. **Digital Signatures **

* Combine **encryption + hashing** to verify sender authenticity & data integrity.
* Example: Digital certificates used in online payment systems (SSL/TLS).

---

### **FinTech Example  Preventing Tampering of Transaction Logs **

* Every bank and FinTech system maintains **transaction logs** (record of transfers, deposits, withdrawals).
* If an attacker tampers with logs (e.g., adds fake transactions):

  *  Financial fraud occurs
  *  Customers lose trust
  *  Bank faces legal penalties

**How Integrity Helps:**

* Hashing ensures logs havent been changed.
* Digital signatures verify transaction authenticity.
* Blockchain-based ledgers (immutable) provide **tamper-proof integrity**.

##  **2.4 Availability **

**Definition:**
Availability ensures that **authorized users have continuous access** to data, applications, and systems **whenever they need it**.

 In FinTech, **availability = money flow**  if systems go down, transactions stop, and trust is broken.

---

### **How to Ensure Availability**

1. **Redundancy **

   * Having multiple servers/components ready to take over if one fails.
   * Example: Payment gateways hosting **backup servers in multiple regions**.

2. **Failover Systems **

   * Automatic switch to a standby system during failure.
   * Example: If one ATM network node fails, another takes over seamlessly.

3. **Backups **

   * Regularly saving system/data copies to restore after crashes or ransomware.
   * Example: Banks maintaining **daily transaction backups** in secure storage.

4. **DDoS (Distributed Denial of Service) Protection **

   * DDoS attacks flood systems with fake requests, making them unavailable.
   * Mitigation: Firewalls, load balancers, Cloudflare/Akamai protection.

---

### **FinTech Example  24/7 Availability of Payment Gateways**

* Online payment systems (Visa, Mastercard, PayPal, Easypaisa, JazzCash) must be operational **24/7/365**.
* Even **5 minutes of downtime** can cause:

  *  Loss of millions in failed transactions.
  *  Customer frustration  loss of trust.
  *  Reputation damage for banks & FinTech apps.

##  **2.5 Balancing the Triad **

While the CIA Triad works together, in reality, there are **conflicts** between principles. Organizations must **prioritize** based on business needs.

---

### **Common Conflicts**

1. **Confidentiality  vs Availability **

   * Strong encryption  slows down system performance.
   * Example: A mobile banking app may delay transactions due to heavy encryption.

2. **Integrity  vs Availability **

   * Strict validation checks ensure accurate transactions but may reduce speed.
   * Example: Too many fraud checks may cause payment delays.

3. **Confidentiality  vs Integrity **

   * Frequent system updates (for integrity) might expose data temporarily.

---

### **Real-World Case: Banking Systems Prioritizing Availability**

* Banks **cannot afford downtime**. Customers expect **ATMs, mobile banking, and POS systems** to work 24/7.
* To maintain uptime, banks may:

  * Use **weaker encryption** on some internal systems for faster processing.
  * Delay applying certain security patches to **avoid disruptions**.

 This shows **availability is often prioritized over confidentiality** in real-world financial systems.

# **3. Threats, Vulnerabilities, and Risks**

Understanding these three core terms is crucial because **most cybersecurity failures happen when organizations confuse them**.

---

##  **3.1 Definitions**

---

### ** Threat = Potential Cause of Harm**

* A **threat** is anything (person, event, or action) that has the **potential** to cause damage to systems, networks, or data.
* Think of it as the **danger itself**, whether or not it has been exploited yet.

**Examples in FinTech:**

*  Hackers trying to breach a mobile banking app.
*  Phishing emails targeting customers.
*  Natural disasters destroying data centers.

<img src="https://media1.giphy.com/media/v1.Y2lkPTc5MGI3NjExNjgxaTJ6eTEzYmYxYWsxZHBjYmxqenprcnFodG8zazRweTV4c2t4NiZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9cw/9Xw21tq79NuPeRJ3OE/giphy.gif" width="350" alt="Blockchain Animation">

---

### ** Vulnerability = Weakness Exploited**

* A **vulnerability** is a flaw or weakness in a system that can be **exploited by a threat**.
* Vulnerabilities do not cause harm by themselves, but they **open the door** for attacks.

**Examples in FinTech:**

*  Weak passwords in customer accounts.
*  Unpatched software in online payment systems.
*  Misconfigured firewalls exposing APIs.

<img src="https://i.pinimg.com/originals/aa/02/7f/aa027f0556f46c0cb7a1567964d157dc.gif" width="350" >

---

### ** Risk = Likelihood  Impact**

* **Risk** is the combination of:

  1. The **likelihood** that a threat will exploit a vulnerability.
  2. The **impact** (damage) if it happens.

**Formula:**

$$
\text{Risk} = \text{Likelihood} \times \text{Impact}
$$

**Examples in FinTech:**

* SIM swap attack 

  * Likelihood = Medium (depends on telco security)
  * Impact = High (account takeover, financial loss)
  * Risk =  High

* ATM downtime 

  * Likelihood = Low
  * Impact = Medium
  * Risk =  Moderate



---

 **Quick Analogy for Students:**

* **Threat** = A thief in the street 
* **Vulnerability** = An open window in your house 
* **Risk** = Chance the thief notices & enters, multiplied by what he can steal

# **3.2 Types of Threats**

Cyber threats come from different sources. In FinTech, these threats can target **people , technology , or the environment **.

---

##  **1. Human Threats **

These are threats caused by **malicious or careless people**.

### **Examples:**

* **Hackers **

  * Try to steal money or customer data.
  * Example: Hackers breaching **crypto exchanges** like Mt. Gox.

* **Insiders **

  * Employees or contractors with access to internal systems.
  * Example: Bank staff leaking customer details for fraud.

* **Phishing Attacks **

  * Fake emails/SMS pretending to be from banks to steal login details.
  * Example: SMS scam tricking users to share their Easypaisa PIN.

These are threats caused by **flaws or failures in technology**.

### **Examples:**

* **Malware **

  * Viruses, worms, ransomware infecting systems.
  * Example: Ransomware locking bank servers until ransom is paid in Bitcoin.

* **Software Bugs **

  * Coding errors that attackers exploit.
  * Example: Capital One breach (2019) due to AWS firewall misconfiguration.

* **Misconfigurations **

  * Weak system setup leaving back doors open.
  * Example: API misconfiguration exposing mobile wallet transactions.

  * Example: A hurricane destroying a financial data center in Florida.

* **Power Outages **

  * Loss of electricity shutting down ATMs and payment gateways.

# **3.3 Types of Vulnerabilities**

A **vulnerability** is a weakness in a system that can be exploited by a threat.
 Think of it as the **open window ** in your house that a thief can use.

In cybersecurity, vulnerabilities fall into **three main categories**:

---

##  **1. Software Vulnerabilities **

These occur due to **flaws or bugs** in applications or operating systems.

### **Examples:**

* **Unpatched Systems **

  * Outdated software leaves holes for attackers.
  * Example: Equifax breach (2017) due to an **unpatched Apache Struts vulnerability**.

* **Zero-Day Exploits **

  * Vulnerabilities unknown to the vendor (no patch yet).
  * Hackers exploit them before they are fixed.

* **Insecure Mobile Apps **

  * Poor coding practices in FinTech apps may expose customer data.

 *FinTech Example:*
A digital banking app running an **old version** exposes customer details because no patch was applied.

<img src="https://media2.giphy.com/media/v1.Y2lkPTc5MGI3NjExdjYycmF1eTdjdXJocnE5NXRsb21ib2ZkMmJtNXpwYW13OGZ2N3RvNyZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/UQOwUpOSg2SuykKmOh/giphy.gif" width="350" >

---

##  **2. Network Vulnerabilities **

These weaknesses exist in the **communication infrastructure**.

### **Examples:**

* **Weak Firewall Rules **

  * Misconfigured firewalls allow unnecessary traffic.
  * Example: A bank accidentally exposing its **admin panel** to the internet.

* **Open Ports **

  * Services running on ports left unsecured.
  * Example: Attackers exploiting **port 22 (SSH)** or **port 1433 (SQL server)**.

* **Lack of Encryption **

  * Data sent in plain text (e.g., HTTP instead of HTTPS).
  * Example: Customer login credentials stolen over an open Wi-Fi network.

 *FinTech Example:*
Hackers sniff unencrypted **credit card transactions** from a payment gateway.

<img src="https://media1.giphy.com/media/v1.Y2lkPTc5MGI3NjExdDQ4ZGtncHNzeGY0MTUxMG5kaWdnenF2dzNuZThqOHJ6eXE1aHpkMiZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/rQcEj8UVOC6QvJs36s/giphy.gif" width="350" >

---

##  **3. Human Vulnerabilities **

Humans are often the **weakest link** in cybersecurity.

### **Examples:**

* **Weak Passwords **

  * Users set 123456 or password  easily cracked.
  * Example: Bank employees using weak credentials to access systems.

* **Social Engineering **

  * Tricking people into giving up secrets.
  * Example: A phishing email asking users to verify their bank details.

* **Insider Mistakes **

  * Employees accidentally clicking malicious links or sharing confidential files.

 *FinTech Example:*
An attacker calls a mobile wallet helpline pretending to be a customer  tricks the agent into resetting the PIN (SIM swap fraud).

# **3.4 Risk Analysis **

Risk analysis helps us **measure and prioritize threats** so organizations know **what to fix first**.
 Formula reminder:

$$
\text{Risk} = \text{Likelihood} \times \text{Impact}
$$

---

##  **1. Qualitative Risk Analysis **

* Uses **descriptions** instead of numbers.
* Risks classified as **Low, Medium, or High** based on expert judgment.
* **Fast & simple**, but subjective.

**Example:**

* Phishing email received in a bank.

  * Likelihood = High (employees get them daily)
  * Impact = Medium (may lead to credential theft)
  * Risk = High 

<img src="https://media2.giphy.com/media/v1.Y2lkPTc5MGI3NjExZHllMWQ5MzR2bGg5ZXN0eHZwcm1uMnZ1dTk4Yzh0a3dyb2dvOW1jNyZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/9jhuQcF26JbzJ6I34a/giphy.gif" width="350" >

---

##  **2. Quantitative Risk Analysis **

* Uses **numerical values** (probability & monetary impact).
* Risk expressed in \$\$ or percentages.
* More accurate but **requires data & calculations**.

**Formula Example:**

$$
\text{Annual Risk Exposure} = \text{Probability of Attack} \times \text{Estimated Financial Loss}
$$

* If SIM swap fraud costs \~\$100,000 per incident
* Probability = 0.2 (20% chance per year)
* Risk Exposure = \$20,000

<img src="https://media2.giphy.com/media/v1.Y2lkPTc5MGI3NjExeDVuYmZ2d3JwMnFwcGN4ano3dHZuOTFwNjZ1N3BnYm0ybTlvbGE1YSZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/tWhSt6azAiDYhW9VhG/giphy.gif" width="350" >

| ------------------------- | ---------- | ------------- | ----------- |
| **Low**                 | Low        | Low           | Medium      |
| **Medium**              | Low        | Medium        | High        |

# **3.5 Security Controls **

Security controls are the **measures we put in place** to protect systems from threats and reduce risks.
They are grouped into **three main types**: **Preventive, Detective, and Corrective**.

---

##  **1. Preventive Controls  (Stop attacks before they happen)**

**Definition:**
Controls that aim to **block threats** and reduce vulnerabilities **before** damage occurs.

**Examples:**

* **Firewalls **  Block unauthorized traffic.
* **Encryption **  Protects sensitive data in transit & storage.
* **Multi-Factor Authentication (MFA) **  Stops account takeovers.
* **Access Controls **  Role-based restrictions (RBAC).

**FinTech Examples:**

* Mobile banking apps using **TLS/SSL** encryption for secure logins.
* Payment gateways requiring **OTP/2FA** before approving transactions.
* Banks limiting employees access to only what they need (**least privilege**).

<img src="https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExem1nNGsxaXEwem1xbHNyMmJ2MHRsazc5MnExaTlkdjA2bm1idDFvdCZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/bKj0qEKTVBdF2o5Dgn/giphy.gif" width="350" >

---

##  **2. Detective Controls  (Identify attacks in progress)**

**Definition:**
Controls that **monitor systems** and help detect suspicious activity.

**Examples:**

* **IDS (Intrusion Detection System) **  Monitors network traffic for attacks.
* **IPS (Intrusion Prevention System) **  Detects & blocks suspicious activity.
* **Logging & Monitoring **  Recording all system events for analysis.
* **SIEM (Security Information and Event Management) **  Combines logs from multiple sources for real-time alerts.

**FinTech Examples:**

* Detecting unusual login attempts on mobile wallets (e.g., logging in from 2 countries within minutes ).
* Monitoring **large, unusual transactions**  potential fraud detection.
* Real-time alerts when a banking API receives **suspicious requests**.

<img src="https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExNHV1YTNwNTh3eHhpam1pNnc1YzFsOWZhb2F6anU2ZDF4aTl0bW1qMyZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/3o6ZtpyL1AwhIrTPiw/giphy.gif" width="350" >

---

##  **3. Corrective Controls  (Fix after an attack happens)**

**Definition:**
Controls that **restore systems** and fix vulnerabilities **after an incident**.

**Examples:**

* **Patching **  Fixing vulnerabilities with software updates.
* **Backup & Recovery **  Restoring data after ransomware or outages.
* **Incident Response Plans **  Steps to recover after breaches.

**FinTech Examples:**

* Bank applies emergency **patches** after a vulnerability is exploited.
* Payment provider restores customer data from **daily backups** after a server crash.
* Digital wallet service activating **disaster recovery site** after data center failure.

<img src="https://media1.giphy.com/media/v1.Y2lkPTc5MGI3NjExcHVuYWhwMmo0N2VjazU1YzBqaTB2bzR5ZGdnOGVmcHhxbzhzMXQwZiZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/2xjIZ4DPAhDLAfGVER/giphy.gif" width="350" >

---

##  **Summary Table for Students**

| **Type**          | **Purpose**                   | **Examples**                      | **FinTech Context**                   |
| ----------------- | ----------------------------- | --------------------------------- | ------------------------------------- |
| **Preventive ** | Stop attack before it happens | Firewalls, encryption, MFA        | Secure logins, OTPs for wallets       |
| **Detective **  | Detect attacks in progress    | IDS, IPS, SIEM, logs              | Detecting fraud & unusual logins      |
| **Corrective ** | Fix after attack              | Patching, backups, recovery plans | Restoring payment system after outage |

# **4. Real-World Cyber Attack Examples (Case Studies)**

Studying past cyber-attacks helps us **learn from mistakes** and understand how threats exploit vulnerabilities.

---

##  **4.1 Major Global Incidents**

---

### **1. JP Morgan Chase Breach (2014) **

* **What Happened?**

  * One of the biggest banks in the US suffered a **data breach**.
  * Attackers stole personal info from **76 million households** and **7 million small businesses**.

* **Cause:**

  * Hackers exploited a **server misconfiguration**  missing two-factor authentication (2FA).

* **Impact:**

  * Customer names, phone numbers, and emails leaked.
  * Shaken trust in banking security.
  * JP Morgan had to **spend billions on upgrading security**.

* **CIA Triad Mapping:**

  * **Confidentiality **  Broken (customer data exposed).
  * **Integrity **  Not significantly impacted.
  * **Availability **  Services remained online.

<img src="https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExMHdqc241eGY2NWJ3a3BlanppdmVpd3p3NndiMXc4b2xtcjlraThnbCZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/xC4MeWCMCezbJFMpID/giphy.gif" width="350" >

---

### **2. Bangladesh Bank SWIFT Attack (2016) **

* **What Happened?**

  * Hackers infiltrated the **SWIFT (Society for Worldwide Interbank Financial Telecommunication) system**.
  * They tried to steal **\$951 million** from Bangladesh Banks account at the US Federal Reserve.
  * Luckily, only **\$81 million was successfully transferred** before being stopped.

* **Cause:**

  * **Weak network security** (cheap routers, no firewalls).
  * Lack of proper monitoring/logging of SWIFT transactions.

* **Impact:**

  * Huge financial loss (\$81M).
  * Damaged reputation of SWIFT system worldwide.
  * Highlighted the need for **stronger financial cybersecurity** in developing countries.

* **CIA Triad Mapping:**

  * **Confidentiality **  Partially affected (transaction details exposed).
  * **Integrity **  Broken (fraudulent transactions injected).
  * **Availability **  Services continued, but trust shaken.

<img src="https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExczllenZ2NmdpdTYybW90OThwdmxhbzhjcXV1bzJuOHU0OW56ZmI4aiZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/xTiTnqUxyWbsAXq7Ju/giphy.gif" width="350" >

---

### **3. Capital One Breach (2019) **

* **What Happened?**

  * A hacker gained access to **AWS cloud servers** used by Capital One.
  * Exposed data of **106 million customers** in the US & Canada.

* **Cause:**

  * **Misconfigured AWS firewall (WAF)**  allowed unauthorized access.

* **Impact:**

  * Exposed personal data: names, addresses, SSNs, credit scores.
  * Bank faced lawsuits and a **\$80 million fine** from regulators.
  * Example of how **cloud misconfigurations** can be catastrophic.

* **CIA Triad Mapping:**

  * **Confidentiality **  Broken (sensitive customer data stolen).
  * **Integrity **  Data not altered.
  * **Availability **  Services remained active.

# **4.2 FinTech-Specific Incidents **

FinTech platforms (mobile wallets, payment gateways, crypto exchanges) are highly targeted because they deal with **money flow + personal data**.

---

##  **1. Mobile Wallet Frauds (SIM Swap Cases) **

* **What Happens?**

  * An attacker tricks the telecom provider into issuing a **duplicate SIM card**.
  * The attacker now receives all **OTP codes & SMS notifications**.
  * Victims **mobile wallet (Easypaisa, JazzCash, PayPal, etc.) is hijacked**.

* **Cause:**

  * Weak identity verification at telecom customer service.
  * Reliance on SMS OTP only (no multi-factor authentication).

* **Impact:**

  * Victim loses funds from wallet .
  * Fraudsters may also **reset bank credentials** linked to mobile number.
  * Telecom and wallet provider face reputation damage.

* **CIA Triad Mapping:**

  * **Confidentiality **  Broken (attacker accesses private OTPs & accounts).
  * **Integrity **  Transactions manipulated by attacker.
  * **Availability **  User loses access to their wallet.

<img src="https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExM2Zkd2JobmllbWR2d3ZzZzY1ZGYxMG1semV6aDc1Y2dnbWg5ZnVibyZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/fKqBFh5pebTrycP2wo/giphy.gif" width="350" >

---

##  **2. Payment Gateway Hacks (API Vulnerabilities) **

* **What Happens?**

  * Attackers exploit insecure **APIs (Application Programming Interfaces)** in payment gateways.
  * Example: Poor authentication  attackers send **fraudulent transactions**.

* **Cause:**

  * Weak API authentication & authorization.
  * Poor input validation (SQL injection, code injection).
  * Misconfigured firewalls exposing admin endpoints.

* **Impact:**

  * Fraudulent transactions & stolen customer card details.
  * Business partners lose trust in gateway security.
  * PCI-DSS compliance violations  heavy fines.

* **CIA Triad Mapping:**

  * **Confidentiality **  Exposed card numbers.
  * **Integrity **  Fake or altered transactions.
  * **Availability **  Gateway may crash due to exploitation.

##  **3. Cryptocurrency Exchange Hacks (Mt. Gox, Binance) **

* **Mt. Gox (2014)**

  * Once handled **70% of global Bitcoin transactions**.
  * Hackers stole **850,000 BTC** (worth \$450M at the time; over \$30B today ).
  * Cause: Poor internal security, no proper auditing.

* **Binance (2019)**

  * Hackers stole **7,000 BTC (\~\$40M)**.
  * Attackers used **phishing + API key exploits**.
  * Binance reimbursed customers from a **Secure Asset Fund**.

* **Impact:**

  * Massive financial loss.
  * Drop in customer trust.
  * Governments increased regulation on crypto exchanges.

* **CIA Triad Mapping:**

  * **Confidentiality **  Wallet keys & data stolen.
  * **Integrity **  Exchange transaction records altered.
  * **Availability **  Platforms temporarily shut down.

# **4.3 Lessons Learned **

From studying **global + FinTech-specific attacks**, we can identify **patterns** in why breaches happen and how to prevent them.

---

##  **1. Common Causes of Cyber Attacks **

1. **Weak Authentication **

   * Single passwords or SMS OTPs are easily stolen.
   * Example: SIM swap attacks on mobile wallets.

2. **Misconfigured Servers **

   * Poor firewall rules, open ports, weak cloud setups.
   * Example: Capital One breach (2019) due to AWS WAF misconfiguration.

3. **Insider Threats **

   * Employees abusing access or making mistakes.
   * Example: Bank staff leaking transaction data for fraud.

4. **Unpatched Systems **

   * Old software not updated  attackers exploit known bugs.
   * Example: Equifax breach (2017) from an unpatched Apache vulnerability.

<img src="https://media4.giphy.com/media/v1.Y2lkPTc5MGI3NjExNXk4eWN4NGRkbGJ1dWFpZXRpbzh4NjF1ZngwYXN0bXY0ZmpuZnZwayZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/vtm4qejJIl1ERPIrbA/giphy.gif" width="350" >

---

##  **2. Impact on the CIA Triad **

* **Confidentiality **  Customer data exposed (emails, card numbers, SSNs).
* **Integrity **  Fraudulent transactions injected into financial systems.
* **Availability **  DDoS attacks & outages making services unusable.

 Example Mapping:

* JP Morgan (2014)  **Confidentiality breach**.
* Bangladesh Bank SWIFT attack (2016)  **Integrity violation**.
* DDoS on payment gateways  **Availability failure**.

<img src="https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExZHBrN3dwdnBscWpsZ25jZjlreXZwOHRyeW1peWg1eWs4dmp2bWJ1cSZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/PwpK5oV3yozgedXSqU/giphy.gif" width="350" >

---

##  **3. Mitigation Measures **

1. **Two-Factor Authentication (2FA) / Multi-Factor Authentication (MFA) **

   * Protects against SIM swaps & phishing.
   * Example: Using biometrics (fingerprint, face ID) + OTP for mobile wallets.

2. **Regular Auditing & Monitoring **

   * Detects misconfigurations & suspicious activity early.
   * Example: Continuous cloud security checks for AWS/Azure systems.

3. **Network Segmentation **

   * Divides systems into smaller secure zones.
   * Example: Separating customer-facing APIs from internal payment processing systems.

4. **Patch Management **

   * Regular updates & fixes for vulnerabilities.
   * Example: FinTechs applying latest TLS versions & fixing API bugs quickly.

5. **Employee Training **

   * Reduce human vulnerabilities (phishing, social engineering).
   * Example: Staff recognizing fake bank admin emails.

# **5. Summary and Reinforcement **

---

##  **5.1 Recap of CIA Triad and Core Concepts**

 **CIA Triad (Pillars of Security):**

* **Confidentiality **  Keeping data secret (encryption, access control).
* **Integrity **  Keeping data accurate & untampered (hashing, digital signatures).
* **Availability **  Keeping systems accessible (redundancy, failover, DDoS protection).

 **Core Cybersecurity Concepts:**

* **Threat ** = Potential danger.
* **Vulnerability ** = Weakness that can be exploited.
* **Risk ** = Likelihood  Impact.
* **Controls ** = Preventive, Detective, Corrective.



---

##  **5.2 Mapping Threats and Attacks to CIA Triad**

| **Attack**                          | **Threat Type**            | **CIA Pillar Violated**             | **FinTech Example**               |
| ----------------------------------- | -------------------------- | ----------------------------------- | --------------------------------- |
| JP Morgan Breach (2014)             | Human + Technical          |  Confidentiality                  | Customer data stolen              |
| Bangladesh Bank SWIFT Attack (2016) | Insider + Technical        |  Integrity                         | Fake transactions injected        |
| DDoS on Payment Gateway             | Technical                  |  Availability                      | Customers unable to make payments |
| SIM Swap Fraud                      | Human (social engineering) |  Confidentiality +  Integrity    | Wallet takeover, false transfers  |
| Crypto Exchange Hack                | Technical                  |  Confidentiality +  Availability | Funds stolen, service downtime    |



---

##  **5.3 Importance of Cybersecurity Mindset in FinTech **

* FinTech = **trust-based industry**  one breach can destroy customer confidence.
* Cybersecurity is **not just ITs job**  everyone (developers, employees, customers) plays a role.
* Proactive security mindset saves:

  *  Money (reduced fraud & fines).
  *  Reputation (customers keep trusting services).
  *  Compliance (GDPR, PCI-DSS, ISO27001).

# **Cybersecurity Careers in FinTech **

---

##  **1. Security Analyst **

**Role:**

* First line of defense  monitor systems, analyze alerts, respond to incidents.
* Identify suspicious activity in banking apps, mobile wallets, and payment gateways.

**Tools:**

* SIEM (Splunk, IBM QRadar, ELK Stack)
* IDS/IPS (Snort, Suricata)
* Wireshark (packet analysis)
* Antivirus / EDR solutions

**FinTech Relevance:**

* Monitoring **fraudulent transactions** or **unauthorized access** in real-time.

---

##  **2. Penetration Tester (Ethical Hacker) **

**Role:**

* Simulates attacks to find weaknesses before criminals do.
* Test mobile wallets, APIs, and smart contracts.

**Tools:**

* Kali Linux  (Metasploit, Burp Suite, Nmap)
* OWASP ZAP (web app testing)
* Hydra (password cracking)
* Wireshark (traffic sniffing)

**FinTech Relevance:**

* Test if **banking APIs** can be hacked, or if mobile wallets resist **SIM swap fraud**.

---

##  **3. Security Engineer **

**Role:**

* Builds and maintains secure systems.
* Implements firewalls, encryption, and authentication systems.

**Tools:**

* Firewalls (Palo Alto, Cisco ASA)
* Cloud Security (AWS Security Hub, Azure Sentinel)
* VPNs, TLS/SSL, PKI
* SAST/DAST tools for secure coding

**FinTech Relevance:**

* Designing **secure payment gateways** and **encrypted data storage**.

---

##  **4. Incident Responder **

**Role:**

* Acts when a breach occurs  contain, analyze, and recover.
* Digital forensics and root-cause analysis.

**Tools:**

* SIEM + SOAR platforms
* Volatility (memory forensics)
* Autopsy (disk forensics)
* EnCase (law enforcement-grade forensics tool)

**FinTech Relevance:**

* Responding to **fraud incidents**, ransomware, or API exploitation in financial systems.

---

##  **5. Compliance & Risk Specialist **

**Role:**

* Ensures FinTech follows regulations: PCI-DSS, ISO 27001, GDPR, PSD2.
* Performs audits & risk assessments.

**Tools:**

* GRC tools (RSA Archer, ServiceNow GRC)
* Risk assessment frameworks (NIST RMF, ISO 27005)
* Compliance scanners (Qualys, Nessus for vulnerabilities)

**FinTech Relevance:**

* Keeping digital wallets & payment systems **regulatory compliant**.

---

##  **6. Cryptographer **

**Role:**

* Designs algorithms for encryption & secure communication.
* Works in blockchain-based FinTech security.

**Tools:**

* OpenSSL, Hashcat
* Python (PyCrypto, Cryptography library)
* Blockchain frameworks (Ethereum, Hyperledger)

**FinTech Relevance:**

* Protecting **transactions, tokens, and customer credentials**.

---

##  **7. Cybersecurity Consultant **

**Role:**

* Advises banks & FinTechs on best security practices.
* Works across multiple domains: risk, compliance, cloud, payments.

**Tools:**

* Mix of all above + frameworks (OWASP, MITRE ATT\&CK)
* Cloud services (AWS, GCP, Azure security tools)

**FinTech Relevance:**

* Guide companies in building **secure FinTech platforms** from scratch.

---

##  **8. SOC Analyst (Security Operations Center) **

**Role:**

* Works in 24/7 monitoring centers.
* Tracks **alerts, logs, and anomalies** continuously.

**Tools:**

* Splunk, QRadar, ELK
* Threat Intel feeds (AlienVault OTX, MISP)
* SIEM dashboards

**FinTech Relevance:**

* Detecting **suspicious login attempts** or **DDoS on payment systems** in real-time.

---

# **Core Skills Needed Across Roles **

*  Knowledge of **CIA Triad & Security Controls**
*  Networking fundamentals (OSI, TCP/IP)
*  Linux basics (command line, scripts)
*  Python for automation & scripting
*  Problem-solving & critical thinking
*  Knowledge of regulations (PCI-DSS, GDPR, ISO 27001, PSD2)

## **3 Search Skills**  *(Free)*

**Room Link:** [https://tryhackme.com/room/searchskills](https://tryhackme.com/room/searchskills)

### ** Tasks**

* **Task1:**  *Introduction*
  Basics of information gathering and effective searching.
* **Task2:**  *Evaluation of Search Results*
  Learn how to verify credibility and cross-check sources.
* **Task3:**  *Search Engines*
  Master advanced Google operators (`" "`, `site:`, `filetype:`).
* **Task4:**  *Specialized Search Engines*
  Explore **Shodan**, **Censys**, and **VirusTotal** for OSINT.
* **Task5:**  *Vulnerabilities & Exploits*
  Learn to locate CVEs and use Exploit DB effectively.
* **Task6:**  *Technical Documentation*
  Use Linux `man` pages and official docs for troubleshooting.
* **Task7:**  *Social Media OSINT*
  Use LinkedIn, Facebook, and GitHub for open-source intelligence.
* **Task8:**  *Conclusion*
  Wrap up with a recap of all OSINT and search techniques.
