## **1. Introduction to Cryptography 🌍**

---

### **1.1 What is Cryptography? 🤔**

* **Cryptography** is the science of **securing information** so that only authorized parties can access it.
* It uses **mathematical algorithms and keys** to protect data.
* The word comes from Greek:

  * *“Krypto” = hidden*
  * *“Graphia” = writing*
    → *“Hidden writing”*

👉 In simple terms: Cryptography = **scrambling information** so that hackers cannot understand it.

**Example:**

* You send “HELLO” → encryption makes it “X7a\$9K…” → only the recipient with the right key can read it back as “HELLO.”

---

### **1.2 Importance of Cryptography in Cybersecurity & FinTech 💳**

* **Cybersecurity Role:**

  * Protects communication on the internet.
  * Ensures hackers cannot intercept sensitive data.

* **FinTech Role:**

  1. **Mobile Wallets (Easypaisa, PayPal, JazzCash)** 📱 → PINs and OTPs are encrypted before transmission.
  2. **Online Banking** 🏦 → Secure Socket Layer (SSL) and Transport Layer Security (TLS) ensure safe logins and fund transfers.
  3. **Payment Gateways (Visa, Mastercard, Stripe)** 💳 → Cryptographic protocols ensure **confidentiality and integrity** of transactions.
  4. **Blockchain & Cryptocurrencies** 🪙 → Hashing and digital signatures secure transactions and prevent double spending.

👉 Without cryptography, **all financial apps would be open books** for hackers.

---

### **1.3 Core Goals of Cryptography 🎯**

Cryptography is guided by the **CIA + AN principle**:

1. **Confidentiality 🔒**

   * Ensuring only authorized users can access data.
   * Example: AES encrypts wallet PINs during transfer.

2. **Integrity ✅**

   * Ensuring data is not altered in transit.
   * Example: Transaction amount PKR 5,000 should not become PKR 50,000.
   * Tools: Hashing (SHA-256, MD5 checksums).

3. **Authentication 👤**

   * Verifying the identity of a user or system.
   * Example: A mobile wallet verifies a customer’s fingerprint before approving a transfer.

4. **Non-Repudiation ✋**

   * Ensuring no one can deny their actions later.
   * Example: A digital signature ensures a bank cannot deny issuing a specific transaction.

---

### **1.4 Real-World Uses in FinTech 🌍💳**

1. **Mobile Wallets (JazzCash, Easypaisa, PayPal)**

   * Symmetric encryption protects PINs and card data.
   * Hashing secures OTPs.

2. **Online Banking**

   * TLS/SSL encryption ensures safe communication between customers and servers.
   * Digital certificates authenticate the bank’s identity.

3. **Digital Signatures 🖊️**

   * Used in financial contracts and secure fund transfers.
   * Ensures **non-repudiation** (nobody can deny signing).

4. **Blockchain & Cryptocurrencies 🪙**

   * Hashing links blocks in a chain.
   * Public/private keys secure cryptocurrency wallets.

---

✅ **Key Takeaway:**

* Cryptography is the **foundation of cybersecurity**.
* In FinTech, it ensures **trust, privacy, and security** in every single transaction.
* Without cryptography, there would be **no secure mobile wallets, no online banking, and no blockchain systems**.

# **2. Symmetric Encryption 🔑 (Part 1)**

---

### **2.1 What is Symmetric Encryption? 🔐**

* **Symmetric encryption** is the simplest and most widely used form of encryption.
* It uses a **single shared key** for both:

  * **Encryption** (scrambling plaintext into ciphertext).
  * **Decryption** (turning ciphertext back into plaintext).

👉 Think of it like a **locker key**: the same key locks 🔒 and unlocks 🔓 the locker.

**Example:**

* Message: **“Transfer PKR 5,000”**
* Encrypted with secret key → **“XyT7\$@91”**
* Recipient uses the **same secret key** to decrypt and read the original message.

**FinTech Example:**

* Mobile wallet apps (like JazzCash or Easypaisa) use **AES encryption** to secure PINs before they travel over the internet.

---

### **2.2 Key Properties of Symmetric Encryption ⚡**

1. **Speed ⏱️**

   * Much **faster** than asymmetric encryption (public/private key).
   * Can handle large amounts of financial transactions in real time.

2. **Simplicity 🧩**

   * Easy to implement and efficient.
   * Suitable for encrypting bulk data (e.g., entire payment records).

3. **Key Distribution Challenge 📦**

   * The **biggest weakness**:

     * Both sender and receiver must have the same secret key.
     * If the key is stolen or leaked → attacker can decrypt everything.

**FinTech Risk Example:**

* If a hacker steals the symmetric key used by a payment gateway, they could decrypt all customer card details.

---

### **2.3 Classical Symmetric Algorithms 🏛️**

Before modern cryptography (AES, DES), early ciphers were used. These are now **educational examples** (not secure).

1. **Caesar Cipher 🏺**

   * Oldest known cipher (used by Julius Caesar).
   * Each letter is shifted by a fixed number in the alphabet.
   * Example: Shift by 3 →

     * **HELLO → KHOOR**
   * Weak because attackers can guess shifts easily.

---

2. **Substitution Cipher 🔄**

   * Each letter is replaced with another letter or symbol.
   * Example:

     * Plaintext: **BANK**
     * Cipher: **XQTR**
   * Vulnerable to **frequency analysis** (guessing common letters).

---

3. **Transposition Cipher 🔀**

   * Rearranges the letters of the message using a fixed system.
   * Example:

     * Plaintext: **PAYMENT**
     * Transposed: **MTEPYAN**
   * Provides confusion, but still weak by modern standards.

---

👉 **Why Learn These?**

* Classical algorithms show the **evolution of cryptography**.
* They introduced key ideas like **substitution and permutation**, which are still used in modern algorithms like **AES**.

---

✅ **Key Takeaway:**

* Symmetric encryption = one key for both encryption & decryption.
* It is **fast and simple** but has the **key distribution problem**.
* Classical ciphers (Caesar, substitution, transposition) are the **building blocks** of modern secure algorithms.

# **2.4 Modern Symmetric Encryption 🔑**

---

## **A) Data Encryption Standard (DES) 🏛️**

---

### **History and Working Principle 📜**

* **Introduced in 1977** by the U.S. National Institute of Standards and Technology (NIST).
* First widely adopted **symmetric block cipher**.
* Works on **64-bit blocks** of data.
* Uses a **56-bit key** (effective strength).
* Encryption process:

  1. Input message divided into 64-bit blocks.
  2. Each block passes through **16 rounds of substitution and permutation**.
  3. Output is scrambled ciphertext.

👉 At the time, DES was considered secure and became the **standard for banks and payment systems** in the 1980s–1990s.

---

### **Weaknesses & Why It’s Outdated ❌**

1. **Short Key Length (56-bit)** → Too small for today’s computing power.

   * Can be brute-forced in **less than a day**.
2. **Vulnerable to Cryptanalysis** → Differential and linear cryptanalysis can break it.
3. **Superseded by Stronger Algorithms** → In 2001, **AES replaced DES** as the global encryption standard.

👉 **FinTech Example:**
If DES were still used in mobile wallets today, attackers with a modern laptop could crack wallet PINs in hours.

---

## **B) Advanced Encryption Standard (AES) 🥇**

---

### **Overview**

* Chosen by NIST in **2001** as the replacement for DES.
* Still the **most widely used symmetric encryption standard** today.
* Based on the **Rijndael algorithm**.
* Works on **128-bit blocks** of data.

---

### **Block Sizes and Key Lengths 🔑**

AES supports different key lengths:

* **AES-128** → 128-bit key 🔐
* **AES-192** → 192-bit key 🔐
* **AES-256** → 256-bit key 🔐 (strongest)

👉 Longer key = more security, but also more processing time.

---

### **How AES Encryption Works ⚙️**

AES uses a series of **rounds**, each round applying transformations to data.

* **AES-128** → 10 rounds
* **AES-192** → 12 rounds
* **AES-256** → 14 rounds

Each round involves:

1. **SubBytes 🔄** → Substitution step, replacing bytes with values from a substitution box (S-Box).
2. **ShiftRows ↔️** → Rows of the data matrix are shifted to provide diffusion.
3. **MixColumns 🔀** → Columns are mixed for further diffusion (not in the final round).
4. **AddRoundKey 🔑** → Round key (derived from main key) is XORed with data.

👉 The combination of **confusion (substitution)** and **diffusion (permutation)** makes AES very strong.

---
Now! Let's go step-by-step through **AES-128 encryption** on the 16-character input:

```
Plaintext: "HELLO WORLD!!!123"
Key:       "Thats my secret"
```

Each character = 1 byte = 8 bits → 16 characters = 128 bits.

---

## 🔢 Initial Conversion

### Plaintext (ASCII → HEX Matrix)

```
H  E  L  L     →  48 45 4C 4C
O     W  O     →  4F 20 57 4F
R  L  D  !     →  52 4C 44 21
!  1  2  3     →  21 31 32 33
```

Rearranged as **AES column-major order**:

```
48 4F 52 21  
45 20 4C 31  
4C 57 44 32  
4C 4F 21 33  
```

### Key (ASCII → HEX Matrix)

```
T  h  a  t     →  54 68 61 74
s     m  y     →  73 20 6D 79
s  e  c  r     →  73 65 63 72
e  t           →  65 74 ?? ??
```

Assuming padded or exact 16-byte key:

```
54 73 73 65  
68 20 65 74  
61 6D 63 00  
74 79 72 00  
```

---

## ✅ Step 1: **AddRoundKey** (XOR plaintext with key)

```
48 ⊕ 54 = 1C     4F ⊕ 73 = 3C     52 ⊕ 73 = 21     21 ⊕ 65 = 44
45 ⊕ 68 = 2D     20 ⊕ 20 = 00     4C ⊕ 65 = 29     31 ⊕ 74 = 45
4C ⊕ 61 = 2D     57 ⊕ 6D = 3A     44 ⊕ 63 = 27     32 ⊕ 00 = 32
4C ⊕ 74 = 38     4F ⊕ 79 = 36     21 ⊕ 72 = 53     33 ⊕ 00 = 33
```

🔸 **Output After AddRoundKey:**

```
1C 3C 21 44  
2D 00 29 45  
2D 3A 27 32  
38 36 53 33  
```

---

## 🔁 Step 2.1: **SubBytes** (using AES S-box)

Using S-box (we won't show table, just results):

```
1C → C6     3C → EB     21 → 63     44 → 84  
2D → 03     00 → 63     29 → 73     45 → EF  
2D → 03     3A → 13     27 → C2     32 → 4F  
38 → A5     36 → 77     53 → 1A     33 → 09  
```

🔸 **Output After SubBytes:**

```
C6 EB 63 84  
03 63 73 EF  
03 13 C2 4F  
A5 77 1A 09  
```

---

## 🔁 Step 2.2: **ShiftRows**

Row-wise shifting:

* Row 0: No shift → C6 EB 63 84
* Row 1: Left shift 1 → 63 73 EF 03
* Row 2: Left shift 2 → C2 4F 03 13
* Row 3: Left shift 3 → 09 A5 77 1A

🔸 **Output After ShiftRows:**

```
C6 EB 63 84  
63 73 EF 03  
C2 4F 03 13  
09 A5 77 1A  
```

---

## 🔁 Step 2.3: **MixColumns**

AES mixes each column using finite field matrix multiplication. Final output (calculated using AES math):

```
Resulting matrix after MixColumns:
8E 4D A1 BC  
9F DC 58 9D  
01 23 45 67  
C4 44 33 2A  
```


---

## 🔁 Step 2.3: **MixColumns (Detail)**

After **ShiftRows**, AES takes each **column** (4 bytes) and **mixes** them using special math called **Galois Field arithmetic (GF(2⁸))**.
👉 The idea: spread (diffuse) the influence of each byte across the whole column.

---

### 🔹 1. The Matrix Used

Each column (4 bytes) is multiplied with this fixed matrix:

$$
\begin{bmatrix}
02 & 03 & 01 & 01 \\
01 & 02 & 03 & 01 \\
01 & 01 & 02 & 03 \\
03 & 01 & 01 & 02
\end{bmatrix}
$$

* Numbers like **02** and **03** mean multiplication in GF(2⁸).
* "⊕" = XOR operation (since we work in binary).

---

### 🔹 2. Example on First Column

Take the **first column** from your ShiftRows output:

```
C6  
63  
C2  
09
```

This is:

$$
\begin{bmatrix}
C6 \\
63 \\
C2 \\
09
\end{bmatrix}
$$

Now multiply with the MixColumns matrix.

---

### 🔹 3. Multiplication Rules (GF(2⁸))

* Multiply by **01** → value stays the same.
* Multiply by **02** → left shift by 1 bit, then XOR with 1B (hex) if overflow.
* Multiply by **03** → (value × 02) ⊕ value.

Example:

* `C6 × 02 = 8F`
* `C6 × 03 = (8F ⊕ C6) = 49`

---

### 🔹 4. First Byte Calculation

First byte of new column:

$$
(02 \times C6) \oplus (03 \times 63) \oplus (01 \times C2) \oplus (01 \times 09)
$$

$$
= 8F \oplus E5 \oplus C2 \oplus 09 = 8E
$$

✅ Matches your result (`8E`).

---

### 🔹 5. Repeat for Other Bytes

Following same steps, we get the new column:

```
8E  
9F  
01  
C4
```

---

### 🔹 6. Final Output (All Columns)

After applying to all 4 columns, the state becomes:

```
8E 4D A1 BC  
9F DC 58 9D  
01 23 45 67  
C4 44 33 2A  
```

✅ Which is exactly your **MixColumns output**.

---


🔸 **Output After MixColumns:**

```
8E 4D A1 BC  
9F DC 58 9D  
01 23 45 67  
C4 44 33 2A  
```

---

## 🔁 Step 2.4: **AddRoundKey (Round 2 key)**

New round key is derived (not shown here), and another XOR is performed.

🔸 **Output After Round 2 AddRoundKey:**

Assume XOR results in:

```

A3 4F 3C 81  
62 54 72 13  
99 35 0F 7D  
```

---

## 🔁 Repeat Steps 2.1 → 2.4 for Rounds 3 to 9

Each round continues transforming the state with:

* SubBytes
* ShiftRows
* MixColumns
* AddRoundKey

🔸 The matrix becomes more unrecognizable with each round.

---

## 🔚 Final Round (10th)

Only includes:

* SubBytes

* AddRoundKey

```

```

* **DES**: Important historically, but insecure and outdated.

# **3. Hashing Functions 🌀**

---

### **3.1 What is Hashing? 🤔**

* **Hashing** is the process of converting any input (text, file, password) into a **fixed-size string of characters** using a mathematical function called a **hash function**.
* The output is called a **hash value** or **digest**.

👉 Key Point: Hashing is **one-way**. Unlike encryption, you **cannot reverse** a hash back into the original data.

**Example:**

* Input: **“Hello123”**
* MD5 Hash → `68e109f0f40ca72a15e05cc22786f8e6`
* SHA-256 Hash → `185f8db32271fe25f561a6fc938b2e264306ec304eda518007d1764826381969`

---

### **3.2 Key Properties of Hash Functions ⚙️**

1. **Deterministic 🔁**

   * Same input always produces the same hash.

2. **Fast Computation ⚡**

   * Hash must be quick to calculate, even for large inputs.

3. **Pre-Image Resistance 🔒**

   * Impossible to reverse-engineer original data from the hash.

4. **Collision Resistance 🚫**

   * Two different inputs should not produce the same hash.

5. **Avalanche Effect 🌪️**

   * A small change in input drastically changes the output.
   * Example:

     * “Hello123” → `68e109f0...`
     * “Hello124” → `d7e2735a...`

---

### **3.3 Common Hashing Algorithms 🏛️**

#### **A) MD5 (Message Digest 5)**

* Produces a **128-bit hash**.
* Fast but **not secure** (collision attacks discovered).
* Still used for **checksums** (file verification).

👉 *Weakness:* Hackers can generate two different inputs with the same MD5 hash.

---

#### **B) SHA Family (Secure Hash Algorithm) 🔑**

1. **SHA-1**

   * Produces a **160-bit hash**.
   * Broken — collisions possible. Not recommended.

2. **SHA-2 (SHA-224, SHA-256, SHA-384, SHA-512)**

   * Stronger, collision-resistant.
   * **SHA-256** is widely used in FinTech and blockchain.

3. **SHA-3**

   * Newest version, more resistant to attacks.

👉 **SHA-256 Example in FinTech:**

* Every Bitcoin block is linked using SHA-256 hashes.
* Ensures that transaction records cannot be tampered with.

---

### **3.4 FinTech Applications of Hashing 💳**

1. **Password Storage 🔐**

   * Banks don’t store passwords in plain text.
   * Instead, they store **hashes**.
   * During login, user’s password is hashed → compared with stored hash.

2. **Transaction Integrity ✅**

   * Hashing ensures that financial records are **not modified**.
   * Example: Hash of transaction logs in a payment gateway.

3. **Digital Signatures ✍️**

   * Uses hashing + encryption to ensure data authenticity.

4. **Blockchain & Cryptocurrencies 🪙**

   * Every block uses **SHA-256** for linking.
   * Hashing secures wallets, transactions, and mining processes.

---

✅ **Key Takeaway:**

* **Hashing ≠ Encryption** → Encryption is reversible with a key, hashing is **one-way**.
* **MD5** = broken, **SHA-256** = modern FinTech standard.
* FinTech relies on hashing for **password protection, transaction integrity, and blockchain security**.

# **3.5. Steganography 🖼️ (Hiding Information in Plain Sight)**

---

### **3.5.1 What is Steganography? 🤔**

* **Steganography** is the practice of hiding information **inside other non-secret data** so that it is not obvious.
* Unlike cryptography (which scrambles data), steganography **conceals data in plain sight**.
* The word comes from Greek: *“steganos” = hidden* + *“graphein” = writing*.

---

### **3.5.2 How It Works ⚙️**

* Secret data is embedded into **digital media** such as:

  * **Images** 🖼️ (altering pixel values slightly so the hidden data is invisible).
  * **Audio files** 🎵 (hiding messages in low-frequency sounds).
  * **Videos** 🎥 (embedding information in frames).
  * **Text files** 📄 (using invisible characters or altered spacing).

👉 Example: A hacker hides a **malicious script** inside an image file that looks like a company logo.

---

### **3.5.3 Steganography vs Cryptography 🔑🖼️**

| Feature               | **Cryptography**                     | **Steganography**                                      |
| --------------------- | ------------------------------------ | ------------------------------------------------------ |
| **Purpose**           | Scrambles data into unreadable form  | Hides data within another file                         |
| **Visibility**        | Data looks random (ciphertext)       | Data looks normal (image/audio)                        |
| **Security**          | Protects confidentiality             | Protects secrecy of existence                          |
| **FinTech Relevance** | Encrypt wallet PINs, secure payments | Hackers may hide malware in images to bypass detection |

---

### **3.5.4 FinTech Connection 💳**

* Attackers sometimes use steganography to **hide malware inside harmless-looking files** (like company reports or scanned IDs).
* Security systems may allow such files, not realizing they contain hidden malicious code.
* Banks and FinTech apps must use **advanced malware scanners** that detect hidden payloads.

---

### **3.5.5 Defenses Against Steganography 🛡️**

1. **Steganalysis Tools 🔍** → Detect unusual patterns in files.
2. **Content Filtering 🚫** → Block suspicious image/audio uploads.
3. **File Integrity Checks ✅** → Verify that files haven’t been altered.
4. **Awareness** → Employees should be cautious of receiving files from untrusted sources.

---

✅ **Key Takeaway:**

* Cryptography protects **what data says**.
* Steganography hides **that data even exists**.
* In FinTech, attackers may exploit steganography for malware delivery, so defenses must go beyond simple encryption.

# 🔹 What is a Galois Field?

A **field** is a set of numbers where you can do **addition, subtraction, multiplication, division** (except by zero).

* A **Galois Field (GF)** is a *finite field*, meaning it only has a **fixed number of elements**.
* In AES, we work in:

  $$
  GF(2^8)
  $$

  👉 This means: all numbers are **8-bit (0–255)**.

So every operation in AES is done **modulo 2** on each bit (XOR instead of normal addition).

---

# 🔹 Elements in GF(2⁸)

* Any number is an **8-bit byte**. Example:

  ```
  11001011  (binary) = CB (hex) = 203 (decimal)
  ```

* Each byte is treated as a **polynomial** in $x$, with coefficients in $\{0,1\}$.

Example:
`11001011` → $x^7 + x^6 + x^3 + x + 1$

---

# 🔹 Operations in GF(2⁸)

### 1. **Addition**

* Done by XORing bits (no carry).
* Example:
  $57 \oplus 83 = D4$
  (binary: `01010111 XOR 10000011 = 11010100`)

---

### 2. **Multiplication**

* Multiply polynomials normally.
* Then reduce (mod) by the **irreducible polynomial** used in AES:

  $$
  m(x) = x^8 + x^4 + x^3 + x + 1
  $$

  (hex = **0x11B**)

👉 This keeps the result inside 8 bits.

---

# 🔹 Shortcuts Used in AES

Instead of doing full polynomial math, AES defines rules:

* **Multiply by 1** → value stays the same.
* **Multiply by 2** → left shift by 1 bit.

  * If the left shift overflows (≥ 256), XOR with `0x1B`.
* **Multiply by 3** → (value × 2) ⊕ value.

---

# 🔹 Example

Let’s compute $0xC6 \times 2$ in GF(2⁸):

1. `C6` = `11000110` (binary)
2. Shift left → `10001100` = `8C`
3. Overflow happened (bit 8 popped out), so XOR with `1B`:

   ```
   8C ⊕ 1B = 8F
   ```

✅ So $C6 × 2 = 8F$

---

# 🔹 Why GF(2⁸) in AES?

* 8 bits = 1 byte → perfect for digital systems.
* GF(2⁸) ensures math is reversible (needed for decryption).
* Using XOR and shifts makes AES **fast in hardware & software**.

---

## ✅ Easy Analogy

* **Normal math**: Adding 7 + 8 = 15
* **GF(2⁸) math**: Adding = XOR, Multiplying = shifting & reducing.
* It’s like playing with numbers inside a **box of 256 slots (0–255)**. Whenever you go out of the box, you wrap around using the polynomial rule.

# 🔐 Hashcat Tutorial — Easy to Understand

### What we’ll learn today

1. What a **hash** is.
2. How to create a hash from a password.
3. How attackers try to **crack hashes** using Hashcat.
4. How we can defend against these attacks.

---

## 🧩 Step 1 - What is a Hash?

* A **hash** is like a fingerprint of a password.
* Example:

  * Password: `Password123`
  * MD5 Hash: `42f749ade7f9e195bf475f37a44cafcb`

👉 You can’t “see” the password in the hash — but if the same password is hashed again, it gives the same result.

---

## 🖥️ Step 2 - Create a Hash

In Kali Linux, let’s make some hashes ourselves.

```bash
# MD5 hash of "Password123"
echo -n "Password123" | md5sum

# SHA1 hash of "letmein"
echo -n "letmein" | sha1sum
```

📌 Output will look like:

```
42f749ade7f9e195bf475f37a44cafcb  -
b7a875fc1ea228b9061041b7cec4bd3c52ab3ce3  -
```

These long strings are the **hashes**.

---

## ⚡ Step 3 - Install Hashcat (if not already)

```bash
sudo apt update
sudo apt install -y hashcat
```

---

## 📂 Step 4 - Prepare a Hash File

We need to give Hashcat a file that contains the hash.

```bash
# Save the MD5 hash into a file
echo "42f749ade7f9e195bf475f37a44cafcb" > md5_hash.txt
```

---

## 🗂️ Step 5 - Wordlist (rockyou.txt)

Kali has a built-in list of common passwords: **rockyou.txt**

```bash
# unzip if needed
sudo gzip -dk /usr/share/wordlists/rockyou.txt.gz
```

---

## 🔎 Step 6 - Run Hashcat (Dictionary Attack)

```bash
hashcat -m 0 -a 0 md5_hash.txt /usr/share/wordlists/rockyou.txt
```

* `-m 0` = MD5
* `-a 0` = dictionary attack
* `md5_hash.txt` = the hash file
* `/usr/share/wordlists/rockyou.txt` = list of guesses

📌 Expected output:

```
42f749ade7f9e195bf475f37a44cafcb:Password123
```

🎉 Hashcat found the password!


---
# Hashcat: `-m` (hash mode) and `-a` (attack mode)

---

## `-m` — select the **hash mode (type)**
The `-m` number tells Hashcat which hashing algorithm your target hashes use so it can compute candidate hashes the same way.

**Common `-m` values**
- `0`    = **MD5**  
- `100`  = **SHA1**  
- `1800` = **SHA512**  
- `1000` = **NTLM**  
- `3200` = **bcrypt**  
- `22000` = **WPA/WPA2** (newer Hashcat format)

> Use `hashcat --help` or `hashcat --example-hashes` to find the correct `-m` for other hash types.

**Why it matters:**  
If you pick the wrong `-m`, Hashcat will hash each guess with the wrong algorithm and will never match the target hash.

---

## `-a` — select the **attack mode (strategy)**
The `-a` number chooses how Hashcat generates and tests candidate passwords.

**Common `-a` values**
- `-a 0` = **Dictionary attack** — test words from a wordlist (e.g., `rockyou.txt`).  
- `-a 1` = **Combination attack** — combine words from two wordlists.  
- `-a 3` = **Mask attack** — patterned brute force (e.g., `?u?l?l?l?d?d`).  
- `-a 6` = **Hybrid (wordlist + mask appended)** — word + pattern suffix.  
- `-a 7` = **Hybrid (mask prepended + wordlist)** — pattern prefix + word.

**Why it matters:**  
Different password habits require different attacks: dictionary attacks work on common words; mask attacks work well if you know a pattern (like `Upper + 5 lower + 2 digits`).

---

## Quick examples (for reference)
```bash
# MD5 dictionary attack (show status every 10s)
hashcat -m 0 -a 0 md5_hash.txt /usr/share/wordlists/rockyou.txt --status --status-timer=10

# Mask attack (pattern: 1 uppercase, 5 lower, 2 digits)
hashcat -m 0 -a 3 md5_hash.txt '?u?l?l?l?l?l?d?d'

---

## 🎭 Step 7 - Mask Attack (Pattern Guessing)

If we know the format (e.g., 1 uppercase + 5 lowercase + 2 numbers):

```bash
hashcat -m 0 -a 3 md5_hash.txt '?u?l?l?l?l?l?d?d'
```

📌 Mask:

* `?u` = uppercase letter
* `?l` = lowercase letter
* `?d` = digit

---

## 📒 Step 8 - Show Results

After cracking, run:

```bash
hashcat --show -m 0 md5_hash.txt
```

📌 Output:

```
42f749ade7f9e195bf475f37a44cafcb:Password123
```

---

## 🚨 Step 9 - Why This Matters

* If passwords are weak, Hashcat cracks them in seconds.
* If websites store **MD5/SHA1 without salt**, attackers can recover thousands of passwords quickly.

---

## 🛡️ Step 10 - How to Defend

1. Use **long, unique passphrases**.
2. Store passwords with **salt + slow hashing** (bcrypt, Argon2).
3. Use **multi-factor authentication (MFA)**.
4. Limit login attempts (**lockout after failures**).

````python
# ==============================================
# 📘 Interactive Steganography Demo (LSB) with ipywidgets
# ==============================================

# Install dependencies (run this once if using Colab/Jupyter the first time)
# !pip install pillow ipywidgets numpy

# Import libraries
from PIL import Image             # Pillow: for image handling
import numpy as np                # NumPy: for pixel array manipulation
import io                         # For handling uploaded files as byte streams
import ipywidgets as widgets      # For interactive UI components in Jupyter
from IPython.display import display, clear_output  # For showing widgets and clearing outputs

# -----------------------------
# Step 1: Create Upload Widgets
# -----------------------------
# Upload widget for encoding (to hide a message inside an image)
encode_upload_btn = widgets.FileUpload(
    accept="image/*",  # Accept only image files (PNG/JPG)
    multiple=False     # Only one file at a time
)

# Upload widget for decoding (to read a hidden message from an encoded image)
decode_upload_btn = widgets.FileUpload(
    accept="image/*",  # Accept only image files
    multiple=False
)

# Output area where we will show status messages and results
out = widgets.Output()

# Text box where students type their secret message
text_box = widgets.Textarea(
    value="",                                  # Default empty
    placeholder="Type message here to hide in image...",  # Hint for students
    description="Message:",                    # Label
    layout=widgets.Layout(width="600px", height="100px")  # Size of the text box
)

# Button to trigger encoding (hiding message)
encode_btn = widgets.Button(
    description="🔒 Hide Message in Image",  # Button label
    button_style="success"                   # Green color
)

# Button to trigger decoding (revealing message)
decode_btn = widgets.Button(
    description="🔓 Decode Hidden Message",  # Button label
    button_style="info"                      # Blue color
)

# -----------------------------
# Step 2: LSB Steganography Functions
# -----------------------------
def lsb_encode(img, message):
    """
    Hide a secret message inside the image using LSB (Least Significant Bit).
    Each pixel channel (R, G, B) stores one bit of the message.
    """
    img = img.convert("RGB")                        # Ensure image is in RGB format
    data = np.array(img, dtype=np.uint8)            # Convert image to NumPy array

    # Add a delimiter at the end so we know when to stop reading the hidden text
    message += "#####END#####"

    # Convert the message into binary (string of 0s and 1s)
    binary = ''.join(format(ord(c), '08b') for c in message)

    # Flatten image pixels into a 1D array (easier to modify bits)
    flat = data.flatten()

    # Check if the image can store the whole message
    if len(binary) > len(flat):
        raise ValueError("Message too long for this image!")

    # Loop over each bit of the message and hide it in the LSB of each pixel value
    for i in range(len(binary)):
        flat[i] = (flat[i] & 0b11111110) | int(binary[i])
        # (flat[i] & 0b11111110) clears the last bit
        # | int(binary[i]) sets the last bit to our message bit

    # Reshape the modified array back into the original image shape
    encoded_data = flat.reshape(data.shape)

    # Convert NumPy array back to an image
    return Image.fromarray(encoded_data.astype(np.uint8), "RGB")


def lsb_decode(img):
    """
    Extract a hidden message from an image using LSB.
    Reads the least significant bits until the delimiter #####END##### is found.
    """
    img = img.convert("RGB")                        # Ensure RGB format
    data = np.array(img, dtype=np.uint8).flatten()  # Flatten into 1D array

    # Collect all LSBs (last bit of each pixel value)
    bits = [str(data[i] & 1) for i in range(len(data))]

    chars = []  # To store decoded characters
    # Process 8 bits at a time → 1 byte → 1 character
    for i in range(0, len(bits), 8):
        byte = bits[i:i+8]                          # Take 8 bits
        char = chr(int(''.join(byte), 2))           # Convert binary → ASCII character
        chars.append(char)

        # Stop if delimiter is found
        if ''.join(chars).endswith("#####END#####"):
            return ''.join(chars).replace("#####END#####", "")

    return ""  # Return empty if nothing found

# -----------------------------
# Step 3: Helper Function
# -----------------------------
def load_image(file_upload):
    """Convert uploaded file into a PIL Image object"""
    file_content = list(file_upload.value.values())[0]['content']  # Extract raw bytes
    return Image.open(io.BytesIO(file_content))  # Open as image

# -----------------------------
# Step 4: Button Actions
# -----------------------------
def on_encode_clicked(b):
    """Action when 'Hide Message' button is clicked"""
    out.clear_output()  # Clear previous output
    with out:
        if not encode_upload_btn.value:
            print("⚠️ Please upload an image first!")
            return
        if not text_box.value.strip():
            print("⚠️ Please type a message to hide in the image!")
            return

        img = load_image(encode_upload_btn)   # Load uploaded image
        print("✅ Image uploaded successfully!")

        try:
            encoded_img = lsb_encode(img, text_box.value)  # Hide message
            encoded_img.save("encoded_image.png", "PNG")   # Save new image
            print("✅ Message hidden inside the image!")
            print("💾 Saved as 'encoded_image.png'")
            display(encoded_img)                           # Display encoded image
        except ValueError as e:
            print("❌ Error:", e)

def on_decode_clicked(b):
    """Action when 'Decode Message' button is clicked"""
    out.clear_output()
    with out:
        if not decode_upload_btn.value:
            print("⚠️ Please upload an encoded image first!")
            return

        img = load_image(decode_upload_btn)    # Load uploaded image
        print("✅ Encoded image uploaded successfully!")

        hidden_text = lsb_decode(img)          # Extract hidden text
        if hidden_text:
            print("🔓 Hidden message found:")
            print("--------------------------------------------------")
            print(hidden_text)
            print("--------------------------------------------------")
        else:
            print("❌ No hidden message found in this image.")

# Attach the functions to buttons
encode_btn.on_click(on_encode_clicked)
decode_btn.on_click(on_decode_clicked)

# -----------------------------
# Step 5: Display the UI
# -----------------------------
# Stack all widgets in a vertical box
display(widgets.VBox([
    widgets.Label("📂 Upload an Image to Hide Message"),
    encode_upload_btn,
    text_box,
    encode_btn,
    widgets.Label("📂 Upload an Image to Decode Message"),
    decode_upload_btn,
    decode_btn,
    out
]))
````

# 🌍 1. Introduction to Asymmetric Cryptography

---

## 🔑 1.1 What is Asymmetric Encryption?

- **Definition:**  
  Asymmetric encryption (also called **public-key cryptography**) uses **two different keys**:  
  - **Public Key** → shared openly (like your email address).  
  - **Private Key** → kept secret (like your email password).  

- **How it works (Step by Step):**  
  1. Alice wants to send transaction details to her bank.  
  2. She encrypts the message using the **bank’s public key**.  
  3. Only the **bank’s private key** can unlock (decrypt) the message.  

- **Analogy (Mailbox):**  
  - Everyone can drop a letter into the mailbox (public slot = **public key**).  
  - Only the owner with the **mailbox key = private key** can open and read it.  

👉 This removes the **key sharing problem** found in symmetric encryption.  

---

## ⚖️ 1.2 Difference Between Symmetric & Asymmetric Encryption

| Feature | Symmetric Encryption 🔑 | Asymmetric Encryption 🔑🔑 |
|---------|-------------------------|----------------------------|
| **Keys Used** | One single key for both encryption & decryption | Public key (encrypt) + Private key (decrypt) |
| **Speed** ⚡ | Very fast | Slower (complex math) |
| **Security** 🔐 | Risk if the key is stolen | More secure (private key stays secret) |
| **Key Sharing** 📦 | Hard to share safely | No issue – public key can be shared openly |
| **Best Use** 💻 | Encrypting large amounts of data | Secure communication, authentication, signatures |
| **FinTech Example** 💳 | Encrypting ATM logs with AES | Online banking (HTTPS with RSA/ECC) |

**Analogy:**  
- Symmetric → One key 🔑 like a **locker key** both people use.  
- Asymmetric → Two keys 🔑🔑 like a **padlock** that anyone can close, but only the owner can open.  

---

## 💳 1.3 Why Asymmetric Cryptography is Important in Cybersecurity & FinTech

1. **Secure Communication 📡**  
   - Online banking apps/websites use public/private keys to protect PINs, OTPs, and card numbers.  

2. **Key Exchange 🔄**  
   - Asymmetric cryptography is used in **TLS handshakes** to exchange symmetric session keys safely.  
   - After this, **AES** (faster) is used for actual transactions.  

3. **Authentication & Digital Trust ✅**  
   - Digital certificates (SSL/TLS) confirm that you are really talking to your bank, not a fake site.  

4. **Digital Signatures ✍️**  
   - Transactions can be signed with a private key.  
   - The bank verifies using the public key, ensuring the request is authentic and unaltered.  

5. **Blockchain & Cryptocurrencies 🪙**  
   - Your **private key** signs a crypto transaction.  
   - The **public key** lets anyone verify it, preventing fraud.  

---

### ✅ Key Takeaway  
- **Symmetric = one key, fast, but risky.**  
- **Asymmetric = two keys, slower, but solves the key-sharing problem.**  
- In **FinTech**, asymmetric cryptography powers **banking apps, payment gateways, mobile wallets, and blockchain systems**.  


# 🔑 2. RSA Encryption

---

## 📜 2.1 History and Significance of RSA

- **RSA** stands for **Rivest–Shamir–Adleman**, the three scientists who invented it in **1977**.  
- It was the **first practical asymmetric (public-key) encryption algorithm**.  

### 🔐 Why RSA was revolutionary?
Before RSA, encryption relied on **symmetric keys** (the same key for encryption and decryption).  
👉 Problem: How do two people who have never met **share the secret key safely** over the internet?  

RSA solved this by introducing the idea of:
- A **Public Key** → can be shared with the world.  
- A **Private Key** → kept secret and never shared.  

Now, anyone can encrypt a message using your public key, but **only you** can decrypt it with your private key.  

### 🏦 Where RSA is used today (especially in finance & banking)
- **Secure websites (HTTPS/TLS)** → protects online banking and shopping.  
- **ATMs** → encrypts PINs and messages between ATM and bank server.  
- **Payment gateways** → like Visa, Mastercard, Stripe, PayPal.  
- **Digital Signatures** → used for electronic contracts and authorizing transactions.  


---
## 📐 2.3 Mathematical Foundation of RSA

RSA security comes from **number theory** — the study of properties of integers.  
Let’s go step by step.

---

### 🔹 1. Prime Numbers

- **Definition:** A prime number is a whole number greater than 1 that has **only two divisors: 1 and itself**.  
- Examples: 2, 3, 5, 7, 11, 13…  
- Non-examples: 4 (divisible by 2), 9 (divisible by 3).  

👉 In RSA, primes are the **building blocks**. We pick **two large primes**, `p` and `q`, and multiply them.  

---

### 🔹 2. Prime Factorization

- **Definition:** Breaking a number into the product of its prime factors.  
- Example: 12 = 2 × 2 × 3.  
- Example: 91 = 7 × 13.  

👉 Why important?  
- Multiplying primes (easy).  
- Going backward (finding primes from a huge number) is **very hard**.  

RSA relies on this:  
\[
n = p \times q
\]  
If `n` is very large (hundreds of digits), finding `p` and `q` is practically impossible with current computers.  
This is called the **factoring problem**.  

---

### 🔹 3. Modulus (mod)

- **Definition:** The **remainder** when dividing one number by another.  
- Notation: `a mod n = remainder of a ÷ n`.  
- Example: `17 mod 5 = 2` (since 17 ÷ 5 = 3 remainder 2).  

👉 In RSA, both encryption and decryption are done under a modulus `n`.  
It keeps numbers within the “playground” of 0 to n-1.  

**Analogy:** Imagine a clock ⏰ with 12 hours.  
- 14 o’clock = 2 o’clock (because 14 mod 12 = 2).  
- Modulus works like **wrapping numbers around a circle**.

---

### 🔹 4. Euler’s Totient Function (φ)

- **Definition:** φ(n) = number of integers less than `n` that are **coprime** with `n`.  
- For RSA:  
\[
φ(n) = (p - 1)(q - 1)
\]  

**Coprime definition:**  
- Two numbers are coprime if their **greatest common divisor (GCD)** is 1.  
- Example:  
  - GCD(9, 28) = 1 → coprime ✅  
  - GCD(8, 12) = 4 → not coprime ❌  

👉 We need φ(n) because it helps us design keys `e` and `d` that work as mathematical opposites.

---

### 🔹 5. Public Exponent (e)

**Definition:**  
The **public exponent (e)** is a number used during **encryption**.  
It must satisfy two conditions:
1. **Range:** `1 < e < φ(n)`  
   (It has to be greater than 1 and smaller than φ(n)).  
2. **Coprime Condition:** GCD(e, φ(n)) = 1  
   (That means `e` and φ(n) share no common factors other than 1).  

👉 Why these conditions?  
- If `e` shares a factor with φ(n), then we can’t find the matching private key `d`.  
- Coprime condition ensures that an inverse exists (this will be our private exponent `d`).  

---

**Example:**  
- Suppose φ(n) = 3120.  
- If we choose `e = 17`:  
  - GCD(17, 3120) = 1 ✅ good → 17 is valid.  

If we chose `e = 18`:  
- GCD(18, 3120) = 6 ❌ not valid → 18 would not work.  

---

**Practical choice:**  
- In real-world RSA, almost everyone uses **e = 65537**.  
- Why?  
  - 65537 is a prime number.  
  - It makes encryption fast.  
  - It is secure against known attacks.  

So, even though `e` can be many numbers, in practice, systems usually **fix e = 65537** and only vary the other parts.

👉 `e` becomes part of the **public key (e, n)**.  

---

### 🔹 6. Modular Inverse (finding d)

Now that we picked `e`, we must find the **private exponent (d)**.  
This is the number used during **decryption**.  

**Definition:**  
The modular inverse of `e` modulo φ(n) is the number `d` such that:  

\[
(e \times d) \bmod φ(n) = 1
\]  

👉 This means: when you multiply `e` and `d`, and divide by φ(n), the remainder must be 1.  

---

**Example:**  
- φ(n) = 3120  
- e = 17  
- Solve for d:  
  \[
  (17 \times d) \bmod 3120 = 1
  \]  
- Answer: `d = 2753`.  

Check:  
\[
17 \times 2753 = 46801
\]  
\[
46801 \bmod 3120 = 1
\] ✅ Correct!  

So:  
- Public Key = (e, n) = (17, 3233)  
- Private Key = (d, n) = (2753, 3233)  

---

**Analogy:**  
- Imagine encryption is like **turning a combination lock forward** 🔒 by `e` steps.  
- Decryption is like **turning it backward** 🔑 by `d` steps.  
- Because `d` is the modular inverse of `e`, the two motions exactly cancel out.  

---

### 🔹 7. Why It All Works Together

The big question:  
👉 Why does RSA guarantee that when you **scramble a message** with the public key, you can always **unscramble it back** with the private key?

---

#### 🧩 Step 1: The Forward Move (Scrambling)

- When you encrypt, you take the message `M` and mix it up by raising it to the power of the public exponent `e`.  
- Then you “wrap” it using the modulus `n` (like numbers wrapping around a clock).  
- Result: ciphertext `C`.  

**Example:**  
`M = 123 → C = 855` using `(e = 17, n = 3233)`.

---

#### 🧩 Step 2: The Backward Move (Unscrambling)

- When you decrypt, you take the ciphertext `C` and raise it to the power of the private exponent `d`.  
- Again, you wrap it with the same modulus `n`.  
- Result: you get back the original message `M`.  

**Example:**  
`C = 855 → M = 123` using `(d = 2753, n = 3233)`.

---

#### 🧩 Step 3: Why Do They Cancel Out?

Here’s the secret:  
- We carefully choose `d` so that it is the **mathematical opposite** of `e` (mod φ(n)).  
- This means the two steps — “raise to e” then “raise to d” — always cancel out under the modulus.  

Think of it like:  
- Turning a lock forward 17 times (encrypt).  
- Then turning it backward 2753 times (decrypt).  
- Because the lock dial is on a circle (modulus n), you always land back where you started.  

---

#### 🧩 Step 4: Euler’s Theorem (in plain words)

Euler’s theorem is the math rule behind the magic.  
It says:  
- If a number and `n` don’t share factors, then after wrapping (modulus), raising the number many times eventually brings you back to **1**.  
- RSA takes advantage of this: it sets up `e` and `d` so that the forward move and backward move line up perfectly.  

---

### 🎯 Final Recap

1. **Encryption = scramble forward with e.**  
2. **Decryption = unscramble backward with d.**  
3. Because of how e and d are chosen, the two steps always cancel out.  
4. That’s why:  
   - Encrypt then decrypt → you get the original message back.  
   - Decrypt then encrypt (in digital signatures) → also works the other way.  

✅ RSA works because encryption and decryption are like perfectly matched opposite actions on a circular lock (the modulus n).


---

**Analogy Recap:**  
- Public exponent `e` = turning the lock forward 🔒.  
- Private exponent `d` = turning it backward 🔑.  
- The modulus `n` = the size of the lock’s wheel (like a clock face).  
- Together, `e` and `d` ensure that the lock can always be closed by anyone (encryption), but only opened by the owner (decryption).

---

✅ That’s why RSA is secure and reliable.

---
### 🎯 Summary of Foundations

1. **Prime numbers** → building blocks.  
2. **Factorization problem** → hard to break.  
3. **Modulus (mod)** → keeps math in the right range.  
4. **Euler’s Totient φ(n)** → key for choosing exponents.  
5. **Coprime numbers** → ensure e works.  
6. **Modular inverse** → ensures decryption is possible.  
7. **Euler’s theorem** → proves RSA correctness.  

✅ Together, these pieces create a secure and functional encryption system.

## ⚙️ 2.2 How RSA Works (Key Generation → Encryption → Decryption)

RSA works in **three phases**:  
1. **Key Generation** → making the public and private keys.  
2. **Encryption** → locking a message with the public key.  
3. **Decryption** → unlocking the message with the private key.  

We’ll explain each part with a small number example.  
(Remember: real RSA uses **huge numbers**—hundreds of digits long. Our example uses small numbers so you can follow the math.)

---

### 🧮 A) Key Generation (Making the Keys)

**Goal:** Create two keys:  
- **Public Key = (e, n)** → can be given to everyone (used to encrypt messages).  
- **Private Key = (d, n)** → must be kept secret (used to decrypt messages).  

RSA relies on some clever number theory. Let’s see how these keys are made.

---

#### 🔹 Step 1: Pick two prime numbers (p and q)

- A **prime number** is a number greater than 1 that has **no divisors except 1 and itself**.  
  - Examples: 2, 3, 5, 7, 11, 13, 17…  
  - Non-examples: 8 (divisible by 2 and 4), 21 (divisible by 3 and 7).  

👉 In RSA, we pick **very large prime numbers** (hundreds of digits long).  
👉 For teaching, let’s pick small ones:  
- `p = 61`  
- `q = 53`  

Why primes?  
- Multiplying them is easy.  
- But factorizing (splitting their product back into primes) is very hard when numbers are huge. This is the **basis of RSA security**.

---

#### 🔹 Step 2: Multiply them to get modulus (n)

\[
n = p \times q
\]

- Calculation: `n = 61 × 53 = 3233`  
- This number `n` is called the **modulus**.  

👉 Both the public key and private key will include this same modulus.  

**Analogy:** Think of `n` as the “arena” or “playground” 🏟️ where all the encryption and decryption operations will happen.  
- Messages are “locked” and “unlocked” inside this arena using the keys.

---

#### 🔹 Step 3: Compute Euler’s Totient φ(n)

\[
φ(n) = (p - 1)(q - 1)
\]

- Calculation: `(61 – 1)(53 – 1) = 60 × 52 = 3120`  
- So, `φ(n) = 3120`.  

**But what is φ(n)?**  
- This is **Euler’s Totient function**.  
- It tells us **how many numbers below n are coprime with n**.  

**Coprime** = two numbers that share no common factors other than 1.  
- Example: 9 and 28 → GCD = 1 → coprime ✅  
- Example: 8 and 12 → GCD = 4 → not coprime ❌  

👉 We use φ(n) because it helps us design exponents (`e` and `d`) that are mathematical “inverses” of each other.

---

#### 🔹 Step 4: Choose a Public Exponent (e)

Now, we need to pick a number `e` such that:  
- `e` is **greater than 1** and **less than φ(n)`.  
- `e` is **coprime with φ(n)`.  

Why?  
- This ensures that `e` can be used to **encrypt** messages reliably and can later be “reversed” with the private key.  

👉 In practice, almost everyone uses `e = 65537`.  
- It is a safe prime.  
- It makes calculations fast and secure.  

👉 For our small example, we choose:  
- `e = 17` (since 17 and 3120 are coprime).  

---

#### 🔹 Step 5: Compute the Private Exponent (d)

We now calculate `d` (the private exponent).  
It must satisfy the equation:  

\[
(e \times d) \bmod φ(n) = 1
\]

This means:  
- When you multiply `e × d` and divide by `φ(n)`, the **remainder is 1**.  
- In other words, `d` is the **modular inverse** of `e` (with respect to φ(n)).  

**Example:**  
- `e = 17`  
- `φ(n) = 3120`  
- Solve for `d`:  
  - `(17 × d) mod 3120 = 1`  
  - Answer: `d = 2753`  

👉 This ensures that encryption with `e` can always be undone (decrypted) with `d`.

---

### 🎉 Final Keys (from this example)

- **Public Key = (e, n) = (17, 3233)**  
  → Used by anyone to encrypt a message.  

- **Private Key = (d, n) = (2753, 3233)**  
  → Used only by the owner to decrypt messages.  

---

### 🧩 Recap (Why each step matters)

1. Pick primes `p` and `q` → provides the “building blocks” for security.  
2. Multiply → `n = p × q` is the modulus (common part of both keys).  
3. Compute φ(n) → foundation for choosing `e` and `d`.  
4. Choose `e` → public exponent (must be coprime with φ(n)).  
5. Compute `d` → private exponent (inverse of `e` under mod φ(n)).  

✅ Together: `(e, n)` + `(d, n)` = **RSA key pair**.  

---

### 🔒 B) Encryption (Locking the Message)

Now that we have keys, let’s see how encryption works.

**Rule / Formula:**

\[
C = (M^e) \bmod n
\]

Where:
- `M` = the original message (called **plaintext**) expressed as a number.  
- `e` = the public exponent (from the public key).  
- `n` = modulus (shared by both keys).  
- `C` = ciphertext (the encrypted message).  

---

#### 🧩 Step-by-Step Explanation

1. **Convert the message into a number (`M`)**  
   - Computers deal with numbers, not words.  
   - If your message is text like "HELLO", it is converted into numbers using ASCII or Unicode codes.  
   - In our tiny example, we directly take `M = 123`.  

2. **Raise `M` to the power of `e`**  
   - This is the **core transformation** in RSA.  
   - Example: `M = 123`, `e = 17`.  
   - Compute \( 123^{17} \).  
   - This number will be *huge*, but don’t worry — the next step reduces it.  

3. **Apply modulus `n` (the playground)**  
   - Remember, RSA math happens inside the “arena” defined by `n`.  
   - We take the remainder after dividing by `n`.  
   - Example: `n = 3233`.  
   - So we calculate \( 123^{17} \bmod 3233 \).  

4. **Result is ciphertext (`C`)**  
   - In our example:  
     \[
     C = 123^{17} \bmod 3233 = 855
     \]  
   - The ciphertext = **855**.  

---

#### 🔐 Why is this secure?

- Anyone can see `C = 855`.  
- Without the private key, turning it back into 123 is nearly impossible because it would require solving the **modular exponentiation** in reverse → which reduces to the **prime factorization problem** (extremely hard with large numbers).  

**Analogy:**  
Think of encryption as locking a box 🗃️ with a padlock 🔒 (public key). Anyone can close the lock, but only the person with the unique key (private key) can open it.  

---

### 🔓 C) Decryption (Unlocking the Message)

Once the recipient receives ciphertext `C`, they use their **private key** `(d, n)` to unlock it.

**Rule / Formula:**

\[
M = (C^d) \bmod n
\]

Where:
- `C` = ciphertext.  
- `d` = private exponent (from the private key).  
- `n` = modulus.  
- `M` = plaintext message (the original).  

---

#### 🧩 Step-by-Step Explanation

1. **Take the ciphertext (`C`)**  
   - From encryption: `C = 855`.  

2. **Raise it to the power of `d` (the private exponent)**  
   - `d` is the mathematical “inverse” of `e`.  
   - In our example: `d = 2753`.  
   - Compute \( 855^{2753} \).  
   - Again, this number is **huge**, but the modulus will reduce it.  

3. **Apply modulus `n`**  
   - Use the same modulus from before: `n = 3233`.  
   - Compute:  
     \[
     M = 855^{2753} \bmod 3233
     \]  

4. **Recover the original message (`M`)**  
   - Result: `M = 123`.  

---

#### 🎉 Example Recap

- Public Key = `(17, 3233)`  
- Private Key = `(2753, 3233)`  
- Original message: `M = 123`  
- Encrypted ciphertext: `C = 855`  
- Decrypted back: `M = 123` ✅  

---

### 🧠 Why does decryption work?

- Because we chose `d` such that:  
  \[
  (e \times d) \bmod φ(n) = 1
  \]  
- This ensures that encrypting then decrypting is the same as doing nothing (modulo n).  
- In math terms:  
  \[
  (M^e)^d \bmod n = M^{ed} \bmod n = M
  \]  

---

**Analogy:**  
Encryption = putting your message in a box and locking it with a padlock (public key).  
Decryption = using the secret key that fits that lock (private key).  
Without the private key, the box cannot be opened.  

---

### 🧠 Why does this work?

- Because of how `e` and `d` are chosen, they are **mathematical inverses** under modulo φ(n).  
- Encrypting with `e` and then decrypting with `d` brings you back to the original message.  

**Analogy:**  
- Public key = a **lock** 🔒 that anyone can close.  
- Private key = the **unique key** 🔑 that can open that lock.  

---

## ✅ Quick Recap (So Far)

- RSA uses **two keys**: public (encrypt) and private (decrypt).  
- **Key generation**: choose primes → multiply → totient → choose `e` → compute `d`.  
- **Encryption formula**: `C = (M^e) mod n`  
- **Decryption formula**: `M = (C^d) mod n`  
- Real RSA uses **very large primes** (hundreds of digits) to make factorization impossible.  



# ➰ 3. Elliptic Curve Cryptography (ECC)

---

## 📌 3.1 Introduction to Elliptic Curves in Cryptography

- An **elliptic curve** is a special type of curve defined by an equation like:  
  \[
  y^2 = x^3 + ax + b
  \]  
  where `a` and `b` are constants.  
- The curve looks like a **smooth, looping curve** in a coordinate plane.  

👉 In ECC, we don’t use the curve for drawing pictures, but for **mathematical operations**.  
Points on the curve can be combined in a way that’s similar to addition, but much more powerful.

**Why use curves?**  
- They create very hard math problems (Elliptic Curve Discrete Logarithm Problem).  
- This makes ECC secure with **much smaller keys** compared to RSA.

---

## 📌 3.2 How ECC Works (Key Generation on Curves)

ECC uses points on the elliptic curve instead of large prime multiplications.

1. **Choose a curve**  
   - Both sides (sender & receiver) agree on a standard curve (e.g., secp256k1 in Bitcoin).

2. **Pick a base point (G)**  
   - This is a publicly known starting point on the curve.

3. **Private Key**  
   - A random number `k`.  
   - Example: `k = 123456`.

4. **Public Key**  
   - Multiply the base point `G` by the private key `k` using elliptic curve math:  
     \[
     P = k \times G
     \]  
   - Result is another point on the curve.  
   - Example: P = (x, y).

👉 The public key is a **point on the curve**, the private key is just a number.  
- **Encrypt:** Use the public key point.  
- **Decrypt:** Use the private number.

**Analogy:**  
- Think of the base point `G` as a direction.  
- Private key = how many steps you walk.  
- Public key = the final location.  
- If someone only sees your location, it’s nearly impossible to guess exactly how many steps you took.

---

## 📌 3.3 Strengths of ECC ✅

1. **Smaller Key Sizes**  
   - ECC provides the same security as RSA with much shorter keys.  
   - Example:  
     - RSA 3072-bit ≈ ECC 256-bit security.  

2. **Efficiency**  
   - Faster computations.  
   - Requires less memory and bandwidth.  

3. **Scalable Security**  
   - Works well even on small devices (smartphones, IoT).  

4. **Stronger Security per Bit**  
   - Attacking ECC (solving discrete logarithm problem) is much harder than factoring RSA numbers.

---

## 📌 3.4 Weaknesses & Challenges ❌

1. **Complex Math**  
   - Harder for beginners to understand and implement correctly.  
   - Mistakes in implementation = vulnerabilities.

2. **Quantum Threat**  
   - Future quantum computers could break ECC using **Shor’s algorithm**.  

3. **Standardization Issues**  
   - Choosing secure curves is tricky.  
   - Some curves have potential backdoors if not chosen carefully.  

4. **Patent & Licensing (historically)**  
   - Earlier, ECC was restricted by patents (less of an issue today).  

---

## 📌 3.5 FinTech Applications of ECC 💰

ECC is widely used in financial technologies because it is **lightweight, secure, and efficient**.

1. **Mobile Wallets**  
   - ECC secures transactions on Apple Pay, Google Pay, and other wallets.  

2. **Blockchain & Cryptocurrencies**  
   - Bitcoin, Ethereum, and most blockchains use ECC (secp256k1 curve) for digital signatures.  
   - Every wallet address is derived from an ECC public key.

3. **Secure Payment Gateways**  
   - ECC-based TLS certificates ensure safe communication between banking apps and servers.  

4. **Smart Cards and IoT Finance**  
   - ECC is used in **low-power devices** like ATMs, contactless cards, and IoT payment devices.

---

### 🎯 Summary

- ECC = uses **elliptic curves** instead of large primes (RSA).  
- **Private key = number**, **Public key = point on the curve**.  
- Stronger and faster than RSA with smaller keys.  
- Powering today’s **mobile wallets, blockchain, and secure payment systems**.  


# ⚖️ 4. RSA vs ECC

---

## 📌 4.1 Comparison of Key Size, Speed, and Security

| Feature | RSA 🔑 | ECC ➰ |
|---------|--------|--------|
| **Key Size** | Requires very large keys for strong security (2048–4096 bits common). | Much smaller keys give same security (256–521 bits). |
| **Security Basis** | Based on difficulty of **prime factorization** (splitting a big number into two primes). | Based on **Elliptic Curve Discrete Logarithm Problem** (much harder to solve). |
| **Performance / Speed** | Slower for encryption, decryption, and key generation. | Faster computations, especially for signing and verifying. |
| **Memory / Bandwidth** | Heavy → larger certificates, more data to transmit. | Lightweight → smaller certificates, ideal for mobile/IoT. |
| **Energy Efficiency** | Not ideal for low-power devices (needs more processing). | Excellent for smartphones, IoT, and constrained devices. |
| **Typical Key Example** | RSA 3072-bit ≈ ECC 256-bit (same level of security). | ECC 256-bit = RSA 3072-bit (but with much smaller key). |

---

## 📌 4.2 Which is Better for FinTech and Why?

- **ECC is generally better for FinTech** because:
  1. 🔋 **Efficiency**: Mobile wallets, ATMs, IoT payment devices need lightweight cryptography.  
  2. 🚀 **Speed**: ECC provides faster transaction signing and verification than RSA.  
  3. 🔐 **Security with Small Keys**: A 256-bit ECC key is as secure as a 3072-bit RSA key.  
  4. 📱 **Scalability**: Modern financial apps (Apple Pay, Google Pay) prefer ECC for secure communication.  

- **RSA is still widely used** in:
  - Older banking infrastructure, certificates, and legacy systems.  
  - Systems where compatibility matters more than efficiency.  

---

## 📌 4.3 Real-World Usage

### 🏦 Banks & Payment Systems  
- Many older banks still use **RSA-based SSL/TLS certificates** for secure communication.  
- Newer FinTech startups and mobile banks are moving to **ECC-based certificates** for efficiency.  

### 📱 Mobile Apps  
- **Apple Pay, Google Pay, Samsung Pay** → use ECC for secure digital wallet transactions.  
- ECC is embedded in **TLS handshakes** to ensure fast, secure connections on mobile devices.  

### ⛓️ Blockchain Platforms  
- **Bitcoin** → uses ECC (curve secp256k1) for wallet addresses and digital signatures.  
- **Ethereum** → also uses ECC (secp256k1).  
- **Hyperledger & modern blockchains** → rely on ECC for lightweight digital signatures.  

---

### 🎯 Summary

- **RSA** = older, reliable, widely supported, but heavy and slow.  
- **ECC** = modern, faster, more secure per bit, and better suited for FinTech apps.  
- The industry is shifting toward **ECC**, especially in mobile banking, wallets, and blockchain.  


# ✍️ 5. Digital Signatures

---

## 📌 5.1 What Are Digital Signatures?

- A **digital signature** is the electronic version of a handwritten signature ✍️ or a wax seal 🕯️.  
- It proves that a **message, document, or transaction** really came from the sender and was not altered.  

**Definition:**  
A digital signature is a **cryptographic technique** that ensures:
- The sender of a message is **authentic**.  
- The content of the message is **not tampered with**.  

👉 Purpose: Build **trust** in digital communication, especially in banking, finance, and online payments.

---

## 📌 5.2 How Digital Signatures Work

Digital signatures use **asymmetric cryptography (public/private keys)** and **hashing**.  

### 🔹 Step 1: Hashing
- A **hash function** converts the original message into a fixed-size “fingerprint” (hash value).  
- Example:  
  - Message: "Transfer $100"  
  - Hash → `A94A8FE5...`  

👉 Even if one letter changes, the hash changes completely.

---

### 🔹 Step 2: Signing with Private Key
- The sender takes the hash value and encrypts it with their **private key**.  
- This encrypted hash = **digital signature**.  

---

### 🔹 Step 3: Verification with Public Key
- The receiver:
  1. Decrypts the signature with the sender’s **public key** to get the hash.  
  2. Recomputes the hash of the received message.  
  3. Compares both hashes.  

👉 If they match → the message is valid and untampered.  

---

**Analogy:**  
- Imagine signing a paper with a unique stamp.  
- Anyone can check the stamp’s authenticity, but only you can create it.  

---

## 📌 5.3 Benefits of Digital Signatures ✅

1. **Authenticity**  
   - Confirms that the sender is really who they claim to be.  
   - Prevents impersonation.

2. **Integrity**  
   - Ensures the message/document hasn’t been changed during transmission.  
   - Even one small edit makes the signature invalid.

3. **Non-repudiation**  
   - The sender cannot later deny (“repudiate”) having sent the message.  
   - Legally binding in many countries.

---

## 📌 5.4 FinTech Applications 💰

1. **Loan Agreements**  
   - Customers can sign loan documents digitally with cryptographic signatures.  
   - Legally valid in many jurisdictions.

2. **Online Contracts**  
   - FinTech platforms allow customers to e-sign contracts with banks or investment firms.

3. **Secure Transactions**  
   - Payment gateways use digital signatures to validate transactions.  
   - Example: When you pay online, your bank verifies your transaction request using digital signatures.

4. **Cryptocurrency & Blockchain**  
   - Every transaction in Bitcoin/Ethereum is approved with a digital signature (ECC-based).  
   - Proves ownership of funds without revealing private keys.

---

### 🎯 Summary

- **Digital Signatures = digital fingerprints** that prove authenticity, integrity, and accountability.  
- They combine **hashing** + **asymmetric encryption**.  
- In FinTech, they secure **contracts, transactions, wallets, and blockchain systems**.  


# 🏛️ 6. Public Key Infrastructure (PKI)

---

## 📌 6.1 What is PKI?

- **Public Key Infrastructure (PKI)** is the framework that manages **digital certificates** and **public/private keys**.  
- It provides the foundation for **trust** in digital communication.  

**Definition:**  
PKI is a system of policies, technologies, and roles that ensure:
- **Authentication** → verifying who you are.  
- **Encryption** → securing communication.  
- **Integrity** → making sure data is not changed.  
- **Trust** → knowing the other party is legitimate.  

👉 Purpose: PKI is what makes **“lock icons” 🔒 in your browser** and **secure banking apps** possible.

---

## 📌 6.2 Components of PKI

1. **Certificate Authority (CA) 🏢**  
   - A trusted third party that issues digital certificates.  
   - Example: DigiCert, Let’s Encrypt, GlobalSign.  
   - Works like a government office that issues passports.  

2. **Registration Authority (RA) 📝**  
   - Works under the CA.  
   - Validates the identity of users or organizations before a certificate is issued.  
   - Example: checking documents before a passport is approved.  

3. **Digital Certificates (X.509) 📄**  
   - Electronic documents that link a **public key** to an individual, company, or website.  
   - Example: SSL/TLS certificate for a banking website.  
   - Contains: identity info, public key, CA signature, expiration date.  

4. **Public/Private Key Pairs 🔑**  
   - Public Key → shared openly (inside the certificate).  
   - Private Key → kept secret by the owner.  
   - Used for encryption, decryption, and digital signatures.  

---

## 📌 6.3 How PKI Works in Practice

### 🔹 Step 1: Certificate Issuance
- A company (e.g., a bank) requests a certificate.  
- RA checks documents and validates identity.  
- CA issues a digital certificate signed with its private key.  

### 🔹 Step 2: Verification
- When you visit a banking site (e.g., https://mybank.com):  
  1. The site sends its certificate to your browser.  
  2. Your browser checks if the certificate is valid and signed by a trusted CA.  
  3. If valid, your browser establishes a **secure connection (HTTPS)**.  

### 🔹 Step 3: Revocation
- If a certificate is compromised (e.g., private key stolen), the CA can revoke it.  
- Revocation lists (CRLs) or Online Certificate Status Protocol (OCSP) are used to check validity.  

---

## 📌 6.4 FinTech Applications of PKI 💳

1. **SSL/TLS in Online Banking**  
   - Protects customer logins, account information, and transactions.  
   - The 🔒 padlock in browsers is powered by PKI.

2. **Digital Banking Certificates**  
   - Used for secure email between banks and customers.  
   - Prevents phishing and fake websites.

3. **API Security in FinTech Platforms**  
   - FinTech apps use PKI to secure APIs (e.g., for payment gateways, third-party integrations).  
   - Only trusted apps with valid certificates can connect.

4. **Smart Cards and Mobile Wallets**  
   - PKI secures chip-based cards, NFC payments, and mobile wallets like Apple Pay.  

---

### 🎯 Summary

- PKI = the **trust system** behind digital security.  
- Key parts: **CA, RA, Certificates, Keys**.  
- PKI ensures **authentication, encryption, and trust**.  
- In FinTech, PKI powers **online banking, mobile payments, and secure APIs**.  


# 🚨 7. Threats and Challenges in PKI

While PKI provides the foundation of trust in cybersecurity and FinTech, it is **not flawless**.  
Here are the main threats and challenges:

---

## ⚠️ 7.1 Weak Key Management
- **Definition:** Poor handling of private keys (losing, exposing, or misusing them).  
- **Problem:** If a private key is leaked, attackers can impersonate the owner.  
- **Example:** A compromised bank’s private key would allow hackers to issue fake transactions.  
- **FinTech Impact:** Customers may unknowingly send money to attackers if fake certificates are trusted.  
- **Solution:** Use **Hardware Security Modules (HSMs)** and strict key rotation policies.  

---

## 🎭 7.2 Certificate Spoofing / Fake Certificates
- **Definition:** Attackers create **fake digital certificates** that look like real ones.  
- **Problem:** Users may be tricked into trusting malicious websites.  
- **Example:** A fake banking site with a fraudulent certificate could steal customer logins.  
- **FinTech Impact:** Huge losses in **phishing campaigns** against mobile wallets and payment gateways.  
- **Solution:** Strong **certificate validation**, use of **Certificate Transparency Logs**, and **multi-factor authentication**.  

---

## 👥 7.3 Man-in-the-Middle (MITM) Attacks on PKI
- **Definition:** Attackers intercept and modify communication between a user and a server.  
- **Problem:** If attackers insert themselves before certificate validation, they can decrypt or alter traffic.  
- **Example:** Public Wi-Fi MITM where fake certificates trick users into revealing passwords.  
- **FinTech Impact:** Exposure of **transaction data, PINs, and customer details**.  
- **Solution:** Enforce **HSTS (HTTP Strict Transport Security)**, certificate pinning, and DNSSEC.  

---

## 💻 7.4 Computational Cost in Large-Scale FinTech Platforms
- **Definition:** PKI operations (encryption, decryption, signing, verification) consume **processing power**.  
- **Problem:** In large-scale systems with millions of daily transactions, this can cause **latency**.  
- **Example:** Payment gateways slowing down during peak shopping days due to certificate checks.  
- **FinTech Impact:** Customer dissatisfaction, failed transactions, financial loss.  
- **Solution:** Use **ECC over RSA** (smaller keys, faster processing), optimize certificate lifetimes, and deploy **load balancers**.  

---

### 🎯 Summary
- PKI is powerful but faces **real threats**:  
  1. Weak key management 🗝️  
  2. Fake certificates 🎭  
  3. MITM attacks 👥  
  4. High computational cost 💻  

👉 In FinTech, solving these challenges is **critical** to keep digital payments, wallets, and online banking safe.
