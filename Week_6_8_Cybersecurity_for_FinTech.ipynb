## 1. Introduction to Authentication and Access Control

---

### 🔐 1.1 What is Authentication?

- **Definition:**  
  Authentication is the process of **verifying the identity** of a user or system.  
  In simple words → **"Are you really who you claim to be?"**

- **Examples in Daily Life:**  
  - Showing your ID card to enter a university 🎓  
  - Unlocking your phone with a fingerprint or PIN 📱  
  - Logging into your Gmail with a password 📧  

- **FinTech Example:**  
  - When you open your mobile wallet app, entering your **PIN or fingerprint** is authentication.  
  - The system checks if you are the **legitimate owner** of that account.  

---

### 🛡️ 1.2 What is Access Control?

- **Definition:**  
  Access control decides **what you can do** after authentication.  
  In other words → **"Now that we know who you are, what are you allowed to access?"**

- **Types of Access Control:**  
  1. **Read-only** (you can view account balance).  
  2. **Write** (you can transfer funds).  
  3. **Admin rights** (bank manager can approve large transactions).  

- **Example in Daily Life:**  
  - You enter a concert 🎶 (authentication: ticket checked).  
  - But you can only go to the **seats you paid for** (access control).  

- **FinTech Example:**  
  - A customer can check their balance and transfer money.  
  - A bank employee might see **customer details** but not change them.  
  - An auditor may only get **read-only access** to transaction logs.  

---

### 💳 1.3 Why Authentication and Access Control Are Critical in FinTech?

1. **Protecting Money and Data 💰**  
   - FinTech apps hold sensitive data: PINs, credit card numbers, transaction logs.  
   - Without strong authentication, attackers can impersonate users.  

2. **Preventing Unauthorized Transactions 🚫**  
   - Imagine if anyone could access a payment gateway without restrictions.  
   - Attackers could move funds, alter balances, or shut down services.  

3. **Compliance with Regulations 📜**  
   - Financial institutions must follow laws (like **PCI-DSS, GDPR**) that require strict authentication and access control.  

4. **Maintaining Customer Trust 🤝**  
   - A single breach can destroy trust.  
   - Strong authentication (like 2FA) and proper access controls ensure **safe, reliable digital banking**.  

---

### 🎯 Summary of Section 1
- **Authentication** = Proving **who you are**.  
- **Access Control** = Deciding **what you can do**.  
- In FinTech, both are vital to **secure money, protect customer data, and comply with regulations**.

# 🔐 2. Password-Based Authentication

Passwords are the **oldest and most common** method of authentication. Even though newer methods exist (like biometrics and tokens), passwords still remain the **first line of defense** in many systems, including **banking and FinTech apps**.

---

## 📜 2.1 History of Passwords

- **Early Computing (1960s):**  
  - First used in MIT’s **Compatible Time-Sharing System (CTSS)**.  
  - Each user had a password to access shared computers.  

- **Evolution:**  
  - 1970s–80s → Mainframes & UNIX started hashing passwords.  
  - 1990s → Internet boom, websites began using passwords for logins.  
  - 2000s → Rise of **online banking and e-commerce**, making passwords critical.  

- **Today:**  
  - Passwords are everywhere → emails, banking, wallets, even IoT devices.  
  - Still widely used despite **known weaknesses**.  

---

## 🥊 2.2 Weak vs Strong Passwords

- **Weak Passwords ❌**  
  - Easy to guess or crack.  
  - Examples:  
    - "123456"  
    - "password"  
    - User’s birthdate 🎂  
  - Problem: Attackers can break them in seconds.  

- **Strong Passwords ✅**  
  - Long (12+ characters).  
  - Includes uppercase, lowercase, numbers, symbols.  
  - Avoids dictionary words & personal info.  
  - Example: `M!croB@nk2025$ecure`  

👉 **Rule of Thumb:** Longer = Stronger.  

---

## 🛠️ 2.3 Password Cracking Techniques

Attackers use powerful tools and techniques to break passwords:

1. **Brute Force Attack ⚡**  
   - Tries all possible combinations until it finds the correct one.  
   - Example: Trying every 4-digit PIN from `0000` to `9999`.  

2. **Dictionary Attack 📖**  
   - Uses a precompiled list of common words and passwords.  
   - Example: `sunshine`, `iloveyou`, `qwerty`.  

3. **Rainbow Tables 🌈**  
   - Precomputed hash values for many possible passwords.  
   - Instead of hashing every guess, attackers just look up the hash.  
   - Faster than brute force.  

---

## 🛡️ 2.4 Best Practices for Password Security

1. **Hashing**  
   - Store only **hashes of passwords**, not plain text.  
   - Example algorithms: **SHA-256, bcrypt, Argon2**.  

2. **Salting**  
   - Add a random string ("salt") to the password before hashing.  
   - Prevents attackers from using rainbow tables.  

3. **Password Managers**  
   - Tools like **LastPass, Bitwarden, 1Password** help create & store strong passwords.  

4. **Policies**  
   - Enforce **minimum length** (12+), complexity, and periodic changes.  

---

## 💳 2.5 FinTech Example: Securing Online Banking Logins

- **Scenario:**  
  - A customer logs into an online banking portal.  
  - The system:  
    1. Validates their **username + password**.  
    2. Applies hashing + salting to protect stored passwords.  
    3. Adds extra **layers of protection** like 2FA (OTP via SMS/app).  

- **Why Important?**  
  - Password-only systems are vulnerable to phishing, brute force, and leaks.  
  - Combining strong passwords with **multi-factor authentication** is the **gold standard** in FinTech security.  

---

### 🎯 Summary
- Passwords are old but still widely used.  
- Weak passwords = easy target. Strong passwords = much safer.  
- Attackers use brute force, dictionary, and rainbow tables to crack them.  
- FinTech apps secure logins using **hashing, salting, and 2FA**.

# 📲 3. Multi-Factor Authentication (MFA) & Two-Factor Authentication (2FA)

---

## 🔐 3.1 What is MFA and 2FA?

- **Authentication Factor** = the category of proof you use to verify identity.  

👉 **Multi-Factor Authentication (MFA)**  
- Uses **two or more** different categories of factors.  
- Example: ATM → Card (something you have) + PIN (something you know).  

👉 **Two-Factor Authentication (2FA)**  
- A specific type of MFA that uses **exactly two factors**.  
- Example: Banking app → Password + OTP sent via SMS.  

---

## 🧩 3.2 Types of Authentication Factors

1. **Knowledge Factor (Something You Know)**  
   - Examples: password, PIN, security question.  
   - Weak alone because people forget them or attackers steal them.  

2. **Possession Factor (Something You Have)**  
   - Examples: mobile phone (OTP), hardware token, smart card.  
   - Harder for attackers unless they physically steal the item.  

3. **Inherence Factor (Something You Are)**  
   - Examples: fingerprint, face scan, voice recognition.  
   - Biometric → unique to each person.  

💡 **Best Security** = combining factors from **different categories**.  

---

## 📱 3.3 SMS-Based 2FA

- **How it works:** A one-time password (OTP) is sent to the user’s phone via SMS.  
- **Pros ✅**  
  - Easy to set up.  
  - Works on any mobile phone.  
- **Cons ❌**  
  - Vulnerable to **SIM swapping** (attackers hijack your number).  
  - SMS can be intercepted or delayed.  

---

## 📲 3.4 App-Based 2FA

- **Examples:** Google Authenticator, Authy, Microsoft Authenticator.  
- **How it works:**  
  - User scans a QR code once.  
  - The app generates rotating 6-digit codes (TOTP = Time-based One-Time Password).  
- **Pros ✅**  
  - Works offline (no SMS needed).  
  - More secure than SMS (no SIM swapping).  
- **Cons ❌**  
  - If you lose your phone and no backup, you’re locked out.  

---

## 🔑 3.5 Hardware Tokens

- **Examples:** YubiKey, RSA SecurID, smart cards.  
- **How it works:** A physical device generates or stores secure codes/keys.  
- **Pros ✅**  
  - Very secure (hard to hack remotely).  
  - Used in high-security industries (government, banks).  
- **Cons ❌**  
  - Costly to deploy at large scale.  
  - Inconvenient if lost or damaged.  

---

## 💳 3.6 FinTech Example: 2FA in Mobile Wallets & Online Transactions

- **Scenario:**  
  1. A customer logs into a mobile wallet (PayPal, Google Pay, JazzCash, Easypaisa).  
  2. Step 1 → Username + Password (Knowledge factor).  
  3. Step 2 → OTP via SMS or App-based code (Possession factor).  

- **Transaction Example:**  
  - While transferring funds above a certain limit, the bank sends a **push notification or OTP** for approval.  
  - Protects against attackers who might have stolen only the password.  

- **Impact in FinTech:**  
  - 2FA prevents **account takeovers**.  
  - Increases **customer trust** in digital transactions.  
  - Meets compliance with financial regulations.  

---

### 🎯 Summary
- **MFA** = more than one factor; **2FA** = exactly two.  
- Three factor types: **Know, Have, Are**.  
- SMS 2FA is simple but risky; App-based is more secure; Hardware tokens are strongest.  
- FinTech relies heavily on 2FA for **secure wallets, mobile banking, and payment gateways**.

# 👥 4. Role-Based Access Control (RBAC)

---

## 📌 4.1 Definition and Concept of RBAC

- **Definition:**  
  Role-Based Access Control (RBAC) is a security model where **permissions are assigned to roles**, and **users are assigned to those roles**.  
  - Users don’t get permissions directly.  
  - Instead, their role determines what they can or cannot do.  

👉 In short: **Users → Roles → Permissions**.  

- **Analogy:**  
  Think of a theater 🎭:  
  - **Actors** have permission to perform on stage.  
  - **Audience** can only watch.  
  - **Staff** can manage lights or sound.  
  Their role decides their access.  

---

## 📌 4.2 Roles, Users, and Permissions

- **User:** An individual in the system (customer, employee).  
- **Role:** A job function or position with defined rights (teller, manager, auditor).  
- **Permission:** Actions that a role can perform (read, write, approve).  

👉 Example Mapping:  
- **User** = Sarah (Bank Employee)  
- **Role** = Teller  
- **Permissions** = View account balances, process deposits/withdrawals  

---

## 📌 4.3 Example in Banking

1. **Bank Teller 🏦**  
   - Role: Teller  
   - Permissions: View balances, process deposits/withdrawals, but **cannot approve large loans**.  

2. **Branch Manager 👔**  
   - Role: Manager  
   - Permissions: Approve loans, override teller transactions, access customer records.  

3. **System Administrator 💻**  
   - Role: Admin  
   - Permissions: Manage users, configure systems, but **should not conduct transactions**.  

👉 RBAC ensures **separation of duties**: each role has just enough access to perform its job.  

---

## 📌 4.4 Benefits and Challenges of RBAC

### ✅ Benefits
1. **Principle of Least Privilege** → Users only get the access they need.  
2. **Simplified Management** → Assigning roles is easier than configuring each user separately.  
3. **Improved Security** → Reduces risk of insider threats.  
4. **Compliance** → Helps meet financial regulations (GDPR, PCI-DSS).  

### ❌ Challenges
1. **Role Explosion** → Too many overlapping roles can become unmanageable.  
2. **Maintenance Overhead** → Needs frequent updates as employees join/leave or change positions.  
3. **Rigid Structure** → Sometimes too strict for dynamic environments like startups.  

---

## 📌 4.5 Real-World Case: Access Segregation in Payment Gateways

- **Scenario:**  
  - A payment gateway (like PayPal, Stripe) handles **millions of transactions daily**.  
  - Employees are divided into roles:  

  1. **Customer Support Staff 👩‍💻** → Can view transaction status but **cannot refund**.  
  2. **Finance Officers 💼** → Can approve or reject refunds above a limit.  
  3. **Developers 👨‍💻** → Can fix system bugs but **cannot access customer financial data**.  
  4. **Auditors 📊** → Have read-only access to logs for compliance checks.  

👉 By using RBAC, the company ensures:  
- **No single employee has unlimited control.**  
- **Sensitive actions require higher-level roles.**  
- **Audit trails** can track who accessed what and when.  

---

### 🎯 Summary
- RBAC assigns **permissions to roles, not users directly**.  
- Banking example: Teller vs Manager vs Admin.  
- Benefits: stronger security, easier management, compliance.  
- Challenges: role explosion, complexity in maintenance.  
- In FinTech, RBAC protects sensitive operations in **payment gateways and digital banking systems**.

# 🔗 5. Modern Authentication Protocols (OAuth & Beyond)

---

## 📌 5.1 What is OAuth? (Delegated Authorization)

- **Definition:**  
  OAuth (Open Authorization) is a protocol that allows a user to **grant a third-party app limited access** to their resources **without sharing their password**.  

👉 Example:  
- You want to use a budgeting app 📊 that needs access to your bank account transactions.  
- With OAuth, you log in to your bank directly, approve access, and the app only gets a **token** (not your username/password).  

**Analogy:**  
- Imagine leaving your car at a valet 🚗.  
- You give them a **valet key** (limited rights – they can drive the car, but not open the trunk or glovebox).  

---

## 📌 5.2 OAuth 2.0 Workflow

OAuth 2.0 has four main players:

1. **Resource Owner (User)** → The person who owns the data (e.g., customer).  
2. **Client (App)** → The application requesting access (e.g., FinTech budgeting app).  
3. **Authorization Server** → Issues tokens after user approval (e.g., Google, Bank’s auth server).  
4. **Resource Server** → Holds the protected data (e.g., banking API).  

**Step-by-Step Flow:**  
1. Client app asks user for permission.  
2. User authenticates with Authorization Server.  
3. Authorization Server issues an **Access Token** to the Client.  
4. Client uses token to access data from Resource Server.  

👉 The user’s **password is never shared** with the client app.  

---

## 📌 5.3 Example: Logging into a FinTech App with Google or Facebook

- You open a FinTech app → It shows “Login with Google / Facebook”.  
- Instead of creating a new password:  
  1. You log into Google.  
  2. Google asks: “Do you allow this app to access your email profile?”  
  3. You click **Allow**.  
  4. The FinTech app gets an **OAuth token** → which it uses to fetch your Google profile (not your password).  

---

## 📌 5.4 OpenID Connect (OIDC)

- **Problem with OAuth:** OAuth alone only handles **authorization** (what you can do), not **authentication** (who you are).  
- **Solution:** OpenID Connect (OIDC) is built on top of OAuth 2.0 to add **identity verification**.  

👉 Example:  
- With OAuth, the app knows your wallet app has **permission to fetch balances**.  
- With OpenID Connect, the app also knows **you are actually Usama Arshad (identity confirmed)**.  

**How it works:**  
- Adds an **ID Token** (JSON Web Token – JWT) alongside the access token.  
- Used heavily in **“Login with Google”** flows.  

---

## 📌 5.5 Security Concerns in OAuth

1. **Phishing Attacks 🎭**  
   - Fake authorization pages trick users into entering credentials.  

2. **Token Leakage 🔓**  
   - If access tokens are stolen, attackers can access resources.  

3. **Improper Scope Management 🚪**  
   - Apps might request more permissions than necessary.  

4. **Insecure Redirect URIs ↩️**  
   - Attackers can hijack the flow if the redirect URL is not strictly validated.  

---

## 📌 5.6 API Access Control in FinTech Using OAuth

- FinTech relies on **APIs** for payments, wallets, and transactions.  
- OAuth helps secure these APIs by issuing **tokens instead of passwords**.  

👉 Examples:  
1. **Payment Gateway API (Stripe, PayPal, Easypaisa)**  
   - Apps get OAuth tokens to process payments securely.  

2. **Banking APIs (Open Banking / PSD2 in EU)**  
   - Customers can share account data with third-party apps via OAuth.  

3. **Mobile Wallets**  
   - OAuth tokens allow apps to verify identity and enable transactions **without exposing credentials**.  

---

### 🎯 Summary

- **OAuth** = Delegated Authorization (limited access without sharing passwords).  
- **OAuth 2.0** uses tokens to access protected resources.  
- **OpenID Connect** adds identity (authentication).  
- **FinTech** uses OAuth for secure API integrations, mobile logins, and payment systems.  
- **Risks** → token leakage, phishing, misconfigured redirects.

````python
# fintech_dashboard_persistent_commented.py
# 🏦 Persistent FinTech Dashboard with RBAC + 2FA
# Author: Dr. Usama's teaching demo

# --------------------------
# 📦 Import required libraries
# --------------------------
import streamlit as st          # Streamlit is used to build the web app
import pyotp                   # pyotp helps us generate and verify one-time passwords (OTP)
import qrcode                  # qrcode generates QR codes for Google Authenticator setup
import io                      # io is used to handle in-memory image buffers
import json                    # json is used to save/load user data persistently
import os                      # os is used to check if data files exist
from PIL import Image          # PIL (Pillow) is used to display and resize QR codes
import random                  # random helps simulate large transfer requests for admin demo

# --------------------------
# 📁 Persistent Storage Setup
# --------------------------
DATA_FILE = "users.json"  # File where user accounts and balances are stored

# Function to save users back to file
def save_users(users):
    with open(DATA_FILE, "w") as f:          # Open file in write mode
        json.dump(users, f, indent=4)        # Dump users as pretty-printed JSON

# Function to load users from file, or create defaults if file is missing/corrupted
def load_users():
    if not os.path.exists(DATA_FILE) or os.path.getsize(DATA_FILE) == 0:  # If no file or file empty
        # Default users created the very first time
        users = {
            "customer1": {"password": "secure123", "role": "Customer", "balance": 1500, "secret": pyotp.random_base32()},
            "customer2": {"password": "wallet321", "role": "Customer", "balance": 3200, "secret": pyotp.random_base32()},
            "admin1": {"password": "adminpass", "role": "Admin", "secret": pyotp.random_base32()},
        }
        save_users(users)  # Save to JSON file
        return users
    else:
        try:
            with open(DATA_FILE, "r") as f:  # Open file for reading
                users = json.load(f)         # Load JSON content into dictionary
            # Ensure each user has a secret for OTP
            changed = False
            for u in users:
                if "secret" not in users[u]:
                    users[u]["secret"] = pyotp.random_base32()
                    changed = True
            if changed:                      # If we added new secrets, save back
                save_users(users)
            return users
        except json.JSONDecodeError:         # Handle corrupted JSON file
            users = {
                "customer1": {"password": "secure123", "role": "Customer", "balance": 1500, "secret": pyotp.random_base32()},
                "customer2": {"password": "wallet321", "role": "Customer", "balance": 3200, "secret": pyotp.random_base32()},
                "admin1": {"password": "adminpass", "role": "Admin", "secret": pyotp.random_base32()},
            }
            save_users(users)
            return users

# Load all users into memory
USERS = load_users()

# --------------------------
# 🎨 Streamlit Page Configuration
# --------------------------
st.set_page_config(page_title="Persistent FinTech Dashboard", page_icon="💳", layout="wide")

# --------------------------
# 🏦 App Header
# --------------------------
st.title("💳 Persistent FinTech Dashboard with RBAC + 2FA")   # App title
st.markdown("""                                              # Short description
This is a **mini FinTech system** with:
- ✅ 2FA login (Google Authenticator, per user)
- ✅ RBAC (Role-Based Access Control)
- ✅ Persistent storage (balances & secrets saved in `users.json`)
""")

# --------------------------
# 🔑 Login Form
# --------------------------
st.subheader("🔑 Step 1: Login")  # Login section header

# Input fields for login
username = st.text_input("Username")                 # Username field
password = st.text_input("Password", type="password")# Password field (hidden)
otp_input = st.text_input("Enter OTP (6 digits)", max_chars=6)  # OTP input

# Initialize session state for login tracking
if "logged_in" not in st.session_state:
    st.session_state.logged_in = False
    st.session_state.user = None

# If user typed a username that exists, show QR code for their OTP secret
if username and username in USERS:
    user_secret = USERS[username]["secret"]   # Fetch secret for this user
    # Generate provisioning URI for Google Authenticator
    uri = pyotp.TOTP(user_secret).provisioning_uri(
        name=f"{username}@fintechdemo.com", issuer_name="FinTech Demo App"
    )
    qr = qrcode.make(uri)                     # Generate QR code
    buf = io.BytesIO()                        # Create in-memory buffer
    qr.save(buf, format="PNG")                # Save QR image to buffer
    buf.seek(0)                               # Reset buffer pointer
    qr_img = Image.open(buf).resize((250, 250))  # Open and resize QR image
    st.image(qr_img, caption=f"📱 Scan this QR in Google Authenticator for {username}", use_container_width=False)

# Login button logic
if st.button("Login"):
    if username in USERS and password == USERS[username]["password"]:  # Check username/password
        role = USERS[username]["role"]                                # Fetch role
        user_totp = pyotp.TOTP(USERS[username]["secret"])             # Get user-specific TOTP generator

        if user_totp.verify(otp_input, valid_window=1):  # Verify OTP, allow ±30 sec drift
            st.session_state.logged_in = True
            st.session_state.user = username
            st.session_state.role = role
            st.success(f"🎉 Login successful! Welcome {username} ({role}).")  # Success message
            st.balloons()  # Show balloons animation
        else:
            st.error("❌ OTP invalid or expired.")  # Wrong OTP
    else:
        st.error("❌ Invalid username or password.") # Wrong username/password

# --------------------------
# 👤 Customer Dashboard
# --------------------------
if st.session_state.logged_in and st.session_state.role == "Customer":
    user = st.session_state.user
    st.header("👤 Customer Dashboard")             # Section header
    st.write(f"Welcome **{user}**")                # Show username

    balance = USERS[user]["balance"]               # Get balance
    st.metric("💰 Account Balance", f"${balance}") # Show balance metric

    st.subheader("💸 Make a Transfer")             # Transfer section
    # Select recipient (only other customers)
    recipient = st.selectbox("Select recipient", [u for u in USERS if USERS[u]["role"] == "Customer" and u != user])
    amount = st.number_input("Enter amount", min_value=1, max_value=10000, step=1)  # Transfer amount

    # Transfer button
    if st.button("Transfer Money"):
        if amount <= USERS[user]["balance"]:       # Check balance
            USERS[user]["balance"] -= amount       # Deduct from sender
            USERS[recipient]["balance"] += amount  # Add to recipient
            save_users(USERS)                      # Save updated balances
            st.success(f"✅ Transfer of ${amount} to {recipient} successful!") # Success message
        else:
            st.error("❌ Insufficient funds.")      # Error if not enough balance

# --------------------------
# 👨‍💼 Admin Dashboard
# --------------------------
elif st.session_state.logged_in and st.session_state.role == "Admin":
    st.header("👨‍💼 Admin Dashboard")               # Section header
    st.write(f"Welcome **{st.session_state.user}**")

    st.subheader("📋 View All Customer Accounts")  # Show all balances
    for u, data in USERS.items():
        if data["role"] == "Customer":
            st.write(f"**{u}** → Balance: ${data['balance']}")

    st.subheader("✅ Approve Large Transfers")      # Simulated approval section
    fake_request = {"from": "customer1", "to": "customer2", "amount": random.randint(5000, 10000)}
    st.write(f"Pending Request: {fake_request['from']} → {fake_request['to']} | Amount: ${fake_request['amount']}")

    if st.button("Approve Transfer"):
        if USERS[fake_request["from"]]["balance"] >= fake_request["amount"]:
            USERS[fake_request["from"]]["balance"] -= fake_request["amount"]
            USERS[fake_request["to"]]["balance"] += fake_request["amount"]
            save_users(USERS)
            st.success("✅ Transfer approved and processed!")
        else:
            st.error("❌ Insufficient funds in sender account.")

    if st.button("Reject Transfer"):
        st.warning("❌ Transfer request rejected.")

# --------------------------
# 📚 Teaching Notes
# --------------------------
with st.expander("ℹ️ Teaching Notes"):   # Collapsible teaching notes
    st.markdown("""
    - **Persistence**: All account data (including OTP secrets) is saved in `users.json`.
    - **2FA**: Each user has a unique Google Authenticator secret.
    - **RBAC**: Customers can transfer/view balances; Admins can manage system.
    - **Security Principle**: Authentication (who you are) + Authorization (what you can do).
    """)
````

# 🔑 Week 6 Lab — Authentication & Access Control

**Topic:** Brute-Force SSH Attack with Hydra 🐍

---

## 🛠️ Lab Environment

* **Attacker Machine**: Kali Linux VM (comes with Hydra pre-installed).
* **Target Machine**: Ubuntu VM (or localhost/Docker) running SSH server with weak credentials.
* Network: Both machines on the same NAT/Host-only adapter.

---

## 🔹 Step 1: Prepare Target (Victim Machine)

### 1. Install SSH server

```bash
sudo apt update && sudo apt install openssh-server -y
```

### 2. Start & enable service

```bash
sudo systemctl enable ssh
sudo systemctl start ssh
```

### 3. Create a weak user (easy to crack)

```bash
sudo adduser labuser
# Set password: 123456
```

### 4. Verify SSH login works

```bash
ssh labuser@localhost
```

If login works → target ready.

💡 **Tip:** In multi-VM setup, find target IP with:

```bash
ip addr
```

Example: `192.168.56.101`.

---

## 🔹 Step 2: Prepare Attacker (Kali VM)

### 1. Verify Hydra is installed

```bash
hydra -h
```

If not present:

```bash
sudo apt update && sudo apt install hydra -y
```

### 2. Locate password wordlist

Kali ships with `rockyou.txt`:

```bash
ls /usr/share/wordlists/rockyou.txt
```

If compressed, unzip it:

```bash
sudo gunzip /usr/share/wordlists/rockyou.txt.gz
```

---

## 🔹 Step 3: Run Hydra Attack

### Command (single user brute-force)

```bash
hydra -l labuser -P /usr/share/wordlists/rockyou.txt ssh://<target_IP>
```

* `-l labuser` → username.
* `-P rockyou.txt` → password dictionary.
* `ssh://<target_IP>` → target (e.g., `ssh://192.168.56.101`).

---

## 🔹 Step 4: Observe Results

* If password is found:

  ```
  [22][ssh] host: 192.168.56.101   login: labuser   password: 123456
  ```
* If not: wordlist didn’t contain it.

---

## 🔹 Step 5: Discussion & Mitigation

1. **Account Lockouts** → limit retries (e.g., 5 failed logins).
2. **Strong Passwords** → long, random strings resist brute force.
3. **Two-Factor Authentication (2FA)** → blocks attacker even with cracked password.

---

## ✅ Learning Outcomes

* Saw Hydra crack a weak SSH password.
* Understood why weak credentials are risky.
* Connected brute-force attacks to **MFA, lockouts, and password hygiene**.

## 1. Introduction to Network & Web Security 🔍

### 1.1 Why Network Security Matters in FinTech

* **Financial systems are high-value targets** 💰 → Hackers aim for money, customer data, and transaction manipulation.
* **FinTech = Internet-first** → Mobile wallets, online banking, stock trading apps all depend on secure networks.
* **A single weakness can lead to massive losses** (example: Bangladesh Bank SWIFT attack).
* **Trust = currency in FinTech** → If users don’t feel safe, adoption drops.

👉 In FinTech, **network security = financial security**.

---

### 1.2 Common Network Threats ⚠️

* **Eavesdropping (Sniffing):**
  Hackers secretly listen to network traffic (like wiretapping a phone).

  * Example: Capturing unencrypted HTTP login credentials.

* **Spoofing:**
  Attacker pretends to be someone else.

  * Example: Fake banking site using DNS spoofing to steal PINs.

* **Man-in-the-Middle (MITM):**
  Attacker intercepts communication between two parties.

  * Example: Mobile wallet traffic hijacked on public Wi-Fi.

👉 These attacks can **break confidentiality, integrity, or availability** of financial systems.

---

### 1.3 Importance of Secure Web Communication 🌐🔒

* Online banking, payment gateways, and APIs constantly send **sensitive data** (PINs, account balances, transaction approvals).
* Without **encryption + authentication**, attackers can:

  * Steal credentials
  * Alter transactions
  * Impersonate customers
* Example: Sending `Transfer $1000` over HTTP could be modified to `Transfer $10,000`.

👉 **Secure channels (HTTPS, VPN, DNSSEC, Firewalls)** = backbone of FinTech trust.

## 2. Firewalls 🔥

### 2.1 What is a Firewall?

* A **firewall = digital security guard** 🚨 that filters incoming and outgoing traffic based on rules.
* Acts as a **barrier** between trusted internal networks (banking systems) and untrusted external networks (Internet).
* **Role in perimeter security:**

  * Block malicious traffic
  * Allow legitimate business traffic
  * Log suspicious behavior for auditing

👉 Think of it as a **bouncer at a nightclub** → only lets authorized guests in, kicks troublemakers out.

---

### 2.2 Types of Firewalls

1. **Packet-Filtering Firewalls** 📦

   * Look at basic info (IP, port, protocol).
   * Example: Block all traffic on port 23 (Telnet).
   * Limitation: Cannot inspect payload content.

2. **Stateful Inspection Firewalls** 🔎

   * Track active sessions.
   * Ensure traffic is part of a valid ongoing conversation.
   * Example: If you request a banking page, only responses to *your* request are allowed back.

3. **Application-Layer Firewalls (Proxy Firewalls)** 🌐

   * Work at the application level (e.g., HTTP, SMTP).
   * Can filter malicious requests like SQL injection attempts.
   * Example: Block suspicious API calls in a mobile wallet.

4. **Next-Generation Firewalls (NGFW)** 🚀

   * Combine packet filtering, stateful inspection, intrusion prevention, deep packet inspection.
   * Smart enough to detect malware and DDoS attempts.
   * Widely used in **modern banks & FinTech cloud environments**.

---

### 2.3 Placement of Firewalls in FinTech Architecture 🏦

* **Perimeter firewall:** Between Internet and bank data center → blocks general threats.
* **Internal segmentation firewall:** Separates critical subsystems (payment processor, customer DB).
* **Cloud firewall (virtual):** Protects APIs and cloud-based FinTech apps.
* **API gateway firewall:** Ensures only valid requests reach payment APIs.

👉 Example: A **payment gateway firewall** may block suspicious IP ranges known for fraud.

---

### 2.4 Strengths and Limitations of Firewalls ⚖️

✅ **Strengths:**

* First line of defense against network intrusions
* Enforce security policies
* Stop known attacks like port scans, unauthorized connections

❌ **Limitations:**

* Cannot protect against **insider threats** (e.g., employee leaks data)
* Cannot detect encrypted malicious traffic if not combined with deep inspection
* Often need integration with **IDS/IPS** and monitoring systems

---

### 2.5 Case Study: Firewalls in Banking APIs 🛡️

* **Scenario:** A FinTech app provides a payment API for merchants.
* Threat: Hackers try to flood the API with requests to crash it.
* **Firewall Role:**

  * Blocks unusual traffic patterns
  * Restricts access to trusted IP ranges
  * Logs suspicious activities for audits

👉 Result: Customers can keep paying securely without disruptions.

## 3. Virtual Private Networks (VPNs) 🔑

### 3.1 What is a VPN?

* A **VPN (Virtual Private Network)** creates a **secure tunnel** 🛤️ between your device and a remote server.
* **Tunneling:** Data packets are wrapped inside encrypted packets (like putting letters inside a locked safe before mailing).
* **Encryption:** Makes data unreadable to outsiders.
* **Authentication:** Ensures only trusted devices/people can connect.

👉 Analogy: Imagine driving through a **private tunnel** instead of open roads — no one can watch or interfere.

---

### 3.2 How VPNs Secure Data in Transit 🔒

* Protects against **eavesdropping** on public Wi-Fi.
* Prevents **man-in-the-middle (MITM)** attacks by encrypting traffic.
* Masks IP address → harder for attackers to trace users.

👉 Example: A FinTech trader logs into their account from a café. Without VPN = attackers sniff traffic; with VPN = data is unreadable.

---

### 3.3 VPN Protocols 🛠️

* **IPSec (Internet Protocol Security):** Strong encryption, widely used in enterprises.
* **L2TP (Layer 2 Tunneling Protocol):** Often combined with IPSec, adds extra security.
* **OpenVPN:** Open-source, very secure, flexible.
* **WireGuard:** Modern, faster, lightweight alternative.

👉 In banks, IPSec and OpenVPN are most common.

---

### 3.4 Corporate VPNs in FinTech 🏦

* Connect **remote employees** securely to bank servers.
* Secure communication between **branches and HQ**.
* Protect API calls from external vendors.

👉 Example: A FinTech startup uses a VPN for its developers to access customer databases securely.

---

### 3.5 Pros and Cons ⚖️

✅ **Pros:**

* Encrypts all traffic
* Masks location/IP
* Protects remote workers

❌ **Cons:**

* Can slow down connections
* Not foolproof (malware still possible)
* Needs proper setup

---

### 3.6 Example: VPNs in Trading 📈

A stock trader connects to **NYSE trading servers** via VPN.

* Without VPN: Risk of MITM altering buy/sell orders.
* With VPN: Orders are encrypted, guaranteeing accuracy and confidentiality.

## 4. DNS Security Extensions (DNSSEC) 🌍

### 4.1 Role of DNS

* **DNS = Phonebook of the Internet** 📖
* Converts website names (like `bank.com`) into IP addresses.
* Without DNS, you’d need to remember numbers like `192.168.20.33`.

---

### 4.2 Threats 🚨

* **DNS Spoofing:** Attacker sends fake DNS response → user goes to malicious site.
* **DNS Cache Poisoning:** Corrupts DNS server cache so everyone using it is redirected to fake sites.

👉 Example: Instead of `mybank.com`, user unknowingly lands on `fakebank.com` that steals PINs.

---

### 4.3 What is DNSSEC? 🔑

* **DNSSEC = DNS + Digital Signatures** ✍️
* Uses **cryptographic keys** to verify that DNS responses are **authentic**.
* Ensures that when DNS says "`bank.com = 192.168.20.33`," it truly came from the bank’s DNS server.

👉 Think of it like a **tamper-proof seal** on a medicine bottle — if the seal is broken, don’t trust it.

---

### 4.4 How DNSSEC Prevents Spoofing 🛡️

* Adds **public/private key cryptography** to DNS records.
* DNS resolvers check signatures before accepting responses.
* Attackers cannot easily forge DNS answers without the private key.

---

### 4.5 Limitations of DNSSEC ⚠️

* Not widely deployed (many ISPs don’t support it).
* Increases complexity of DNS management.
* Slower responses due to extra cryptographic checks.

---

### 4.6 FinTech Example 💳

* A user types `securewallet.com`.
* With DNSSEC → Browser gets digitally signed DNS response, ensuring it’s the **real wallet server**.
* Without DNSSEC → User could land on a **phishing clone** that looks identical but steals their credentials.

👉 DNSSEC is crucial in preventing **fake banking sites** from stealing data.

## 5. HTTPS and Secure Web Communication 🔒

### 5.1 Evolution: HTTP vs HTTPS

* **HTTP (Hypertext Transfer Protocol)** 📜

  * Old way of transferring data between browser and server.
  * ❌ Data sent in plain text → attackers can read or modify it.

* **HTTPS (HTTP Secure)** 🔒

  * Adds **encryption (TLS/SSL)** on top of HTTP.
  * ✅ Confidentiality → data is private
  * ✅ Integrity → data cannot be tampered with
  * ✅ Authentication → ensures server is genuine

👉 Example: Logging into online banking over **HTTP** = risky ⚠️. Over **HTTPS** = safe ✅.

---

### 5.2 Role of TLS/SSL in HTTPS

* **TLS (Transport Layer Security)** = modern encryption protocol.
* **SSL (Secure Sockets Layer)** = older version (still used in name, but outdated).
* Provides 🔑:

  * Encryption (protects data in transit)
  * Authentication (validates server identity)
  * Integrity (detects tampering)

---

### 5.3 How HTTPS Works (Handshake 🤝)

1. **Client Hello:** Browser says hello, suggests encryption options.
2. **Server Hello:** Server replies with its digital certificate.
3. **Certificate Verification:** Browser checks certificate authenticity (CA).
4. **Key Exchange:** Secure session key is established (RSA, ECC, Diffie-Hellman).
5. **Encrypted Communication:** All further traffic is encrypted.

👉 Analogy: Like **agreeing on a secret code language** before starting a private conversation.

---

### 5.4 Certificate Authorities (CA) and PKI in HTTPS

* **CA (Certificate Authority):** Trusted orgs (like DigiCert, Let’s Encrypt) that issue digital certificates.
* **PKI (Public Key Infrastructure):** Framework of keys, certificates, and trust.
* **Browser trust store:** Only certificates from trusted CAs are accepted.

👉 Example: `securebank.com` certificate signed by DigiCert = ✅ trusted.

---

### 5.5 Browser Security Indicators 🖥️

* **Padlock Icon 🔒** = connection is secure.
* **HSTS (HTTP Strict Transport Security):** Forces HTTPS, blocks downgrade to HTTP.
* **Certificate Pinning:** App only trusts specific certificates → prevents fake CA attacks.

---

### 5.6 Case Study: API Calls in FinTech Apps 📲

* A mobile wallet calls backend API: `https://api.wallet.com/transaction`.
* HTTPS ensures:

  * 🔐 No attacker can sniff transaction details.
  * ✅ User knows the server is authentic.
  * 🔏 Prevents man-in-the-middle tampering.

---

### 5.7 Common Attacks on HTTPS ⚠️

* **SSL Stripping:** Downgrading from HTTPS → HTTP.
* **Expired Certificates:** Users may ignore warnings, exposing them to MITM.
* **Misconfigured Certificates:** Wrong hostname, self-signed certs.
* **Weak Encryption:** Old SSL/TLS versions still in use.

👉 Real-world: Equifax breach (2017) partly due to failure to patch SSL vulnerabilities.

# 📅 Week 8: Secure Communication Protocols 🔐📡

---

## 1. Introduction to Secure Communication Protocols 🌍

### 1.1 Why Secure Communication is Critical in FinTech

* FinTech systems deal with **highly sensitive data**:

  * Customer information 🧑‍💻
  * PINs, credit card numbers 💳
  * Trading orders 📈
* A **single breach** can mean:

  * Financial loss 💸
  * Regulatory penalties ⚖️
  * Loss of customer trust 🚫

👉 Without secure protocols, **banking and wallets cannot function safely online**.

---

### 1.2 Risks of Insecure Communication ⚠️

* **Man-in-the-Middle (MITM) Attack** 👤🔀👤

  * Hacker intercepts and modifies traffic.
  * Example: User sends “Transfer $100” → Attacker changes it to “$1000”.

* **Replay Attacks** 🔁

  * Attacker records valid data (like login tokens) and reuses it later.
  * Example: Resubmitting a “transfer request” packet again.

* **Packet Sniffing (Eavesdropping)** 👂📡

  * Attacker captures unencrypted traffic over Wi-Fi or networks.
  * Example: Reading banking usernames/passwords sent via plain HTTP.

---

### 1.3 Overview of Key Protocols 🛠️

* **TLS/SSL** → Encrypts web traffic (used in HTTPS).
* **SSH (Secure Shell)** → Secure remote login & file transfers.
* **IPSec** → Protects IP traffic (used in VPNs).
* **PGP / S/MIME** → Secure email communication.

👉 Each protocol solves a **different part of secure communication** in FinTech.

---

## 2. TLS/SSL (Transport Layer Security / Secure Sockets Layer) 🔒

### 2.1 History (SSL → TLS Evolution)

* **SSL (Secure Sockets Layer)**:

  * Developed by Netscape in the 1990s.
  * Versions SSL 2.0 & SSL 3.0 → insecure today.

* **TLS (Transport Layer Security)**:

  * Replacement for SSL (from 1999 onward).
  * Modern versions (1.2 & 1.3) are **industry standards**.

👉 Today when we say **SSL**, we almost always mean **TLS**.

---

### 2.2 TLS Handshake Process 🤝 (Step-by-Step)

When you visit `https://mybank.com`:

1. **Client Hello** 👋

   * Browser sends:

     * Supported TLS versions
     * Encryption algorithms (cipher suites)
     * Random number (used for key generation).

2. **Server Hello** 🖥️

   * Bank’s server replies with:

     * Chosen TLS version
     * Chosen cipher suite
     * Its **digital certificate** (signed by CA).

3. **Certificate Verification** 📜

   * Browser checks if certificate is:

     * Signed by a trusted CA
     * Matches domain name
     * Not expired/revoked.

4. **Key Exchange** 🔑

   * Browser & server agree on a **shared session key** using RSA/ECC/Diffie-Hellman.

5. **Secure Session Established** 🔐

   * All further communication is encrypted with the shared session key.

👉 Analogy: Like two people agreeing on a **secret code language** before talking privately.

---

### 2.3 Symmetric vs Asymmetric Encryption Inside TLS

* **Asymmetric (Public/Private keys)** → Used during handshake for key exchange.

  * Slow but secure.
  * Example: RSA/ECC.

* **Symmetric (Shared key)** → Used for actual communication.

  * Fast and efficient.
  * Example: AES.

👉 TLS **combines both**: asymmetric for setup, symmetric for speed.

---

### 2.4 Certificates, PKI, and Trust Chains 🏛️

* **Digital Certificate** = Server’s ID card ✅.
* **Issued by Certificate Authority (CA)** like DigiCert, Let’s Encrypt.
* **PKI (Public Key Infrastructure)** = System of CAs, keys, and certificates.
* **Trust Chain:**

  * Root CA → Intermediate CA → Server Certificate.

👉 Without valid certificates, a browser shows **“Not Secure” warning** 🚫.

---

### 2.5 Versions of TLS (1.0 → 1.3) 📜

* **TLS 1.0 / 1.1** → Deprecated (weak).
* **TLS 1.2** → Still widely used, secure.
* **TLS 1.3** → Latest, faster, stronger:

  * Zero-RTT handshake (fewer steps).
  * Stronger cipher suites (no outdated algorithms).

👉 Banks and payment providers are migrating to **TLS 1.3** for speed & security.

---

### 2.6 FinTech Example: Online Banking Transactions 🏦

* User logs in at `https://securebank.com`.
* TLS ensures:

  * Username & password cannot be sniffed.
  * Transactions (like “Pay $100”) are encrypted.
  * Certificate proves the server is genuine.

👉 Without TLS → attacker could:

* Steal credentials
* Modify transaction amounts
* Impersonate the bank

## 3. Email Security Protocols 📧

### 3.1 Why Email Needs Protection in Financial Services

* Email is the **backbone of business communication** 📩.
* Financial institutions use email for:

  * Sending **monthly statements** 📑
  * Confirming **transactions or loan approvals** ✅
  * Sharing **compliance/legal notices** ⚖️
* **Risks without protection:**

  * Email spoofing (fake bank emails)
  * Phishing scams 🎣
  * Leakage of confidential documents

👉 Secure email ensures **authenticity + confidentiality + integrity**.

---

### 3.2 PGP (Pretty Good Privacy) 🔐

* **Definition:** A data encryption and decryption program for securing emails.
* **How it works:**

  * Uses **public-key cryptography** (RSA/ECC) + **symmetric encryption (AES)**.
  * Sender encrypts with recipient’s **public key** → recipient decrypts with **private key**.
* **Features:**

  * Digital signatures for authenticity ✍️
  * Encryption for confidentiality 🔒
* **Usage:** Widely used in open-source, developers, and privacy communities.

👉 Example: A FinTech startup founder sends sensitive investor details via **PGP-encrypted email**.

---

### 3.3 S/MIME (Secure/Multipurpose Internet Mail Extensions) 📜

* **Definition:** An email security standard built into enterprise systems.
* **How it works:**

  * Uses **X.509 digital certificates** from trusted Certificate Authorities.
  * Provides **end-to-end encryption** and **digital signatures**.
* **Features:**

  * Seamless integration with Outlook, Gmail, Apple Mail.
  * Easier for enterprise adoption compared to PGP.

👉 Example: A bank sends **digitally signed loan agreements** using S/MIME → customer can verify it’s genuine.

---

### 3.4 Differences: PGP vs S/MIME ⚖️

| Feature        | **PGP**                                   | **S/MIME**                                |
| -------------- | ----------------------------------------- | ----------------------------------------- |
| Key Management | Web of Trust (users verify each other) 🌐 | Centralized (Certificate Authorities) 🏛️ |
| Adoption       | Common among individuals & open-source    | Common in enterprises & banks             |
| Setup          | Manual (install keys, software)           | Automatic (built-in with corporate email) |
| Trust Model    | Decentralized                             | Centralized                               |

---

### 3.5 FinTech Example 💳

* **Use Case:** Protecting financial statements & compliance emails.
* Customer receives an **encrypted PDF bank statement** via email.
* S/MIME ensures only the customer can open it + verifies bank’s signature.
* Protects against **phishing or fake statement scams**.

## 4. IPSec (Internet Protocol Security) 🌐

### 4.1 What is IPSec and Why It Matters

* **Definition:** A protocol suite for securing IP communications by authenticating and encrypting each IP packet.
* **Why it matters in FinTech:**

  * Secures communication between **branches, ATMs, data centers**.
  * Protects financial transactions that travel over the Internet.

👉 Without IPSec, ATM-to-bank traffic could be sniffed or altered.

---

### 4.2 Modes of IPSec 🛠️

1. **Transport Mode:**

   * Encrypts only the **payload** of the IP packet.
   * Source & destination IP headers remain visible.
   * Used for **end-to-end communication** (e.g., client ↔ server).

2. **Tunnel Mode:**

   * Encrypts the **entire packet** (payload + header).
   * New IP header is added.
   * Used for **network-to-network communication** (e.g., branch VPNs).

---

### 4.3 Key Protocols in IPSec 🔑

* **AH (Authentication Header):**

  * Provides authentication and integrity.
  * Ensures packet hasn’t been tampered with.
  * ❌ Does not encrypt payload.

* **ESP (Encapsulating Security Payload):**

  * Provides **encryption + authentication**.
  * Ensures confidentiality and integrity.
  * Widely used in modern VPNs.

👉 Usually, ESP = preferred in banking.

---

### 4.4 Use in VPNs 🔑

* IPSec is the **foundation of corporate VPNs**.
* Used for:

  * Secure branch-to-headquarter links 🏦
  * Remote employee connections 💻
  * Inter-bank secure communications 🔗

---

### 4.5 FinTech Example: ATM Security 🏧

* ATM communicates with central banking server for withdrawals.
* IPSec tunnel ensures:

  * Transaction request (e.g., withdraw $500) cannot be sniffed or altered.
  * Authentication confirms the ATM is genuine, not a fake machine.
* Result: **ATM-to-server communication is secure**.

## 5. SSH (Secure Shell) 🔑

### 5.1 What is SSH and Its Role in Secure Remote Access

* **Definition:** SSH is a cryptographic protocol that provides **secure remote login** to servers and devices over insecure networks.
* **Why it matters:**

  * Without SSH → admins used Telnet (plaintext → passwords exposed).
  * With SSH → everything (commands, passwords, data) is **encrypted**.
* **Role in FinTech:**

  * Bank IT teams manage **cloud servers, APIs, databases** via SSH.
  * Ensures attackers can’t sniff or hijack sensitive commands.

👉 Analogy: Think of SSH as a **locked, private corridor** where admins can talk safely to servers.

---

### 5.2 Public/Private Key Authentication 🔑

* SSH uses **asymmetric cryptography** for secure login.
* **How it works:**

  1. Admin generates a **key pair** (public + private).
  2. Public key is stored on the server.
  3. Admin logs in → server challenges them.
  4. Private key proves identity without sending a password.
* ✅ More secure than passwords.
* ✅ Prevents brute force attacks.

---

### 5.3 Secure File Transfer (SCP, SFTP) 📂

* **SCP (Secure Copy):**

  * Copy files between local machine and remote server.
  * Example: Uploading a config file to a payment server.

* **SFTP (SSH File Transfer Protocol):**

  * Secure, reliable file transfer over SSH.
  * Supports resume, directory listing.
  * Example: Transferring encrypted transaction logs from branch to HQ.

---

### 5.4 FinTech Example: Cloud Server Management ☁️

* Bank’s servers running in AWS or Azure are accessed by admins.
* Instead of risky Telnet, admins use **SSH with key-based authentication**.
* Ensures:

  * Secure maintenance of transaction systems.
  * Protection against MITM sniffing on the internet.

👉 Without SSH, attackers could intercept admin credentials and control entire bank servers.

## 7. Common Threats to Secure Protocols 🚨

Even though protocols like TLS, IPSec, SSH, and PGP are designed to be secure, **implementation flaws and misconfigurations** can create vulnerabilities. Let’s look at the major ones 👇

---

### 7.1 SSL Stripping Attacks 🥷

* **Definition:** An attacker downgrades a secure HTTPS connection into plain HTTP.
* **How it works:**

  * User types `bank.com` → Browser first requests `http://bank.com`.
  * Attacker intercepts and **keeps connection on HTTP**, removing HTTPS upgrade.
  * User still sees the website but with **no encryption**.
* **Impact in FinTech:**

  * Login credentials and transactions sent in plaintext → stolen by attacker.
* **Defense:**

  * Use **HSTS (HTTP Strict Transport Security)** to force HTTPS.
  * Educate users to check 🔒 padlock in browsers.

👉 Example: Customer logs into `mywallet.com` at a café. MITM attacker strips HTTPS → steals login + PIN.

---

### 7.2 Heartbleed Bug 🫀💔 (OpenSSL Vulnerability, 2014)

* **Definition:** A serious flaw in OpenSSL’s TLS heartbeat extension.
* **How it works:**

  * Attackers send malicious “heartbeat” requests.
  * Server leaks chunks of memory back.
  * Sensitive data like passwords, session keys, and even private keys can leak.
* **Impact in FinTech:**

  * Could expose entire banking server’s secrets.
  * Attackers could impersonate bank servers.
* **Defense:**

  * Patch/update OpenSSL immediately.
  * Regular vulnerability scanning.

👉 Example: An online payment gateway left unpatched could leak **customer card numbers** in memory.

---

### 7.3 Downgrade Attacks (Forcing Older TLS Versions) ⬇️

* **Definition:** Attacker tricks client/server into using **weak, outdated versions of TLS or ciphers**.
* **How it works:**

  * During TLS handshake, attacker intercepts and tells both sides:

    * “I only support TLS 1.0.”
  * Both downgrade, using weak encryption.
* **Impact in FinTech:**

  * Easy decryption of traffic with modern computing.
  * Transaction data compromised.
* **Defense:**

  * Disable weak versions (TLS 1.0/1.1, SSLv2/3).
  * Use TLS 1.3 wherever possible.

👉 Example: A stock trading app forced down to TLS 1.0 → attacker decodes live trading instructions.

---

### 7.4 Certificate Mismanagement 📜❌

* **Definition:** Security failures due to poor handling of digital certificates.
* **Examples:**

  * Expired certificates (users ignore warnings).
  * Self-signed certificates (no real trust).
  * Wrongly issued certificates by a compromised CA.
  * Certificates not revoked after compromise.
* **Impact in FinTech:**

  * Users may connect to fake banking sites that “look” valid.
  * Downtime when expired certificates block services.
* **Defense:**

  * Automated certificate management (e.g., Let’s Encrypt + monitoring).
  * Certificate pinning in mobile apps.

👉 Example: In 2020, Microsoft Teams **went down for hours** due to an expired certificate — imagine a payment gateway facing that 😬.

# 📅 Week 8: Secure Communication Protocols 🔐📡

**Tool:** `sslyze` (SSL/TLS Scanner) 🛡️

---

## 🎯 **Objective**

* Understand how computers **secure communication** using SSL/TLS.
* Learn what happens during a **TLS handshake**.
* Know why some **protocols and ciphers** are unsafe today.
* See how **certificates** prove a website’s identity.

---

## 📘 **What is SSL/TLS?**

* **SSL (Secure Sockets Layer)** and **TLS (Transport Layer Security)** are protocols that **encrypt** data sent between your browser and a website.
* They make sure that:

  * 💬 Nobody can read your messages (**Confidentiality**)
  * 🧩 Messages aren’t changed on the way (**Integrity**)
  * 🧾 You’re talking to the real website, not a fake one (**Authentication**)

---

## 🔄 **How a TLS Connection Works**

1. **ClientHello:** The browser says “Hi” and tells the server what versions and ciphers it supports.
2. **ServerHello:** The server replies with its chosen version and cipher, plus its **certificate**.
3. **Certificate Check:** The browser verifies the certificate is valid and signed by a trusted authority.
4. **Key Exchange:** Both sides agree on a secret key to encrypt the data.
5. **Secure Channel:** From now on, everything is safely encrypted 🔒.

---

## 🧠 **TLS Versions**

| Version       | Status         | Notes                          |
| ------------- | -------------- | ------------------------------ |
| SSL v2 / v3   | ❌ Obsolete     | Very weak, never use           |
| TLS 1.0 / 1.1 | ⚠️ Deprecated  | Vulnerable, should be disabled |
| TLS 1.2       | ✅ Secure       | Common and safe                |
| TLS 1.3       | 🏆 Most Secure | Faster and more modern         |

👉 Always prefer **TLS 1.2 or 1.3** only.

---

## 🧮 **What is a Cipher Suite?**

A cipher suite is a *set of algorithms* that decide **how** data is protected.
It defines:

* 🔑 How the key is exchanged (e.g., ECDHE)
* 🪪 How the server proves its identity (e.g., RSA or ECDSA)
* 🧊 How the data is encrypted (e.g., AES-GCM, ChaCha20)

**Good example:** `TLS_ECDHE_RSA_WITH_AES_256_GCM_SHA384`
→ uses strong encryption and authentication.

---

## 🪪 **Certificates & HTTPS**

* Certificates are like **digital ID cards** for websites.
* They are issued by **Certificate Authorities (CAs)**.
* A valid certificate shows:

  * Who owns the site
  * When it expires
  * Which CA signed it

If a certificate is **expired**, **self-signed**, or **mismatched**, the browser shows a warning ⚠️.

---

## 🚨 **Common Problems & Attacks**

| Problem             | Description               | Fix                          |
| ------------------- | ------------------------- | ---------------------------- |
| 🔓 Old TLS versions | Easy to break             | Disable TLS 1.0/1.1          |
| 🧱 Weak Ciphers     | RC4 or 3DES               | Use AES-GCM or ChaCha20      |
| 💀 Heartbleed       | Memory leak in OpenSSL    | Update OpenSSL               |
| 🤖 ROBOT Attack     | Exploits RSA padding      | Disable old RSA key exchange |
| 🧬 Downgrade Attack | Forces weaker TLS version | Enforce TLS 1.2+ and HSTS    |

---

## 🧰 **How to Stay Secure**

- ✅ Use only TLS 1.2 and 1.3
- ✅ Allow only modern ciphers (AES-GCM, ChaCha20-Poly1305)
- ✅ Use certificates from trusted CAs
- ✅ Renew certificates before expiry
- ✅ Enable HSTS (HTTP Strict Transport Security)
- ✅ Keep OpenSSL and web servers updated

---

## 💼 **Why It Matters (FinTech Example)**

In online banking or digital-payment systems:

* Transactions, card data, and personal info flow over TLS.
* If TLS is weak → attackers can steal or alter data.
* Regulations (e.g., PCI DSS) require strong TLS settings and regular security audits.

---

✅ **You Now Know:**

* What TLS does and why it matters
* Difference between protocol versions
* Importance of cipher suites and certificates
* How weak settings can be exploited
