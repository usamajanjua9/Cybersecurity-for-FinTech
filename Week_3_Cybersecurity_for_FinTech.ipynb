## **1. Introduction to Cyber Attacks in FinTech 🌐**

---

### **1.1 What is a Cyber Attack? 🕵️‍♂️**

* A **cyber attack** is a **deliberate attempt** by an individual or group to **steal, damage, or disrupt digital systems**.
* It often targets:

  * **Data** 📂 (stealing customer credentials, credit card numbers).
  * **Systems** ⚙️ (disabling ATMs, mobile wallets, or payment gateways).
  * **Services** 🌍 (blocking access to online banking or exchanges).

**Types of Goals:**

1. **Confidentiality Breach 🔒** – Stealing sensitive financial information.
2. **Integrity Breach ✅** – Altering transactions or account balances.
3. **Availability Breach ⚡** – Bringing systems down so no one can access them.

👉 *FinTech Example:* A hacker stealing mobile wallet PIN codes = **confidentiality attack**, while shutting down Visa’s payment network = **availability attack**.

---

### **1.2 Why Financial Systems are Prime Targets 🎯**

Financial technology platforms (mobile wallets, payment gateways, online banking) are **high-value targets** because:

1. **Money is Involved 💵**

   * Direct financial gain → attackers can steal or redirect funds instantly.

2. **Sensitive Data 📊**

   * Banking apps hold personal identification numbers, card details, and transaction histories.
   * Hackers can sell this data on the **dark web**.

3. **High Uptime Requirements ⏱️**

   * Banks and wallets must run **24/7**.
   * Even one hour of downtime can cost **millions of dollars**.
   * Attackers exploit this to extort ransom.

4. **Large User Base 🌍**

   * Billions of daily transactions.
   * More users → more chances of phishing or malware spreading.

5. **Complex Systems 🔀**

   * FinTech integrates multiple services (apps, APIs, cloud, blockchain).
   * Complexity = more vulnerabilities.

👉 *FinTech Example:* Attackers prefer hitting **payment gateways during peak hours** (like Black Friday sales) because downtime = massive financial losses.

---

### **1.3 Recent Global and FinTech-Specific Incidents 🌍💳**

1. **Bangladesh Bank Heist (2016) 💰**

   * Hackers used the SWIFT interbank messaging system.
   * Stole **\$81 million** by sending fraudulent transfer requests.

2. **Capital One Breach (2019) 🏦**

   * Misconfigured firewall exposed **100 million customer records**.
   * Data included Social Security numbers and bank account details.

3. **Equifax Breach (2017) 🔐**

   * Credit reporting agency hacked.
   * **147 million records** stolen, including financial data.

4. **Mobile Wallet Fraud (Ongoing) 📱**

   * SIM swap attacks → hackers trick telecom providers to issue a duplicate SIM.
   * Gain access to OTPs (One-Time Passwords) for wallets like Easypaisa, PayPal, M-Pesa.

5. **Cryptocurrency Exchange Hacks (Multiple Cases) 🪙**

   * **Mt. Gox (2014):** 850,000 Bitcoins stolen.
   * **Binance (2019):** \$40 million worth of crypto stolen due to compromised API keys.

---

✅ **Key Takeaway:**
Cyber attacks are not random — **FinTech is always in the crosshairs** because of **money, data, and uptime requirements**.

# **2. Social Engineering Attacks 🎭**

---

## **2.1 Phishing 🎣**

**Definition:**
Phishing is a **deceptive cyber attack** where attackers impersonate **trusted institutions** (like banks, wallets, or payment gateways) to trick users into revealing sensitive data such as passwords, card numbers, or One-Time Passwords (OTPs).

Phishing is one of the **most common and successful attack methods** in the FinTech world.

---

### **A) Email Phishing 📧**

* Attackers send emails pretending to be from a **bank or FinTech service**.
* Email often looks **legitimate**, using logos and branding.
* Victims are asked to:

  * Click a fake link.
  * Update their account.
  * Verify suspicious activity.

**Example:**

* A customer receives an email that looks like it came from *PayPal*.
* Subject line: *“Your account has been suspended. Click here to verify.”*
* Clicking the link → redirects to a **fake PayPal website** where login credentials are stolen.

**Impact in FinTech:**

* Leads to **account takeover**.
* Funds drained or sensitive data sold on the **dark web**.

---

### **B) Spear Phishing 🎯**

* More **targeted** than regular phishing.
* Attackers research specific individuals (e.g., bank employees, FinTech developers, finance managers).
* Emails are **personalized** with names, job roles, or recent events.

**Example:**

* A hacker targets a finance manager at a bank with an email:

  * *“Hi Ali, here is the updated quarterly budget file. Please review.”*
* The attachment contains **malware (Trojan)** that compromises the system.

**Impact in FinTech:**

* Insider-level access gained.
* Attackers can move inside the financial network undetected.
* Often a starting point for **large-scale breaches**.

---

### **C) Smishing (SMS Phishing) & Vishing (Voice Phishing) 📱**

**Smishing (SMS + Phishing):**

* Fake SMS messages pretending to be from **mobile wallets or banks**.
* Contain malicious links or request sensitive info.

**Example:**

* An SMS: *“Your Easypaisa account will be blocked. Verify now: [http://fake-link.com”](http://fake-link.com”)*
* Victim clicks → enters **PIN and OTP** → attacker uses OTP to drain wallet.

**Vishing (Voice + Phishing):**

* Attackers call victims pretending to be **bank staff**.
* Convince victims to share OTPs, card details, or login credentials.

**Example:**

* Caller claims: *“This is from your bank’s fraud department. Please share the OTP you received to block a suspicious transfer.”*
* Victim shares OTP → attacker completes fraudulent transaction.

---

### **D) FinTech Example: OTP Theft via SMS Phishing 🔑**

* OTPs (One-Time Passwords) are widely used in FinTech for transaction verification.
* Attackers send fake SMS asking for OTP verification.
* Victim unknowingly shares OTP with attacker.
* Result:

  * Funds stolen from mobile wallets.
  * Unauthorized transactions on credit/debit cards.

---

### **Defenses Against Phishing 🛡️**

1. **User Awareness & Training** 📚 – Teach users to recognize fake emails/SMS.
2. **Multi-Factor Authentication (MFA)** 🔐 – Even if password is stolen, attacker cannot access account without second factor.
3. **Email Filtering Systems** 📧 – Block suspicious messages with advanced spam filters.
4. **Anti-Phishing Solutions** 🛡️ – Detect phishing URLs and fake websites.
5. **Regular Alerts from FinTech Providers** 📢 – Warn users not to share OTPs, PINs, or passwords.

---

✅ **Key Takeaway:**
Phishing is **the easiest and most effective attack** against financial users.

* **Email phishing** tricks general users.
* **Spear phishing** targets employees.
* **Smishing & vishing** exploit OTPs in mobile wallets.

👉 Without awareness and layered defenses, phishing can bypass even the most advanced security systems.

# **2.2 Man-in-the-Middle (MITM) Attacks 🕵️‍♂️**

---

### **What is a MITM Attack?**

A **Man-in-the-Middle attack** happens when an attacker secretly **intercepts and alters communication** between two parties who think they are communicating directly.

* Victim → Attacker → Bank/Wallet Server
* The attacker can:

  * **Listen** 👂 to sensitive data (credentials, PINs, OTPs).
  * **Modify** ✍️ the communication (change transaction details).
  * **Redirect** 🔄 users to fake websites.

---

### **A) How Interception Works 🧩**

1. **Victim initiates a connection** (e.g., logging into PayPal).
2. **Attacker positions themselves** between the victim and the server.

   * Can be done through:

     * ARP poisoning (local network attack).
     * Rogue Wi-Fi hotspot.
     * Compromised router.
3. **Victim’s data passes through attacker’s machine first** before reaching the real server.
4. Attacker can **steal or alter data** without the victim noticing.

👉 **FinTech Risk:** Attacker intercepts login credentials or modifies a transaction from PKR 5,000 to PKR 50,000.

---

### **B) HTTPS Stripping 🔓**

* Secure connections use **HTTPS (Hypertext Transfer Protocol Secure)**.
* Attackers downgrade the connection to plain **HTTP** using tools like `sslstrip`.
* Victims think they are secure, but data is actually sent in **clear text**.

**Example:**

* Customer opens `https://easypaisa.com.pk`.
* Attacker forces downgrade to `http://easypaisa.com.pk`.
* Customer logs in → PIN and OTP visible to attacker.

👉 This is extremely dangerous for **mobile wallets and online banking**.

---

### **C) Wi-Fi Eavesdropping 📡**

* Attackers set up **fake Wi-Fi hotspots** (e.g., “Free\_Cafe\_WiFi”).
* Victims connect, thinking it’s safe.
* All traffic passes through attacker’s laptop first.
* Even on genuine Wi-Fi, if not encrypted, attackers can capture packets.

**What attackers capture:**

* Login usernames and passwords.
* Credit/debit card numbers.
* OTPs sent via unprotected channels.

**FinTech Example:**

* A student connects to free Wi-Fi in a coffee shop and logs into JazzCash.
* Attacker on the same Wi-Fi captures the **unencrypted packets**.

---

### **D) FinTech Example: Hijacking Mobile Wallet Traffic 📱💳**

1. Victim uses mobile wallet app (Easypaisa, PayPal, M-Pesa) on **public Wi-Fi**.
2. Attacker performs **ARP poisoning** or sets up a **rogue access point**.
3. Victim’s traffic is routed through attacker.
4. Attacker:

   * Steals login credentials.
   * Captures OTPs.
   * Alters transactions (PKR 1,000 transfer changed to PKR 10,000).

---

### **E) Defenses Against MITM 🛡️**

1. **Use HTTPS Everywhere 🔒** → Enforce HTTPS with HSTS (HTTP Strict Transport Security).
2. **TLS Certificates 🔑** → Apps must check server certificates to prevent fake servers.
3. **VPNs (Virtual Private Networks) 🌍** → Encrypt all communication, even on unsafe Wi-Fi.
4. **Secure Wi-Fi Practices 📶** → Avoid using financial apps on public hotspots.
5. **Mobile Wallet Security 🚨** → Apps should warn if connection is insecure.

---

✅ **Key Takeaway:**
MITM attacks are **silent but deadly**.

* Attackers don’t just steal data — they can **manipulate transactions in real time**.
* In FinTech, a single MITM attack can lead to **account takeover and large-scale fraud**.

# **3. Network and Infrastructure Attacks 🌍**

---

## **3.1 Distributed Denial of Service (DDoS) 🌊**

---

### **What is a DDoS Attack? ⚠️**

A **Distributed Denial of Service (DDoS) attack** is when attackers flood a target (like a bank server or payment gateway) with **massive amounts of fake traffic**.

* The target becomes **overloaded** and **legitimate users cannot access services**.
* It is like thousands of fake customers crowding into a bank branch so real customers cannot get in.

👉 **Key Feature:** It is “**distributed**” → attack traffic comes from **many compromised computers/devices** worldwide.

---

### **A) How Attackers Launch DDoS 🚨**

1. **Botnets 🤖**

   * Attackers infect thousands or millions of computers, IoT devices, or even smart appliances with malware.
   * All these compromised devices form a **botnet army**.
   * The attacker commands the botnet to flood a target server.

   **Example:** The **Mirai Botnet (2016)** used hacked IoT devices (like webcams) to launch one of the largest DDoS attacks ever.

---

2. **Amplification Attacks 🔊**

   * Attackers exploit open servers (like DNS or NTP servers) to reflect and **amplify traffic**.
   * Small attacker request → huge response sent to the victim.
   * Common techniques:

     * **DNS Amplification**
     * **NTP Amplification**
     * **Smurf Attack (ICMP amplification)**

---

### **B) Impact on Availability of Payment Gateways 💳**

* **Availability is critical in FinTech.** If customers cannot complete transactions, trust is lost.
* **Financial Losses:** Even **one hour** of downtime can cost banks **millions of dollars**.
* **Reputation Damage:** Customers may switch to competitors.
* **Regulatory Fines:** Financial institutions are required to ensure continuous service.

👉 Example Impacts:

* Customers cannot pay online using Visa/Mastercard.
* Mobile wallets (Easypaisa, PayPal) fail during shopping hours.
* E-commerce merchants lose revenue.

---

### **C) FinTech Example: DDoS Attack During High-Volume Shopping Days 🛒**

* Imagine **Black Friday** in Pakistan.
* Thousands of customers use Easypaisa and JazzCash at the same time.
* Attackers launch a DDoS attack using a **botnet of 50,000 compromised IoT devices**.
* The payment gateway servers become overloaded.
* Result:

  * Transactions fail ❌
  * Customers cannot pay for items 🛑
  * Businesses lose millions 💸
  * Customers lose trust in the FinTech platform

---

### **D) Defenses Against DDoS 🛡️**

1. **Load Balancers** ⚖️ → Distribute traffic across multiple servers.
2. **Content Delivery Networks (CDNs)** 🌍 → Absorb and filter attack traffic.
3. **Rate Limiting** ⏱️ → Restrict requests per IP address.
4. **DDoS Mitigation Services** 🛡️ → Providers like Cloudflare, Akamai, and AWS Shield filter malicious traffic.
5. **Anycast Routing** 📡 → Spread attack traffic across many global servers.
6. **Failover & Redundancy** 🔄 → Backup servers take over if primary servers fail.

---

✅ **Key Takeaway:**

* DDoS attacks threaten the **availability** part of the CIA triad.
* FinTech systems must prepare for them, especially during **peak transaction times**.

# **3.2 DNS Attacks 🌍🧪**

---

### **What is DNS?**

* The **Domain Name System (DNS)** is like the **phonebook of the Internet**.
* It translates domain names (e.g., `www.paypal.com`) into **IP addresses** so computers can connect.

---

### **A) DNS Spoofing / Poisoning 🧨**

* Attackers corrupt the DNS cache with **false information**.
* When a user types `www.easypaisa.com`, the DNS server returns the **IP of a fake site** instead of the real one.
* Victim unknowingly logs into the fake site → attacker steals credentials.

**Example:**

* A fake Easypaisa login page collects **PIN + OTP**.
* Users think they are on the real site.

---

### **B) Redirection to Fake Banking Websites 🎭**

* Attackers set up **lookalike domains** (e.g., `paypa1.com` or `jazzcash-support.net`).
* Victims are tricked into clicking malicious links from phishing emails/SMS.
* Login details and card info are captured.

---

### **C) Defenses Against DNS Attacks 🛡️**

1. **DNSSEC (DNS Security Extensions)** → Ensures DNS responses are authentic.
2. **Secure Recursive Resolvers** → Use trusted DNS providers (Google DNS, Cloudflare DNS).
3. **User Awareness** → Check for HTTPS and valid certificates.
4. **Threat Intelligence** → Banks monitor for lookalike phishing domains.

---

### **FinTech Example 💳**

* In 2022, attackers targeted Indian mobile banking users with fake DNS redirections.
* Customers entered OTPs on **phishing domains**, leading to **massive wallet fraud**.

---

# **3.3 SQL Injection and Web Exploits 💻💉**

---

### **A) Injection Flaws in Financial Applications 💉**

* **SQL Injection (SQLi):**

  * Attackers insert malicious Structured Query Language (SQL) statements into an input field.
  * The backend database executes them as valid commands.

**How it works:**

1. Victim enters normal input → “User: Ali, Password: 1234.”
2. Attacker enters malicious input → `’ or 1 = 1 --`
3. Database interprets this as **true for all users** → attacker gains access.

**Impact in FinTech:**

* Exposing **entire banking customer database**.
* Modifying balances or transaction logs.
* Deleting financial records.

---

### **B) Cross-Site Scripting (XSS) 🕷️**

* Attackers inject malicious scripts into a web application.
* When other users visit the page, the script executes in their browser.

**Types:**

1. **Stored XSS** → Script saved in database (appears in forums, comments, support chat).
2. **Reflected XSS** → Script in a URL link that executes immediately.

**Impact in FinTech:**

* Stealing **cookies/session tokens**.
* Hijacking **online banking sessions**.
* Redirecting to **malicious payment gateways**.

---

### **C) Real-World Example 🔎**

* Attackers injected SQL queries into a vulnerable mobile wallet API.
* They retrieved **customer details, PINs, and balance info**.
* XSS vulnerabilities in bank chat portals allowed attackers to **steal session tokens**.

---

### **D) Defenses Against Injection & XSS 🛡️**

1. **Input Validation** → Only allow expected input (e.g., numbers for account fields).
2. **Parameterized Queries / Prepared Statements** → Prevent SQL code from being injected.
3. **Web Application Firewall (WAF)** → Blocks malicious SQL/XSS requests.
4. **Content Security Policy (CSP)** → Prevents unauthorized scripts from executing.
5. **Regular Security Audits** → Identify and patch vulnerabilities.

---

✅ **Key Takeaway:**

* **DNS attacks** trick users into connecting to **fake servers**.
* **SQL Injection and XSS** exploit weaknesses in the **application itself**.
* Together, these threaten **both customer trust and the financial backbone of FinTech apps**.

# **4. Malware and Malicious Software 🦠**

---

## **4.1 Viruses and Worms 🧬**

---

### **A) What are Viruses and Worms?**

* **Virus 🦠**

  * Malicious software that **attaches itself to files or programs**.
  * Needs human action (like opening a file or running an app) to spread.
  * Example: A fake “bank statement PDF” that, when opened, installs a virus.

* **Worm 🪱**

  * Malicious software that **self-replicates and spreads automatically** through networks.
  * Does not need human action.
  * Example: A worm spreads through a bank’s internal network infecting every server.

👉 **Key Difference:**

* Virus = needs user to trigger it.
* Worm = spreads by itself.

---

### **B) Propagation Methods 🚀**

1. **Email Attachments 📧**

   * Fake “account statement” or “tax refund notice” carrying a malicious file.

2. **USB Drives 💽**

   * Infected removable media spreading malware into banking systems.

3. **Network Sharing 🌍**

   * Worms scan for open ports and weak passwords, spreading across computers.

4. **Drive-By Downloads 🌐**

   * Visiting a compromised website triggers hidden malware download.

---

### **C) Example: Banking Worms Stealing Credentials 💳**

* **Zeus Trojan + Worm Hybrid**

  * Spread through phishing emails with malicious attachments.
  * Installed keyloggers to record **usernames, passwords, and PINs**.
  * Targeted **online banking systems worldwide**, stealing millions.

* **Conficker Worm**

  * Exploited Microsoft Windows vulnerability.
  * Spread to millions of machines, many inside corporate and financial networks.
  * Could be used as a **botnet for DDoS attacks** on banks.

**FinTech Scenario:**

* A worm infects a payment gateway server via an unpatched vulnerability.
* It spreads across the bank’s internal systems.
* Attackers collect **customer login details** and sell them on the dark web.

---

### **D) Defenses Against Viruses and Worms 🛡️**

1. **Antivirus and Anti-Malware Software** → Detect and remove known threats.
2. **Patch Management** → Regular updates close vulnerabilities worms exploit.
3. **Email Filtering Systems** → Block malicious attachments.
4. **Network Segmentation** → Prevent a worm from spreading across the entire network.
5. **User Awareness Training** → Employees should not open suspicious files.

---

✅ **Key Takeaway:**

* Viruses and worms are often the **first stage** of a bigger financial cyberattack.
* Once inside, they **steal banking credentials, spread through networks, and open doors for Trojans or ransomware**.

# **4.2 Trojans 🐴**

---

### **A) What is a Trojan?**

* A **Trojan horse** is malicious software that **pretends to be a legitimate program** but secretly performs harmful actions.
* Unlike worms, Trojans **do not self-replicate**.
* They rely on **tricking the victim** into installing them.

**How they get in:**

* Fake banking apps on app stores 📱
* Malicious email attachments 📧
* Cracked or pirated software 💽
* Phishing links 🌍

👉 Named after the **Trojan horse from Greek mythology** — looks harmless but hides attackers inside.

---

### **B) Banking Trojans 🏦💳**

Banking Trojans are designed specifically to **steal financial information** such as:

* Online banking login credentials.
* Card details and PINs.
* One-Time Passwords (OTPs).
* Digital wallet access tokens.

---

#### **1. Zeus Trojan (Zbot)**

* One of the most famous banking Trojans.
* Spread via phishing emails disguised as bank notifications.
* Installed keyloggers to record login details.
* Stole **hundreds of millions of dollars** from online banking accounts worldwide.

👉 **FinTech Example:** A victim logs into an online wallet, Zeus captures keystrokes (username + password), then attacker initiates unauthorized transfers.

---

#### **2. Emotet Trojan**

* Originally a banking Trojan, later evolved into a **malware delivery platform**.
* Spread via malicious email attachments (“invoice,” “payment confirmation”).
* Delivered other banking Trojans and ransomware.

👉 **FinTech Example:** An employee at a financial institution opens an infected invoice. Emotet spreads across the bank network and installs TrickBot to steal credentials.

---

#### **3. TrickBot Trojan**

* Successor to Zeus, one of the **most advanced banking Trojans**.
* Specialized in targeting financial institutions.
* Can:

  * Steal credentials.
  * Intercept browser sessions.
  * Spread across networks like a worm.
* Often used as a **loader** for ransomware (e.g., Ryuk).

👉 **FinTech Example:** TrickBot infects multiple ATM servers, allowing attackers to **manipulate withdrawal limits** and perform large-scale fraud.

---

### **C) Backdoors in FinTech Apps 🔑**

Some Trojans also install **backdoors**:

* Hidden access that allows attackers to:

  * Log in remotely.
  * Execute commands.
  * Bypass security checks.
* In FinTech apps, a backdoor could allow attackers to:

  * Access **transaction APIs** without authentication.
  * Monitor customer activity silently.
  * Steal entire databases of user credentials.

**Example Scenario:**

* A fake mobile wallet app appears on an unofficial app store.
* It looks and feels real but has a **Trojan backdoor**.
* Users install it → attackers gain access to all transactions.

---

### **D) Defenses Against Trojans 🛡️**

1. **App Store Security** → Only download banking/FinTech apps from trusted sources.
2. **Antivirus and Endpoint Protection** → Detects known Trojans like Zeus or TrickBot.
3. **Multi-Factor Authentication (MFA)** → Even if passwords are stolen, attacker cannot log in without OTP or biometrics.
4. **Secure Coding & App Audits** → Ensure no hidden backdoors exist in apps.
5. **Employee Awareness Training** → Prevent clicking on phishing emails with malicious attachments.

---

✅ **Key Takeaway:**

* Trojans like **Zeus, Emotet, and TrickBot** show how **deception + persistence** can devastate financial institutions.
* Backdoors in FinTech apps are especially dangerous because they give attackers **long-term, hidden access** to systems and customer data.

# **4.3 Ransomware 🔒💰**

---

### **A) What is Ransomware?**

* **Ransomware** is a type of malicious software that **encrypts a victim’s files or systems** and then demands payment (a ransom) for restoring access.
* Usually spread through:

  * Phishing emails with malicious attachments 📧
  * Exploiting unpatched systems 🛠️
  * Malicious downloads from fake websites 🌍

👉 In FinTech, ransomware is one of the most feared threats because **downtime = instant financial loss**.

---

### **B) Encryption of Banking Data 🔐**

* Once ransomware infects a banking or payment system, it:

  1. **Scans** for sensitive files (databases, financial records).
  2. **Encrypts** them using strong cryptography.
  3. Replaces files with encrypted versions that cannot be accessed without the attacker’s key.
* A ransom note is then displayed demanding payment (often in cryptocurrency).

**Impact on FinTech:**

* Banks cannot access customer data.
* Payment gateways are shut down.
* ATMs cannot process transactions.

---

### **C) Double Extortion Strategy 🎭**

Modern ransomware groups use **double extortion**:

1. **First Extortion:** Encrypt the victim’s data and demand ransom for the decryption key.
2. **Second Extortion:** Threaten to **leak stolen data** publicly if ransom is not paid.

👉 This makes organizations pay even if they have backups, because **reputation and trust** are at risk.

**Example:**

* A bank’s customer database (with card details, PINs, transactions) is stolen.
* Hackers demand ransom: *“Pay us, or we release all your customers’ data online.”*

---

### **D) FinTech Example: Ransomware Locking ATM Servers 🏧**

* Scenario:

  1. Attackers deliver ransomware via a phishing email to a bank employee.
  2. The malware spreads into the bank’s internal network.
  3. ATM servers are infected and **locked down**.
  4. Customers across the country see ATMs showing an error: *“Out of Service.”*
  5. Bank receives a ransom note demanding millions in cryptocurrency.

**Impact:**

* Financial losses due to ATM downtime.
* Customer trust shattered.
* Regulatory fines for service disruption.

---

### **E) Defenses Against Ransomware 🛡️**

1. **Regular Backups 🔄** → Keep secure, offline backups of all financial data.
2. **Patch Management 🔧** → Apply updates to fix vulnerabilities ransomware exploits.
3. **Email Filtering ✉️** → Block malicious attachments and links.
4. **Endpoint Detection and Response (EDR) 🖥️** → Detect ransomware behavior early.
5. **Network Segmentation 🚪** → Stop malware from spreading across entire banking systems.
6. **User Awareness Training 📚** → Teach staff not to open suspicious files or links.

---

✅ **Key Takeaway:**

* Ransomware is **not just about money**, it’s about **trust**.
* In FinTech, ransomware can freeze **payment systems, ATMs, mobile wallets, and gateways** in minutes.
* The **double extortion model** makes it even more dangerous since both **availability and confidentiality** are at risk.

# **4.4 Spyware and Keyloggers 🕵️‍♀️⌨️**

---

### **A) What is Spyware?**

* **Spyware** is malicious software that secretly monitors user activity and collects sensitive information.
* It often runs silently in the background, without the victim’s knowledge.
* Data stolen may include:

  * Login credentials 🔑
  * Card numbers 💳
  * Browsing history 🌍
  * Banking habits 🏦

---

### **B) What are Keyloggers?**

* A **keylogger** is a type of spyware that **records every keystroke** typed on a device.
* Attackers use it to capture:

  * Online banking usernames and passwords.
  * PIN codes typed into banking apps.
  * One-Time Passwords (OTPs) entered during login.

👉 Keyloggers can be:

* **Software-based** → Hidden programs installed on a system.
* **Hardware-based** → Small devices placed between the keyboard and computer.

---

### **C) Credential Theft from Banking Portals 🏦**

* When customers log into online banking portals or payment gateways:

  1. Keylogger records keystrokes (username + password).
  2. Spyware may also take screenshots of login pages.
  3. Data is sent to the attacker’s server.

**Impact in FinTech:**

* Unauthorized access to accounts.
* Funds transferred out without customer’s knowledge.
* Stolen credentials sold on the **dark web** for further fraud.

**Example:**

* Zeus Trojan (a banking Trojan) included keylogging modules to steal credentials from online banking logins.

---

### **D) Mobile Spyware in FinTech Apps 📱**

* Attackers embed spyware inside **fake or modified FinTech apps**.
* Once installed, spyware can:

  * Steal SMS-based OTPs.
  * Record screen activity.
  * Forward transaction notifications to attackers.
  * Capture biometric data (fingerprint/face ID bypass attempts).

**Real-World Case:**

* **Anubis Banking Trojan (Android)**: Disguised as a currency converter app.
* Once installed, it activated keylogging and screen capture to steal mobile banking credentials.

---

### **E) Defenses Against Spyware and Keyloggers 🛡️**

1. **Antivirus and Anti-Spyware Tools** → Detect hidden spyware programs.
2. **Multi-Factor Authentication (MFA)** → Even if credentials are stolen, login requires OTP/biometrics.
3. **App Store Safety** → Only download apps from trusted sources (Google Play, Apple App Store).
4. **Mobile Security Controls** → Enable Play Protect (Android) or App Store protections.
5. **Encryption of Keystrokes** → Some secure banking apps use keystroke encryption to block keyloggers.
6. **User Awareness** → Do not install unknown apps or click suspicious links.

---

✅ **Key Takeaway:**

* Spyware and keyloggers silently steal **credentials, PINs, OTPs, and card details**.
* In FinTech, they are especially dangerous because **credentials = money**.
* Mobile spyware is rising fast, as wallets and banking apps dominate financial transactions.

# **5. Threat Modeling in Financial Systems 🏦**

---

## **5.1 What is Threat Modeling? 🤔**

* **Threat modeling** is a **structured approach** to identifying, analyzing, and addressing potential **security threats** before they cause damage.
* It answers 4 main questions:

  1. **What are we building?** (banking app, payment gateway, ATM network).
  2. **What can go wrong?** (phishing, SQL injection, DDoS).
  3. **What are we doing about it?** (firewalls, MFA, encryption).
  4. **Did we do a good job?** (testing, audits, penetration tests).

**Why it matters in FinTech:**

* FinTech systems handle **money + sensitive data**.
* Threat modeling ensures security is **built-in from the start**, not just added later.
* Helps balance **CIA triad** (Confidentiality 🔒, Integrity ✅, Availability ⚡).

👉 Example:

* Before launching a new Easypaisa feature, threat modeling identifies risks like **MITM attacks, SIM swap fraud, SQL injection in APIs** → fixes are applied early.

---

## **5.2 Common Threat Modeling Methodologies 🛠️**

There are several frameworks to guide threat modeling. Three popular ones are:

---

### **A) STRIDE Framework 🧩** (by Microsoft)

Focuses on **types of threats**.

**STRIDE =**

* **S – Spoofing Identity** 🕵️ → Pretending to be someone else (e.g., fake user login).
* **T – Tampering with Data** ✍️ → Altering transactions or records.
* **R – Repudiation** ❌ → User denies making a transaction (no proof/logging).
* **I – Information Disclosure** 📂 → Leaking PINs, account numbers.
* **D – Denial of Service** 🌊 → Shutting down ATMs or payment gateways.
* **E – Elevation of Privilege** 🔑 → Gaining admin rights without authorization.

**FinTech Example:**

* An attacker tampers with API calls to **increase wallet balance** → STRIDE helps model this risk.

---

### **B) DREAD Model ⚖️**

Used to **prioritize risks** based on impact and likelihood.

**DREAD =**

* **D – Damage Potential** 💥 (How bad is the impact?)
* **R – Reproducibility** 🔁 (How easy is it to reproduce?)
* **E – Exploitability** 🔓 (How easy is it to exploit?)
* **A – Affected Users** 👥 (How many users affected?)
* **D – Discoverability** 🔍 (How easy is it to find the vulnerability?)

**Scoring:** Each category rated 1–10 → higher score = higher priority.

**FinTech Example:**

* SQL Injection flaw in a payment API:

  * Damage = 10 (customer balances exposed).
  * Reproducibility = 9 (easy to repeat).
  * Exploitability = 8 (basic SQL knowledge needed).
  * Affected Users = 10 (all wallet users).
  * Discoverability = 7 (hackers scanning APIs can find it).
  * **Total = 44/50 → Critical Risk** 🚨

---

### **C) PASTA (Process for Attack Simulation and Threat Analysis) 🎭**

* A **seven-step framework** focused on **attacker perspective**.
* Combines **business impact + technical risks**.

**Steps:**

1. Define business objectives (e.g., ensure 24/7 ATM availability).
2. Define technical scope (apps, servers, APIs).
3. Decompose the system (how components interact).
4. Identify threats.
5. Analyze vulnerabilities.
6. Model attacks (how an attacker would strike).
7. Define risk + mitigation strategies.

**FinTech Example:**

* Modeling how a hacker could exploit a **mobile wallet API** by simulating attacks before launch.

---

✅ **Key Takeaway:**

* **STRIDE** → Identifies *what types* of threats exist.
* **DREAD** → Prioritizes *how severe* they are.
* **PASTA** → Models *how attackers would operate* in real life.

---





---

# **5.3 Identifying Assets 🔐**

Before protecting a financial system, we must identify **what needs protection**. In threat modeling, these are called **assets**.

---

### **A) Customer Data 📂**

* **Personal details** (name, CNIC/passport, phone numbers).
* **Credentials** (usernames, passwords, PINs, biometrics).
* **Financial records** (account balances, transaction histories).
* **Regulatory-sensitive data** (KYC/AML data).

👉 *FinTech Example:* If an attacker steals **wallet PINs + OTPs**, they can drain customer balances instantly.

---

### **B) Payment APIs 🔄**

* APIs are the **backbone** of mobile wallets, banking apps, and payment gateways.
* Used for:

  * **Fund transfers** 💸
  * **Balance checks** 📊
  * **Bill payments** ⚡
* APIs are frequent **targets of injection and MITM attacks**.

👉 *FinTech Example:* A vulnerable API could allow hackers to **manipulate transactions** (e.g., transfer PKR 10,000 instead of PKR 1,000).

---

### **C) Servers and Infrastructure 🖥️**

* **Application servers** (mobile wallet backend).
* **Database servers** (store customer and transaction data).
* **Authentication servers** (validate OTPs and MFA requests).
* **ATM servers** and payment switch networks.

👉 *FinTech Example:* A ransomware attack on an ATM switch server could bring down **entire ATM networks nationwide**.

---

### **D) Communication Channels 📡**

* Internet links, APIs, and mobile networks.
* Weaknesses can lead to **MITM, DNS poisoning, or DDoS attacks**.

👉 *FinTech Example:* Public Wi-Fi session hijacking of a customer’s mobile wallet.

---

---

# **5.4 Identifying Attackers 🎭**

In FinTech threat modeling, it’s not just about “what can go wrong” but also **“who might attack us.”** Different attackers have different motives.

---

### **A) External Hackers 👨‍💻**

* Motivated by money 💵.
* Use phishing, Trojans, and DDoS to steal funds.
* Often sell stolen data on the **dark web**.

👉 *Example:* Hackers launching a **DDoS attack on PayPal** in 2010 as part of “Operation Payback.”

---

### **B) Insider Threats 🧑‍💼**

* Employees, contractors, or partners misusing access.
* Risks:

  * Stealing customer data.
  * Manipulating transaction records.
  * Planting backdoors in systems.

👉 *FinTech Example:* A **bank employee** leaks high-value customer details to fraudsters.

---

### **C) Cybercriminal Groups 🕷️**

* Organized gangs running **fraud campaigns**.
* Use **banking Trojans (Zeus, TrickBot)** or **ransomware**.
* Often coordinate across multiple countries.

👉 *Example:* **Carbanak Group** stole **\$1 billion** from over 100 banks worldwide.

---

### **D) Nation-State Attackers 🏴‍☠️**

* Highly skilled, well-funded attackers sponsored by governments.
* Motives:

  * **Disrupt rival economies**.
  * **Espionage** (stealing financial intelligence).
  * **Sabotage** of critical financial infrastructure.

👉 *Example:*

* **North Korean Lazarus Group** → Behind the **2016 Bangladesh Bank heist**, stealing **\$81 million** via SWIFT.

---

### **E) Hacktivists 🎯**

* Attackers motivated by ideology or protest.
* Use DDoS or website defacement to make a political statement.

👉 *Example:* DDoS on **Mastercard and Visa** after they blocked donations to WikiLeaks.

---

---

✅ **Key Takeaway:**

* **Assets** = what we protect (data, APIs, servers, comms).
* **Attackers** = who we protect against (hackers, insiders, groups, nation-states, hacktivists).

Both must be clearly mapped in threat modeling to design **effective defenses**.

---




---

# **5.5 Risk Prioritization (Likelihood × Impact) ⚖️**

Once we identify **assets** and **attackers**, the next step is to **prioritize risks**.

* Not every threat is equally dangerous.
* Some are **very likely but low impact** (e.g., spam emails).
* Others are **rare but catastrophic** (e.g., nation-state ransomware on ATM networks).

👉 Risk = **Likelihood × Impact**

---

### **A) Likelihood 🔮**

* **High Likelihood:** Phishing, weak passwords, API attacks.
* **Medium Likelihood:** DDoS, insider threats.
* **Low Likelihood:** Nation-state targeted attacks.

---

### **B) Impact 💥**

* **High Impact:** Ransomware on ATM servers, SQL injection on payment databases.
* **Medium Impact:** Phishing on a few customers.
* **Low Impact:** Minor website defacement.

---

### **C) Risk Matrix 🗂️**

| **Threat**                  | **Likelihood** | **Impact** | **Risk Level** |
| --------------------------- | -------------- | ---------- | -------------- |
| Phishing on mobile wallet   | High           | Medium     | **High**       |
| SQL Injection on API        | Medium         | High       | **High**       |
| Insider leaking credentials | Medium         | High       | **High**       |
| DDoS on payment gateway     | High           | High       | **Critical**   |
| Nation-state SWIFT attack   | Low            | Very High  | **Critical**   |
| Website defacement          | Medium         | Low        | Low            |

👉 This helps banks/FinTech firms **focus resources on the most dangerous threats first**.

---

# **5.6 Countermeasures and Controls 🛡️**

Once risks are prioritized, controls must be applied to reduce them.

---

### **A) Preventive Controls 🚫**

* **Encryption (TLS, IPSec)** 🔒 → Protects data from interception.
* **Firewalls & WAFs** 🔥 → Block malicious network and web traffic.
* **Access Control & MFA** 🔑 → Prevent unauthorized logins.
* **Patch Management** 🛠️ → Fix vulnerabilities before attackers exploit them.

---

### **B) Detective Controls 🔍**

* **Intrusion Detection/Prevention Systems (IDS/IPS)** 🛰️ → Spot attacks in real-time.
* **SIEM (Security Information and Event Management)** 📊 → Central monitoring of logs.
* **Threat Intelligence Feeds** 🌍 → Identify known phishing domains or malware signatures.

---

### **C) Corrective Controls 🛠️**

* **Incident Response Plans** 📑 → Steps to follow after an attack.
* **Backup & Disaster Recovery** 💾 → Restore services quickly.
* **Forensic Analysis** 🕵️ → Investigate how the attack happened and prevent recurrence.

---

### **D) FinTech-Specific Countermeasures 💳**

1. **Customer Education** 📚 → Remind users: *Never share OTP or PIN*.
2. **Transaction Monitoring Systems** 👁️ → Detect unusual transfers in real time.
3. **Regulatory Compliance** ⚖️ → PCI-DSS, GDPR, State Bank of Pakistan cybersecurity guidelines.
4. **Zero Trust Architecture** 🚪 → Never trust by default, always verify.

---

✅ **Key Takeaway:**

* **Risk prioritization** ensures the most dangerous threats get attention first.
* **Controls** (preventive, detective, corrective) reduce both **likelihood** and **impact**.
* In FinTech, combining **technical defenses + customer awareness** is the only way to build trust.

3. **Extract information** from the database.

# 🔎 What sqlmap Can Do

When you give sqlmap a vulnerable URL (like DVWA’s SQLi page), it can:

### 1. Detect Injection Points

* It tests the input field with lots of **SQL payloads**.
* Example:

  ```
  id=1' OR '1'='1
  ```
* If it gets unexpected results, it knows the site is vulnerable.

---

### 2. Enumerate Databases

* List all databases inside the server:

  ```bash
  sqlmap -u "http://127.0.0.1/dvwa/vulnerabilities/sqli/?id=1&Submit=Submit#" \
  --cookie="PHPSESSID=xxxx; security=low" --dbs
  ```
* Output:

  ```
  available databases [3]:
  [*] dvwa
  [*] information_schema
  [*] mysql
  ```

---

### 3. Extract Tables & Data

* Show tables inside the `dvwa` database:

  ```bash
  sqlmap ... -D dvwa --tables
  ```

* Output:

  ```
  Database: dvwa
  Tables: users, guestbook
  ```

* Dump the **users** table:

  ```bash
  sqlmap ... -D dvwa -T users --dump
  ```

* Output:

  ```
  +----+----------+----------------------------------+
  | id | user     | password                         |
  +----+----------+----------------------------------+
  |  1 | admin    | 5f4dcc3b5aa765d61d8327deb882cf99 |
  |  2 | gordon   | e99a18c428cb38d5f260853678922e03 |
  +----+----------+----------------------------------+
  ```

👉 Notice: It extracts **hashed passwords**.

---

### 4. Advanced Attacks

* Detect DB **version & OS**.
* Read **files from the server**.
* Even **write files** (if permissions allow).

# 🎓 Ettercap Demo (MITM with Kali Linux)

⚠️ **Safety Note**: Do this in a **lab network or VM environment** (not on production WiFi). You’ll need **two machines**:

* **Attacker**: Kali Linux (with Ettercap installed)
* **Victim**: Another VM or real PC on the same LAN

---

## 1. Install Ettercap (if missing)

```bash
sudo apt update
sudo apt install ettercap-graphical -y
```

---

## 2. Start Ettercap GUI

```bash
sudo ettercap -G
```

👉 GUI window opens.

---

## 3. Select Sniffing Interface

* In menu: **Sniff → Unified Sniffing**
* Choose your network interface (e.g., `eth0` for VM, `wlan0` for WiFi).

---

## 4. Scan for Hosts

* Go to **Hosts → Scan for hosts**
* Then **Hosts → Hosts List**
  👉 You’ll see IP addresses of devices on the network.

---

## 5. Set Targets

* Select victim IP → **Add to Target 1**
* Select router/gateway IP → **Add to Target 2**

---

## 6. Start ARP Poisoning

* Go to **MITM → ARP Poisoning**
* Check **Sniff remote connections**
* Click **OK**

👉 Now your Kali machine is “in the middle” of traffic between victim and router.

---

## 7. Observe Traffic

* Open **View → Connections**
* If the victim browses an unencrypted site (e.g., `http://testphp.vulnweb.com/login.php`), you may capture credentials in **cleartext**.

💡 For HTTPS traffic, students will see it’s **encrypted** — good moment to explain why HTTPS matters.

---

## 8. Stop Attack

* **Stop sniffing** → **Quit Ettercap**
* Everything returns to normal.
