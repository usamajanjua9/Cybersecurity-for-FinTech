## **1. Introduction to Networking Models 🌐**

Networking models are like a **blueprint or map 🗺️** that explains **how information travels** from one computer to another over a network.

---

### **1.1 Why Networking Models Matter in Cybersecurity 🔐**


Without a model:

* Cybersecurity experts would be guessing where problems or attacks occur ❌.

With a model:

* We know exactly **where to place defenses** ✅.
* We can explain **which layer was attacked**.
* We can trace problems faster.

**Why they are important in Cybersecurity:**

1. **Finding the source of problems quickly ⏱️**

   * If a customer’s online bank transfer fails, the issue could be at:

     * The **Application Layer** (the bank app has an error).
     * The **Presentation Layer** (encryption handshake failed).
     * The **Network Layer** (packets not being routed correctly).
   * A networking model helps us **zoom in on the exact layer** instead of guessing.

2. **Choosing the correct defense 🛡️**

   * If a hacker injects fake commands into an application (SQL injection), it is an **Application Layer** problem.
   * If a hacker floods the server with fake requests (Denial of Service attack), it is a **Transport Layer** problem.
   * If a hacker spoofs device addresses (Address Resolution Protocol poisoning), it is a **Data Link Layer** problem.

3. **Speaking a common language 🗣️**

   * Networking models allow developers, network engineers, and security teams to communicate clearly.
   * Example: Saying “This is a Layer Three Internet Protocol spoofing attack” is much clearer than “The network is acting strange.”

**FinTech Example 💳:**
When PayPal, Easypaisa, or JazzCash has downtime:

* Engineers first check the **Application Layer** → is the application code broken?
* Then check the **Transport Layer** → is the Transport Control Protocol session failing?
* Then check the **Network Layer** → is the Internet Protocol routing broken?
* Then check the **Physical Layer** → is there a damaged fiber cable?

### **1.2 OSI Model Overview (Seven Layers) 📚**

The **Open Systems Interconnection Model** has **seven layers**. Each layer has a **specific function**. Data travels **from top to bottom** when sending, and **from bottom to top** when receiving.

**Memory Trick:** *“Please Do Not Throw Sausage Pizza Away” 🍕*

* **P**hysical → **D**ata Link → **N**etwork → **T**ransport → **S**ession → **P**resentation → **A**pplication

image.png
---

#### **Layer Seven: Application Layer 📱**

* Closest to the end-user.
* Provides services like email, web browsing, and mobile applications.
* **FinTech Example:** A bank mobile application (HBL, PayPal, Easypaisa) that lets customers send money.

---

#### **Layer Six: Presentation Layer 🎭**

* Responsible for formatting, translating, encrypting, and compressing data.
* Ensures that the data is “understandable” on both sides of the communication.
* **FinTech Example:** Transport Layer Security encrypting customer credit card data before sending it to the server.

---

#### **Layer Five: Session Layer 🗂️**

* Starts, manages, and ends communication sessions between two systems.
* **FinTech Example:** A secure login session in a digital wallet application, where the user stays logged in until they sign out.

---

#### **Layer Four: Transport Layer 📦**

* Ensures reliable delivery of data. Splits messages into smaller chunks and reassembles them correctly.
* Uses protocols like Transmission Control Protocol (reliable) and User Datagram Protocol (fast).
* **FinTech Example:** Transmission Control Protocol ensures that a payment confirmation packet is delivered completely and not lost.

---

#### **Layer Three: Network Layer 🌍**

* Responsible for addressing and routing data.
* Uses Internet Protocol addresses to send data from one device to another across networks.
* **FinTech Example:** Routing a transaction request from Easypaisa servers in Pakistan to Visa servers in the United States.

---

#### **Layer Two: Data Link Layer 🔄**

* Handles local delivery of data using frames and device addresses (Media Access Control addresses).
* Provides error detection at the local network level.
* **FinTech Example:** An Automated Teller Machine connected securely to the bank’s switch, preventing Media Access Control address spoofing.

---

#### **Layer One: Physical Layer ⚡**

* Deals with the actual hardware, signals, and media used for transmission.
* Examples: copper wires, fiber optic cables, Wi-Fi radio waves.
* **FinTech Example:** An Automated Teller Machine going offline during a fiber optic cable cut or power outage.

---

✅ **Quick Recap Table**

| Layer                       | Function                            | FinTech Example                                                |
| --------------------------- | ----------------------------------- | -------------------------------------------------------------- |
| Layer Seven: Application 📱 | User-facing services                | Mobile banking app, payment gateway                            |
| Layer Six: Presentation 🎭  | Formatting, translation, encryption | Transport Layer Security encrypting card transactions          |
| Layer Five: Session 🗂️     | Connection management               | Secure login session                                           |
| Layer Four: Transport 📦    | Reliable delivery of messages       | Transmission Control Protocol ensuring payment packet delivery |
| Layer Three: Network 🌍     | Routing and addressing              | Transaction routed between servers in different countries      |
| Layer Two: Data Link 🔄     | Local delivery, device addressing   | Automated Teller Machine connected to switch                   |
| Layer One: Physical ⚡       | Hardware and signals                | Fiber optic link used for payments                             |

---

  * Answer: **Layer Seven: Application**.
* *“If a cable is cut at a data center, which OSI layer is impacted?”*

  * Answer: **Layer One: Physical**.

## **1.3 Transmission Control Protocol / Internet Protocol Model Overview (Four Layers) 📡**

The **Transmission Control Protocol / Internet Protocol (TCP IP) Model** is the **practical model** used in today’s Internet. It is simpler than the Open Systems Interconnection Model, with **four layers**.

---

### **Layer Four: Application Layer 📱**

* Provides services and interfaces that applications use.
* Includes protocols such as **Hypertext Transfer Protocol (HTTP, HTTPS)**, **Domain Name System (DNS)**, **File Transfer Protocol (FTP)**, and **Simple Mail Transfer Protocol (SMTP)**.
* **FinTech Example:** Mobile banking applications use **Hypertext Transfer Protocol Secure (HTTPS)** to transfer transaction data securely.

---

### **Layer Three: Transport Layer 📦**

* Ensures reliable delivery of data between applications.
* Uses **Transmission Control Protocol (TCP)** for reliable communication (retransmissions, acknowledgments).
* Uses **User Datagram Protocol (UDP)** for fast but less reliable communication.
* **FinTech Example:** Transmission Control Protocol guarantees that a payment confirmation packet from Visa or Mastercard arrives without loss.

---

### **Layer Two: Internet Layer 🌍**

* Responsible for addressing and routing data packets.
* Uses **Internet Protocol (IP)** for logical addressing and **Internet Control Message Protocol (ICMP)** for diagnostics.
* **FinTech Example:** An online wallet transaction is routed from Pakistan to an international server using Internet Protocol.

---

### **Layer One: Link Layer 🔗**

* Deals with physical addressing and delivery within a local network.
* Includes Ethernet, Wi-Fi, and Address Resolution Protocol (ARP).
* **FinTech Example:** A point-of-sale machine in a store connects to the payment server through Ethernet or Wi-Fi at this layer.

---

✅ **Summary Table: Transmission Control Protocol / Internet Protocol Model**

| Layer          | Functions                          | FinTech Example                                          |
| -------------- | ---------------------------------- | -------------------------------------------------------- |
| Application 📱 | Application services and protocols | Online banking app using HTTPS                           |
| Transport 📦   | Reliable or fast delivery          | Payment confirmation using Transmission Control Protocol |
| Internet 🌍    | Addressing and routing             | Wallet transaction routed using Internet Protocol        |
| Link 🔗        | Physical and local delivery        | Point-of-sale machine connecting via Ethernet            |

## **1.4 Comparison: Open Systems Interconnection Model vs Transmission Control Protocol / Internet Protocol Model ⚖️**

Both models describe **how data travels** through networks, but they differ in complexity and usage.

---

### **Structure**

* **Open Systems Interconnection Model:** Seven layers (detailed, theoretical).
* **Transmission Control Protocol / Internet Protocol Model:** Four layers (practical, used in the Internet).

---

### **Usage**

* **Open Systems Interconnection Model:** Mostly used for teaching, documentation, and troubleshooting.
* **Transmission Control Protocol / Internet Protocol Model:** Used daily in real-world Internet communications.

---

### **Strengths and Weaknesses**

| Aspect             | Open Systems Interconnection Model                                                          | Transmission Control Protocol / Internet Protocol Model                   |
| ------------------ | ------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------- |
| **Strengths**      | Clear separation of seven functions → easy to identify attacks and defenses.                | Simpler, directly maps to Internet protocols → practical in real use.     |
| **Weaknesses**     | Theoretical, rarely implemented exactly. Some layers (Session, Presentation) often overlap. | Less detailed, harder to map specific security controls to each function. |
| **Learning Value** | Excellent for understanding security at each layer.                                         | Excellent for understanding how data is actually transmitted online.      |

---

### **Analogy**

* The **Open Systems Interconnection Model** is like a **textbook manual 📖** with every detail explained.
* The **Transmission Control Protocol / Internet Protocol Model** is like a **real-world toolkit 🧰** that technicians actually use.

---

## **1.5 Importance of Networking Models in Financial Technology Systems 💳**

Networking models are critical in **Financial Technology (FinTech)** because they show **where attacks can occur and how to secure them**.

---

### **A) Mobile Wallets (Easypaisa, PayPal, JazzCash)**

* **Application Layer:** Mobile app authentication, secure Application Programming Interface (API).
* **Presentation Layer (in Open Systems Interconnection):** Encryption with Transport Layer Security (TLS) to protect personal identification numbers (PINs) and one-time passwords (OTPs).
* **Transport Layer:** Transmission Control Protocol ensures reliable packet delivery for transactions.
* **Internet Layer:** Internet Protocol routes the request from the user to the banking server.

---

### **B) Payment Gateways (Visa, Mastercard, Stripe)**

* **Application Layer:** Must follow Payment Card Industry Data Security Standard (PCI-DSS).
* **Presentation Layer:** Enforces HTTPS and modern cryptography.
* **Transport Layer:** Defends against Transmission Control Protocol flooding and session hijacking.
* **Internet Layer:** Uses redundancy and Virtual Private Networks to prevent denial of service.

---

### **C) Online Banking Systems**

* **Application Layer:** Protects against phishing, SQL injection, and cross-site scripting.
* **Presentation Layer:** Encrypts communication between customer and bank.
* **Transport Layer:** Ensures transactions are not lost during communication.
* **Network and Physical Layers:** Ensure data centers and Automated Teller Machines remain online through redundancy and backup power.

---

**Key Message:**
If **any layer fails**, it can:

* Leak sensitive customer information (Confidentiality breach 🔒).
* Corrupt transaction records (Integrity breach ✅).
* Make systems unavailable (Availability breach ⚡).

---

Three attack examples, **map them to a layer**:

* A phishing email leading to fake login → **Application Layer**.
* A Transmission Control Protocol SYN flooding attack → **Transport Layer**.
* A fiber optic cut causing ATM downtime → **Physical/Link Layer**.

# **2. OSI Model – Security at Each Layer 🔒**

---

## **2.1 Physical Layer (Layer One) ⚡**

### **Role**

* The foundation of all networking.
* Deals with the **actual hardware**: cables, fiber optics, network interface cards, wireless signals, and power supply.
* If this layer fails, no communication can happen at higher layers.

---

### **Threats**

1. **Cable Tapping** 🕵️

   * Hackers physically connect to cables and eavesdrop on data signals.
   * Example: Payment card data stolen by tapping into Automated Teller Machine (ATM) lines.

2. **Hardware Theft** 💻

   * Physical stealing of routers, servers, or laptops containing sensitive data.

3. **Jamming Attacks** 📡

   * Disrupting wireless signals, making services unavailable.
   * Example: Disabling Point-of-Sale (POS) terminals in stores by jamming Wi-Fi.

4. **Power Loss / Power Outages** 🔌

   * If electricity fails, servers and payment systems go offline.

---

### **Protections**

* **Closed-Circuit Television (CCTV)** 📹 to monitor equipment rooms.
* **Locks and Restricted Access** 🔑 to server rooms and ATM areas.
* **Redundant Cabling** 🔄 (multiple cable paths in case one is cut).
* **Uninterruptible Power Supply (UPS)** 🔋 and backup generators to prevent downtime.

---

### **FinTech Example**

* A fiber optic cut can bring down a **mobile wallet service** across a region.
* Redundancy ensures that transactions continue smoothly, especially during high demand like Eid shopping.

---

---

## **2.2 Data Link Layer (Layer Two) 🔄**

### **Role**

* Provides **local delivery** of data within the same network.
* Uses **frames** and **Media Access Control (MAC) addresses**.
* Works closely with switches and network interface cards.

---

### **Threats**

1. **MAC Spoofing** 🎭

   * Attacker forges a fake Media Access Control address to impersonate another device.
   * Can bypass access control in payment servers.

2. **Address Resolution Protocol (ARP) Poisoning** 🧪

   * Attacker sends false ARP messages, tricking devices into sending traffic through the attacker.
   * Example: Intercepting payment gateway traffic.

3. **Virtual Local Area Network (VLAN) Hopping** 🛠️

   * Attacker breaks out of their assigned VLAN and gains access to restricted network segments.
   * Example: Unauthorized access to bank internal servers.

---

### **Protections**

* **Port Security** 🚪 → Limits which Media Access Control addresses are allowed on each switch port.
* **Dynamic ARP Inspection** 🔍 → Detects and blocks malicious ARP replies.
* **VLAN Segmentation** ✂️ → Separates customer networks from internal banking systems.

---

### **FinTech Example**

* Hackers inside a company could use ARP poisoning to **intercept card transaction data**.
* VLAN segmentation ensures that **customer payment traffic** never mixes with **bank’s internal admin traffic**.

---

---

## **2.3 Network Layer (Layer Three) 🌍**

### **Role**

* Responsible for **routing** data between different networks.
* Uses logical addresses (Internet Protocol addresses) to send data across the globe.
* This is where the Internet Protocol operates.

---

### **Threats**

1. **Internet Protocol Spoofing** 👻

   * Attacker sends packets with a forged source Internet Protocol address.
   * Used in **Denial of Service** and bypassing security filters.

2. **Route Hijacking** 🚦

   * Attacker advertises false routes, redirecting traffic through malicious servers.
   * Example: Online banking traffic being rerouted through a hacker-controlled server.

3. **Internet Control Message Protocol (ICMP) Floods** 🌊

   * Attacker floods the target with Internet Control Message Protocol packets (like ping), overwhelming it.
   * Example: Disabling payment gateways through excessive traffic.

---

### **Protections**

* **Firewalls** 🔥 → Block malicious or unauthorized traffic at the Internet Protocol level.
* **Access Control Lists (ACLs)** 📑 → Define which Internet Protocol addresses and traffic are allowed.
* **Virtual Private Networks (VPNs)** 🔐 → Encrypt data between client and server to avoid interception.
* **Internet Protocol Security (IPSec)** 🔒 → Provides end-to-end encryption and authentication at the Internet Protocol level.

---

### **FinTech Example**

* Without a Virtual Private Network, customer **transaction data could be intercepted** while traveling between the mobile wallet and the bank server.
* Firewalls protect **payment APIs** from malicious packets coming from unknown Internet Protocol addresses.

---




---

## **2.4 Transport Layer (Layer Four) 📦**

### **Role**

* Responsible for the **end-to-end delivery** of data between two devices.
* Two main protocols work here:

  * **Transmission Control Protocol (TCP):** Reliable communication (acknowledgements, retransmissions).
  * **User Datagram Protocol (UDP):** Faster communication, but not guaranteed to be reliable.

**FinTech Example:**

* Transmission Control Protocol makes sure that a bank transaction packet is delivered completely.
* User Datagram Protocol may be used in real-time financial applications that require speed over reliability.

---

### **Threats**

1. **SYN Flood Attack** 🌊

   * Attacker sends thousands of half-open Transmission Control Protocol connection requests.
   * Overwhelms the server so it cannot serve real customers.

2. **Transmission Control Protocol Session Hijacking** 🎭

   * Attacker takes over an existing connection between a customer and a bank server.
   * Can be used to **inject fake financial transactions**.

3. **User Datagram Protocol Abuse** 🚨

   * User Datagram Protocol can be misused for flooding attacks or amplification attacks.

---

### **Protections**

* **SYN Cookies** 🍪 → Prevent servers from being overwhelmed by half-open connections.
* **Rate Limiting** ⚖️ → Restricts the number of requests per user or Internet Protocol address.
* **Transport Layer Security (TLS) over TCP** 🔒 → Encrypts traffic to stop hijacking and tampering.

---

---

## **2.5 Session Layer (Layer Five) 🗂️**

### **Role**

* Establishes, manages, and ends communication **sessions** between devices.
* Keeps track of login sessions or online transactions.

**FinTech Example:**

* When a customer logs into a mobile wallet and stays logged in, the session layer keeps this communication active until logout.

---

### **Threats**

1. **Session Hijacking** 🥷

   * Attacker steals the session identifier (like a cookie or token).
   * Can impersonate the user to steal money.

2. **Cookie Theft** 🍪

   * If cookies are not protected, they can be stolen and reused to access accounts.

---

### **Protections**

* **Session Timeouts** ⏲️ → Automatic logout after inactivity.
* **Token-Based Authentication** 🔑 → Strong, unique session identifiers that cannot be guessed.
* **Secure Cookies** 🔒 → Cookies are encrypted and tied to secure connections only.

---

---

## **2.6 Presentation Layer (Layer Six) 🎭**

### **Role**

* Handles **data translation, formatting, compression, and encryption**.
* Ensures that data is in a readable format for applications.

**FinTech Example:**

* Transport Layer Security (TLS) ensures that customer Personal Identification Number (PIN) or card details are encrypted before being sent.

---

### **Threats**

1. **SSL Stripping** 🛑

   * Attacker forces a downgrade from secure HTTPS to plain HTTP.
   * Data travels without encryption.

2. **Data Manipulation** ✍️

   * Altering encrypted or compressed data during transmission.

3. **Malware Injection** 🦠

   * Malicious code inserted during translation or formatting processes.

---

### **Protections**

* **Transport Layer Security (TLS) / Secure Sockets Layer (SSL)** 🔐 → Encrypts communication.
* **Data Validation** ✅ → Ensures only valid, expected data formats are accepted.
* **Secure Application Programming Interfaces (APIs)** 🔒 → Prevent attackers from exploiting weak data exchanges.

---

---

## **2.7 Application Layer (Layer Seven) 📱**

### **Role**

* Closest to the **end-user**.
* Provides services like email, web browsing, file transfers, mobile banking apps, and payment gateways.

**FinTech Example:**

* JazzCash mobile app, PayPal online portal, and banking websites all operate here.

---

### **Threats**

1. **Structured Query Language Injection (SQL Injection)** 💉

   * Attacker sends malicious commands to manipulate the bank’s database.

2. **Cross-Site Scripting (XSS)** 🕷️

   * Attacker injects malicious scripts into banking websites to steal login details.

3. **Phishing Attacks** 🎣

   * Fake websites or emails tricking users to enter credentials.

---

### **Protections**

* **Web Application Firewall (WAF)** 🛡️ → Blocks malicious requests like SQL injection or cross-site scripting.
* **Secure Coding Practices** 👩‍💻 → Writing safe code, validating inputs, sanitizing data.
* **Patch Management** 🔧 → Regular updates to remove known vulnerabilities.

# **Scenario: A Mobile Wallet Transaction in Action 💳📱**

---

### **Step 1: Customer Initiates a Transaction (Application Layer 📱)**

* A customer in Islamabad opens the **JazzCash mobile wallet** and enters:

  * Amount: PKR 5,000
  * Recipient: Friend’s mobile number

**Protocol in use:** Hypertext Transfer Protocol Secure (HTTPS).

* Ensures the request is **encrypted** with Transport Layer Security.

**Possible Attack 🎣:** Phishing → customer could be tricked into entering details on a fake wallet app.
**Defense 🛡️:** Secure coding, app store verification, and Web Application Firewalls.

---

### **Step 2: Data Formatting and Encryption (Presentation Layer 🎭)**

* The wallet app encrypts the **Personal Identification Number (PIN)** and other sensitive details.

**Protocol in use:** Transport Layer Security (TLS).

**Possible Attack 🛑:** SSL Stripping → attacker forces downgrade to HTTP, exposing PIN.
**Defense 🛡️:** Enforce HTTPS only, enable HSTS (HTTP Strict Transport Security).

---

### **Step 3: Secure Session Established (Session Layer 🗂️)**

* A session token is generated after the customer login.
* This keeps the customer logged in until they log out.

**Protocol in use:** Token-based authentication within HTTPS.

**Possible Attack 🎭:** Session hijacking → attacker steals session token.
**Defense 🛡️:** Session timeouts, secure cookies, token encryption.

---

### **Step 4: Reliable Delivery of Transaction Data (Transport Layer 📦)**

* Transaction data is broken into **segments** by Transmission Control Protocol (TCP).
* TCP ensures no packets are lost and data arrives in the right order.

**Protocol in use:** Transmission Control Protocol (TCP).

**Possible Attack 🌊:** SYN flood → attacker sends thousands of fake TCP requests to block real transactions.
**Defense 🛡️:** SYN cookies, rate limiting, load balancers.

---

### **Step 5: Routing Across Networks (Network Layer 🌍)**

* The packet leaves the customer’s phone and travels through routers.
* Internet Protocol (IP) provides addressing to send the data to the JazzCash server.

**Protocol in use:** Internet Protocol (IP).

**Possible Attack 👻:** IP spoofing → hacker forges source IP to bypass firewall.
**Defense 🛡️:** Access Control Lists, firewalls, Internet Protocol Security (IPSec).

---

### **Step 6: Local Network Delivery (Data Link Layer 🔄)**

* On the way, the packet passes through Wi-Fi at the customer’s home or 4G towers.
* Media Access Control (MAC) addresses identify devices locally.

**Protocols in use:** Ethernet, Wi-Fi, Address Resolution Protocol (ARP).

**Possible Attack 🧪:** ARP poisoning → attacker redirects traffic to their device.
**Defense 🛡️:** Dynamic ARP inspection, VLAN segmentation.

---

### **Step 7: Physical Transmission (Physical Layer ⚡)**

* The transaction travels as **electrical signals in cables** or **radio waves** in Wi-Fi/4G.
* Eventually reaches the JazzCash payment server.

**Protocols in use:** Fiber optics, copper cables, Wi-Fi signals.

**Possible Attack 🔌:** Fiber cable cut or Wi-Fi jamming disrupts service.
**Defense 🛡️:** Redundant cabling, Uninterruptible Power Supply (UPS), backup connectivity.

---

### **Step 8: Processing at the Payment Gateway 💳**

* JazzCash verifies the customer PIN, balance, and destination number.
* Transaction is routed securely to **Visa/Mastercard servers** if it involves a card.

**Protocols in use:** HTTPS + TLS over TCP.

**Possible Attack 🌊:** Distributed Denial of Service (DDoS) floods gateway with fake traffic.
**Defense 🛡️:** DDoS mitigation services, Web Application Firewalls, load balancing.

---

### **Step 9: Confirmation Sent Back to Customer ✅**

* The confirmation packet travels back through all the layers in reverse order.
* The customer sees **“Transaction Successful – PKR 5,000 sent.”**

**Protocols in use:** TCP (reliable delivery) + HTTPS (encrypted).

**Possible Attack 🕵️:** Man-in-the-Middle → attacker intercepts response and injects fake “success” message.
**Defense 🛡️:** TLS encryption, certificate pinning in apps.

---

# **Summary Flow 🌐**

1. **Application Layer:** HTTPS sends wallet request → 🛑 Phishing
2. **Presentation Layer:** TLS encrypts PIN → 🛑 SSL stripping
3. **Session Layer:** Session token manages login → 🛑 Session hijack
4. **Transport Layer:** TCP segments transaction → 🛑 SYN flood
5. **Network Layer:** IP routes packets → 🛑 IP spoofing
6. **Data Link Layer:** ARP resolves addresses → 🛑 ARP poisoning
7. **Physical Layer:** Data travels as signals → 🛑 Cable cut / jamming

# **4. Wireshark Introduction and Demo 🔍**

---

## **4.1 What is Wireshark? 🐬**

* **Wireshark** is the world’s most popular **network protocol analyzer**.
* It captures **live network traffic** and shows it in detail (every packet).
* Think of it as a **microscope 🔬** for network activity.

**Capabilities:**

* See all packets flowing across a network in real time.
* Understand **which protocols are being used** (HTTP, HTTPS, DNS, TCP, UDP, etc.).
* Detect suspicious or malicious traffic.

**FinTech Example:**

* Imagine JazzCash traffic being sniffed on a public Wi-Fi hotspot.
* With Wireshark, you can observe **unencrypted HTTP traffic** vs **encrypted HTTPS traffic**, and show students how attacks like **man-in-the-middle** become possible if encryption is weak.

---

## **4.2 Why Cybersecurity Professionals Use It 🛡️**

Cybersecurity experts use Wireshark for many reasons:

1. **Troubleshooting** 🛠️

   * Debugging why an online banking application is slow or not responding.
   * Example: Identifying packet loss in TCP transactions.

2. **Incident Response** 🚨

   * Analyzing suspicious activity like ARP poisoning or DNS spoofing.
   * Example: Detecting if customer banking traffic is redirected to a fake server.

3. **Protocol Analysis** 📊

   * Understanding how different protocols work (HTTP, HTTPS, DNS, etc.).
   * Essential for students to learn **where attacks can happen** in the TCP/IP stack.

4. **Attack Detection** 🕵️

   * Spotting port scans, brute force attempts, or malware communication.
   * Example: Spotting unusual DNS queries from a compromised FinTech server.

👉 In short, **Wireshark = visibility**. If you can see the packets, you can understand, protect, and respond.

---

## **4.3 Installing and Running Wireshark (Kali Linux Demo Guide) 🐧**

Wireshark comes **pre-installed in Kali Linux**, so setup is easy.

---

### **Step 1: Verify Installation**

Open the Kali Linux terminal and type:

```bash
wireshark --version
```

* If it shows the version → Wireshark is installed.
* If not installed:

```bash
sudo apt update
sudo apt install wireshark -y
```

---

### **Step 2: Launch Wireshark**

* From terminal:

```bash
sudo wireshark
```

* Or from the Kali Linux application menu: **Applications → Sniffing & Spoofing → Wireshark**

👉 Note: Running as `sudo` is required to capture packets at the system level.

---

### **Step 3: Select a Network Interface**

* Wireshark will show all available interfaces (e.g., `eth0` for Ethernet, `wlan0` for Wi-Fi).
* Select `wlan0` (wireless) if using Wi-Fi, or `eth0` if connected via cable.
* Click **Start Capture** ▶️.

---

### **Step 4: Capture Traffic**

* Once capture starts, packets will flow in real time.
* Each packet shows:

  * Time ⏱️
  * Source and Destination IP 🌍
  * Protocol used 📡
  * Packet length 📦
  * Info field (details about the transaction)

---

### **Step 5: Apply Filters**

Wireshark captures **everything**, but we can filter:

* **HTTP traffic only**

```http
http
```

* **HTTPS traffic only**

```http
tls
```

* **DNS queries only**

```http
dns
```

* **Show only traffic to a specific IP (e.g., PayPal)**

```http
ip.addr == 64.4.250.36
```

👉 Activity: Compare an HTTP login vs an HTTPS login → Check how credentials are exposed in plain HTTP but encrypted in HTTPS.

---

### **Step 6: Stop and Save Capture**

* Click **Stop Capture** ⏹️.
* Save file as `.pcap` (Packet Capture file).
* These files can be analyzed later or shared for exercises.

# **4.4 Capturing Packets 🎥**

---

### **A) Setting up a Capture ⚡**

1. Open Wireshark in Kali Linux:

   ```bash
   sudo wireshark
   ```
2. Choose the **network interface**:

   * `eth0` → Wired Ethernet
   * `wlan0` → Wireless Wi-Fi
   * `lo` → Loopback (local traffic)
3. Click **Start Capturing** ▶️.

   * Packets will begin scrolling in real time.
   * Each row = one packet.

👉 Start browsing the web (open Easypaisa or PayPal in a browser) so students see **real traffic** appear instantly.

---

* In the “Protocol” column of Wireshark, students will see:

  * **TCP** 📦 → Reliable transport
  * **UDP** ⚡ → Fast, connectionless transport
  * **HTTP** 🌐 → Plain web requests
  * **TLS/HTTPS** 🔒 → Secure web traffic
  * **DNS** 🌍 → Domain name lookups
  * **ARP** 🔄 → Address Resolution Protocol

# **4.5 Basic Analysis 🔍**

---

### **A) Viewing Packet Details 🗂️**

1. Click on any packet → Wireshark expands into **three panes**:

   * **Top Pane:** List of all captured packets.
   * **Middle Pane:** Decoded protocol layers (Ethernet → IP → TCP → HTTP).
   * **Bottom Pane:** Raw binary (hexadecimal view).
2. Expand each layer to see:

   * Source & Destination IP addresses 🌍
   * Port numbers (e.g., 443 for HTTPS, 80 for HTTP) 🔢
   * Flags (SYN, ACK) in TCP handshakes

👉 **Demo:** Try a **3-way handshake** (SYN → SYN-ACK → ACK) when a new session starts.

---

### **B) Filtering HTTP vs HTTPS Traffic 🔒 vs 🌐**

* **HTTP only:**

  ```http
  http
  ```

  * Students can open a non-secure site (if available) → see GET/POST requests in plain text.
  * Username and password might appear visible.

* **HTTPS (TLS) only:**

  ```http
  tls
  ```

  * Students will see encrypted packets.
  * No credentials are visible → proving why HTTPS is essential for FinTech apps.

👉 **FinTech Example:**

* Compare:

  * HTTP login page of a test site → credentials visible.
  * HTTPS login page of PayPal → credentials encrypted.
* *“Which one would you trust for your money?”*

---

### **C) Identifying Anomalies ⚠️**

When analyzing traffic, look for suspicious patterns:

1. **Unusual Protocol Use**

   * Example: A FinTech app using plain HTTP instead of HTTPS → dangerous.

2. **Strange DNS Requests** 🌍

   * Repeated requests to unknown domains could indicate malware or data exfiltration.

3. **Flood of Packets** 🌊

   * Large number of SYN packets = possible SYN flood attack.

4. **Unexpected Ports** 🔢

   * Normal web traffic uses port 443 (HTTPS).
   * If you see traffic on random ports → could be tunneling or malicious activity.

👉 **FinTech Example:**

* If attackers poison DNS, instead of `paypal.com → correct IP`, it resolves to a **fake IP**.
* Wireshark helps analysts spot this discrepancy.

# **4.7 Ethical Considerations in Packet Sniffing ⚖️**

Wireshark is a **powerful but dangerous tool**. Students must understand ethical use.

---

### **A) Legal Boundaries ⚖️**

* Capturing traffic on **networks you don’t own** or without permission is **illegal**.
* It can violate privacy laws and result in criminal charges.
* Example: Capturing other people’s online banking data in a coffee shop Wi-Fi is unlawful.

---

### **B) Privacy Concerns 👀**

* Packet sniffing can expose usernames, passwords, emails, and financial information.
* In FinTech, this could mean **identity theft** or **fraud**.
* Students must **never misuse Wireshark** for spying on others.

---

### **C) Responsible Use ✅**

* Use Wireshark only in:

  * Lab environments.
  * Your own computer or test networks.
  * With explicit permission (e.g., penetration testing).
* For educational purposes → focus on comparing encrypted vs unencrypted traffic.

---

### **D) Professional Ethics 👩‍💻**

* Cybersecurity professionals follow a **“do no harm” principle**.
* Tools like Wireshark should be used for **defense, learning, and auditing**, never for malicious gain.

---

👉 **Class Discussion:**

* *“If you saw someone capturing traffic in your university Wi-Fi, what ethical and legal issues do you think apply?”*
* Recognizing **consent, legality, and privacy rights**.

## 🧠 **1. Introduction to Nmap**

### ❓ What is Nmap?

**Nmap (Network Mapper)** is a free and open-source tool used for network discovery and security auditing. It allows users to find:

* Live devices on a network
* Open ports
* Services running on ports
* Operating systems
* Vulnerabilities

### 🌍 Real-World Use Cases

* **Network administrators** use Nmap to monitor infrastructure.
* **Cybersecurity professionals** use it for penetration testing and audits.
* **Ethical hackers** rely on it during reconnaissance in red teaming.

### ⚖️ Legal & Ethical Considerations

* **NEVER** scan systems without permission.
* Only scan **your own machines**, **test environments**, or **allowed domains** like `scanme.nmap.org`.

### 🧰 Installation (if needed)

* **Linux**: `sudo apt install nmap`
* **Windows**: Download from [https://nmap.org/download.html](https://nmap.org/download.html)
* **Kali Linux**: Already pre-installed

---

## 🔍 **2. Host Discovery (Ping Scans)**

These scans help you discover which hosts are alive or reachable.

### ✅ **Command 1: Basic Ping Scan**

```bash
nmap -sn 192.168.1.1
```

➡️ Sends ICMP echo request to a **single host** to check if it's online.

### ✅ **Command 2: Subnet Ping Scan**

```bash
nmap -sn 192.168.1.0/24
```

➡️ Sends ICMP requests to **all hosts in a subnet** (e.g., 192.168.1.1 to 192.168.1.254).

### ✅ **Command 3: Skip Ping (Treat All Hosts as Alive)**

```bash
nmap -Pn scanme.nmap.org
```

➡️ Useful if **ping is blocked** by a firewall (e.g., cloud or hardened systems). Treats host as alive without checking first.

## 🚪 **3. Port Scanning**

Port scanning helps you find **which services are running on a target**, based on open ports.

---

### ✅ **Command 1: Default Port Scan**

```bash
nmap scanme.nmap.org
```

➡️ Scans the **top 1000 TCP ports** of the target using the default settings.

---

### ✅ **Command 2: Scan Specific Ports**

```bash
nmap -p 22,80,443 scanme.nmap.org
```

➡️ Scans **only the listed ports** (SSH, HTTP, HTTPS).

---

### ✅ **Command 3: Scan All 65535 Ports**

```bash
nmap -p- scanme.nmap.org
```

➡️ Scans **all TCP ports (0–65535)** on the target. Thorough but slow.

---

### ✅ **Command 4: Scan Top N Ports**

```bash
nmap --top-ports 100 scanme.nmap.org
```

➡️ Scans the **top 100 most commonly used ports** (faster than full scan).

---

### ✅ **Command 5: Scan IP Range**

```bash
nmap 192.168.1.1-10
```

➡️ Scans **multiple IPs** from `.1` to `.10`.

---

### ✅ **Command 6: Scan from a List of IPs**

```bash
nmap -iL targets.txt
```

➡️ Scans all IP addresses listed in a file (`targets.txt`).

## ⚙️ **4. Scan Techniques**

Nmap supports different scanning techniques. Some are stealthy, others are thorough.

---

### ✅ **Command 1: TCP Connect Scan**

```bash
nmap -sT scanme.nmap.org
```

➡️ Full TCP **3-way handshake**. Used when stealth is not needed or SYN is blocked.

---

### ✅ **Command 2: TCP SYN Scan (Stealth)**

```bash
nmap -sS scanme.nmap.org
```

➡️ Sends **SYN packets** only (half-open scan). **Faster and stealthier**.

---

### ✅ **Command 3: UDP Scan**

```bash
nmap -sU scanme.nmap.org
```

➡️ Scans **UDP ports**, useful for services like DNS, SNMP. **Slower** than TCP scans.

---

### ✅ **Command 4: TCP + UDP Scan**

```bash
nmap -sSU scanme.nmap.org
```

➡️ Scans both **TCP and UDP ports** together (can be very slow).

---

### ✅ **Command 5: Aggressive Scan**

```bash
nmap -A scanme.nmap.org
```

➡️ Combines:

* OS detection (`-O`)
* Version detection (`-sV`)
* Script scanning (`-sC`)
* Traceroute

⚠️ Very detailed, but **noisy and slow**.

---

### ✅ **Command 6: Timing Templates**

```bash
nmap -T4 scanme.nmap.org
```

➡️ Controls scan **speed vs. stealth**:

* `-T0`: Paranoid (very slow, stealthy)
* `-T3`: Normal (default)
* `-T4`: Fast
* `-T5`: Insane (very fast, more detectable)

### ✅ **Command 1: Detect Services and Versions**

➡️ Shows what services (e.g., SSH, HTTP) and versions (e.g., OpenSSH 7.9) are running on the open ports.

### ✅ **Command 2: Combine with Specific Ports**

```bash
nmap -sV -p 22,80 scanme.nmap.org
```

➡️ Detects services on **selected ports only**.

### ✅ **Command 3: Combine with OS Detection**

```bash
nmap -sV -O scanme.nmap.org
```

➡️ Runs **both version and OS detection**.

* Vulnerability scanning

## 🔒 **8. Firewall Evasion and Spoofing Techniques**

Sometimes, **firewalls or intrusion detection systems (IDS)** block or log regular Nmap scans. These options help you **evade detection**, **bypass filters**, or **mask your identity** — useful for ethical hacking **only with permission**.

---

### ⚠️ Use these techniques responsibly — they may **trigger alerts or legal violations** if used without authorization.

---

### ✅ **Command 1: Use Decoy IPs (Hide Your IP)**

```bash
nmap -D RND:5 scanme.nmap.org
```

➡️ Sends fake packets from **5 random decoy IPs** to confuse intrusion detection systems.

---

### ✅ **Command 2: Fragment Packets**

```bash
nmap -f scanme.nmap.org
```

➡️ Splits packets into smaller pieces, making it **harder for firewalls to detect scans**.

---

### ✅ **Command 3: Spoof Source Port**

```bash
nmap --source-port 53 scanme.nmap.org
```

➡️ Spoofs source port as **53 (DNS)** — some firewalls allow DNS traffic through.

---

### ✅ **Command 4: Scan with Increased Speed**

```bash
nmap -T4 scanme.nmap.org
```

➡️ Increases scan speed. Timing values:

* `-T0`: Paranoid (slowest, stealthy)
* `-T3`: Default
* `-T4`: Faster
* `-T5`: Insane (very fast, easy to detect)

---

### ✅ **Command 5: Combine Speed and Stealth**

```bash
nmap -sS -T4 scanme.nmap.org
```

➡️ Fast and stealthy **SYN scan** with **timing template**.

---

🧠 Tip: These methods **do not guarantee full stealth** — skilled network admins will still detect scans using proper tools.
